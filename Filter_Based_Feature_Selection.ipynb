## **Filter-Based Feature Selection Techniques**

Filter-based feature selection techniques are methods used in machine learning to select a subset of features from a larger set based on statistical measures. These methods operate independently of any specific machine learning algorithm and are applied before the actual learning process begins. They are called "filter" methods because they evaluate features independently and filter out those that do not meet certain criteria.

### What it does
The primary goal of filter-based feature selection techniques is to identify and retain the most informative features while discarding redundant or irrelevant ones. By scoring each feature independently and selecting a subset based on these scores, these methods aim to improve model performance, reduce overfitting, and enhance interpretability.

### How it works
Filter-based feature selection techniques work by evaluating each feature independently of the others, using statistical measures to assess its relevance to the target variable. Here's a brief overview of how it works:

1. **Feature Scoring**: Different statistical measures can be used to score features. For example:
   - Correlation: Measures the linear relationship between each feature and the target variable.
   - Mutual Information: Captures the amount of information one variable provides about another.
   - ANOVA: Tests for significant differences in means across multiple groups.
  
2. **Ranking Features**: Once each feature is scored, they are ranked in descending order based on their scores. Features with higher scores are considered more informative.

3. **Selecting Subset**: Finally, a subset of features is selected based on the ranking. This subset can be determined by specifying a fixed number of features to keep (top k), setting a threshold score, or using other criteria.

### Notable Features/Functionality
- **Independence**: Filter methods do not rely on any specific machine learning algorithm and can be applied universally to different types of datasets and models.
- **Efficiency**: Since filter methods evaluate features independently, they are computationally efficient, making them suitable for large datasets with high-dimensional feature spaces.
- **Interpretability**: By selecting features based on their individual scores, filter methods often result in more interpretable models, as only the most relevant features are retained.

### Examples and Use Cases
- **Example**: In a dataset containing various features related to customer demographics and purchasing behavior, filter-based feature selection can help identify the most influential factors in predicting customer churn.
- **Use Case**: In image classification tasks, where each pixel can be considered a feature, filter methods can be used to select a subset of the most discriminative features (pixels) while reducing computational complexity.
  
Filter-based feature selection techniques play a crucial role in preprocessing pipelines, helping to improve model performance, reduce overfitting, and enhance interpretability across various machine learning tasks and domains.

### **VIF for Feature Selection**

VIF (Variance Inflation Factor) is a statistical measure used in regression analysis to detect multicollinearity among predictor variables. Multicollinearity occurs when independent variables in a regression model are highly correlated with each other, which can lead to unreliable and unstable estimates of the coefficients.

#### Structure:

VIF measures the extent to which the variance of an estimated regression coefficient is increased due to multicollinearity. It is calculated separately for each predictor variable in the model.

#### Formula:

The formula to calculate VIF for a predictor variable \( X_i \) is:

 ##  ```VIF_X_i  = 1/(1 - R_X_i**2)```
image.png

Where:
- R_X_i**2 is the coefficient of determination when  X_i is regressed against all other predictor variables.

#### Steps:

1. **Fit a regression model**: Begin by fitting a multiple regression model with all the predictor variables.
2. **Calculate VIF for each variable**: For each predictor variable X_i, calculate the VIF using the formula mentioned above.
3. **Assess VIF values**: Higher VIF values indicate stronger multicollinearity. Typically, a VIF greater than 5 or 10 is considered problematic, although the threshold may vary depending on the context.
4. **Identify problematic variables**: Variables with high VIF values are considered multicollinear and may need to be addressed.
5. **Iterate if necessary**: If multicollinearity is detected, consider removing or transforming the correlated variables and recalculate VIF values to ensure that multicollinearity is sufficiently addressed.

#### Assumptions:

- VIF assumes a linear relationship between variables.
- It assumes that multicollinearity is the only issue affecting the model's stability and reliability.

#### How it Works:

VIF measures how much the variance of the estimated regression coefficient of a particular predictor variable is increased due to multicollinearity with other predictors. A high VIF value indicates that the variance of the coefficient is inflated, suggesting that the variable is highly correlated with other predictors in the model.

#### Notable Features/Functionality:

- **Detecting Multicollinearity**: VIF is primarily used to detect multicollinearity among predictor variables in regression models.
- **Quantitative Assessment**: VIF provides a quantitative measure of the extent of multicollinearity, allowing researchers to identify problematic variables.
- **Model Optimization**: By identifying and addressing multicollinearity, VIF helps in improving the stability and reliability of regression models.

#### Examples and Use Cases:

- **Economics**: In economic models where several factors might be correlated, VIF helps in identifying which variables might be causing multicollinearity issues.
- **Healthcare**: In medical research, where multiple factors might influence a health outcome, VIF helps in selecting the most relevant and independent predictors.
- **Marketing**: In market research, understanding the interrelationships between various marketing strategies or factors can be crucial, and VIF aids in identifying redundant variables.
- **Finance**: When building financial models, multicollinearity among predictor variables can lead to erroneous estimates, and VIF helps in identifying and addressing such issues.

VIF serves as a valuable tool in regression analysis for ensuring the reliability and stability of the estimated coefficients by identifying and addressing multicollinearity issues among predictor variables.

## Points to Consider While Using VIF

1. **Understanding of Multicollinearity**: Have a clear understanding of multicollinearity and its implications in regression analysis. Multicollinearity occurs when independent variables in a regression model are highly correlated with each other, which can lead to unstable coefficient estimates and difficulties in interpreting the model.

2. **Selection of Variables**: Ensure that you have selected appropriate variables for your regression model. Including variables that are conceptually related or highly correlated with each other can increase multicollinearity.

3. **Threshold Value**: Decide on a threshold value for VIF beyond which multicollinearity is considered problematic. Commonly used threshold values for VIF range from 5 to 10, but the specific threshold may vary depending on the context and the degree of tolerance for multicollinearity.

4. **Interpretation**: Understand how to interpret VIF values. A VIF value greater than 1 indicates the presence of multicollinearity, with higher values indicating stronger multicollinearity. A VIF of 1 indicates no multicollinearity.

5. **Assumption Checking**: Use VIF as a diagnostic tool to check for multicollinearity after fitting a regression model. It is important to assess multicollinearity before drawing conclusions about the regression coefficients and their significance.

6. **Remedial Actions**: If multicollinearity is detected, consider taking remedial actions such as removing highly correlated variables, combining correlated variables, or using dimensionality reduction techniques like principal component analysis (PCA).

7. **Contextual Considerations**: Consider the specific context of your analysis and the potential consequences of multicollinearity. In some cases, multicollinearity may not be a major concern if the primary goal is prediction rather than inference.

8. **Documentation**: Document the process of assessing multicollinearity using VIF, including the threshold chosen, the interpretation of results, and any remedial actions taken. This helps ensure transparency and reproducibility in your analysis.


## **Feature Selection using Variance Threshold**

The mathematical formula for calculating the variance threshold is relatively straightforward. Given a dataset with $n$ samples and $p$ features, the variance of each feature $j$ is computed. If the variance of a feature is below a specified threshold $\tau$, the feature is considered low-variance and may be removed from the dataset.

Here's the mathematical formula for the variance of a feature $j$:

$$ \text{Var}(X_j) = \frac{1}{n} \sum_{i=1}^{n} (x_{ij} - \bar{x}_j)^2 $$

where:
- $x_{ij}$ is the value of feature $j$ for sample $i$.
- $\bar{x}_j$ is the mean of feature $j$ across all samples.
- $n$ is the total number of samples.

After computing the variance for each feature, if the variance of a feature $j$ is below the threshold $\tau$, that feature may be removed from the dataset.

The threshold $\tau$ is typically chosen based on domain knowledge, experimentation, or as a hyperparameter in machine learning models. It represents the minimum acceptable variance for features to be retained in the dataset. Features with variance below the threshold are considered to have low predictive value or to contribute little information to the model, and thus may be discarded to simplify the dataset and improve computational efficiency or model performance.

1. **Ignores Target Variable:**
   - Variance Threshold is a univariate method, evaluating each feature independently.
   - It doesn't consider the relationship between features and the target variable.
   - May retain irrelevant features with high variance but no link to the target or discard potentially useful features with low variance but strong target relationship.

2. **Ignores Feature Interactions:**
   - Doesn't account for interactions between features.
   - A feature with low variance might become informative when combined with another feature.

3. **Sensitive to Data Scaling:**
   - The method is sensitive to data scale.
   - Features not on the same scale can yield higher variance for those with larger values.
   - It's essential to standardize features before applying Variance Threshold to mitigate this issue.

4. **Arbitrary Threshold Value:**
   - Users must define what constitutes "low" variance.
   - The threshold isn't universally applicable and may vary between datasets.
   - Determining an optimal threshold requires experimentation and domain expertise.

### How It Works

1. **Evaluation of Features:**
   - Variance Threshold evaluates each feature independently.
   - It calculates the variance of each feature across samples.
   - Features with variance below the specified threshold are marked for removal.

2. **Threshold Determination:**
   - Users define a threshold value indicating the minimum acceptable variance.
   - Features failing to meet this threshold are deemed uninformative and are discarded.

3. **Filtering Process:**
   - Features with variance below the threshold are removed from the dataset.
   - The resulting dataset contains only features with variance above the specified threshold.

### Notable Features

- **Univariate Method:** Considers features individually without regard to their relationships.
- **Simplicity:** Straightforward implementation requiring minimal hyperparameter tuning.
- **Preprocessing Requirement:** Sensitive to data scaling, necessitating feature standardization.
- **User-Defined Threshold:** Users have control over the threshold value, but determining an optimal threshold can be challenging.

### Example Use Cases

1. **Dimensionality Reduction:** Used as a preliminary step to reduce the number of features in high-dimensional datasets.
2. **Noise Reduction:** Eliminates features with low variance, which may be noise or uninformative for modeling.
3. **Data Exploration:** Provides insights into feature variability, aiding in understanding dataset characteristics.

### Conclusion

While Variance Threshold offers simplicity and ease of implementation, its limitations, such as ignorance of feature interactions and sensitivity to data scaling, should be considered. It serves as a useful tool in feature selection pipelines but should be complemented with other methods for comprehensive feature analysis and model building.

## **ANOVA for Feature Selection**

### Overview
- ANOVA (Analysis of Variance) is a statistical method used for feature selection in machine learning.
- It is particularly useful when dealing with numerical input variables and a categorical target variable, but it can also be adapted for other scenarios.

### Steps:
1. **Compute Group Means:** Calculate the mean of the numerical input variable for each category of the categorical target variable.

2. **Calculate Between-group Variance:** Compute the variance between the group means, also known as the "between-group" variance.

3. **Calculate Within-group Variance:** Compute the variance within each group, also known as the "within-group" variance.

4. **Compute F-statistic:** Calculate the F-statistic, which is the ratio of between-group variance to within-group variance.

5. **Determine Significance:** Compare the F-statistic with a critical value from the F-distribution or calculate the p-value associated with the statistic. If the F-statistic exceeds the critical value or the p-value is below a chosen significance level (e.g., 0.05), the feature is considered significant.

6. **Select Features:** Based on the significance test results, select the features with the highest F-statistics or lowest p-values as relevant features for the model.

### Advantages:
- Suitable for numerical input variables and a categorical target variable.
- Identifies features with significant differences among different categories of the target variable.
- Provides a quantitative measure of feature importance.

### Disadvantages:
- Assumes normality and homogeneity of variances, which may not always hold true.
- Sensitive to outliers.
- Limited to categorical target variables.


###**Assumptions and Limitations of ANOVA (Analysis of Variance)**

### 1. Assumption of Normality:
ANOVA assumes that the data for each group follow a normal distribution. This means that the values within each group are symmetrically distributed around the mean, forming the characteristic bell curve. However, this assumption may not always hold true, especially in cases where the data have skewed distributions or exhibit non-normal patterns. Violation of this assumption can lead to inaccuracies in the results obtained from ANOVA.

### 2. Assumption of Homogeneity of Variance:
ANOVA assumes that the variances of the different groups being compared are equal. This assumption, also known as homoscedasticity, implies that the spread of data points around the group means is consistent across all groups. However, in real-world scenarios, this assumption might be violated, leading to unequal variances among groups. When this occurs, the validity of ANOVA results can be compromised, potentially yielding incorrect conclusions.

### 3. Independence of Observations:
ANOVA assumes that observations within and across groups are independent of each other. This means that the values obtained from one observation do not influence the values of other observations. However, in certain datasets, such as time series data or nested data, observations may be related or dependent on each other. When this assumption is violated, ANOVA may produce biased results, undermining its reliability.

### 4. Effect of Outliers:
ANOVA is sensitive to outliers, which are extreme or unusual data points that deviate significantly from the rest of the dataset. Even a single outlier can disproportionately influence the calculation of the F-statistic, which is central to ANOVA analysis, potentially leading to erroneous conclusions. Therefore, it's crucial to identify and appropriately handle outliers to ensure the robustness of ANOVA results.

### 5. Doesn't Account for Interactions:
ANOVA, like other univariate feature selection methods, does not consider interactions between features. It assesses the impact of individual factors on the response variable but does not account for how these factors may interact with each other. Consequently, ANOVA may overlook important relationships and fail to capture complex interactions present in the data. Researchers should complement ANOVA with techniques that account for interactions, such as factorial ANOVA or regression analysis, to obtain a more comprehensive understanding of the data.

### Examples and Use Cases:
- **Example:** A pharmaceutical company conducts an experiment to compare the effectiveness of three different drug formulations in treating a specific condition. They use ANOVA to analyze the mean effectiveness scores obtained from patients assigned to each drug group. However, before interpreting the results, they must ensure that the assumptions of normality and homogeneity of variance are met and that potential outliers are addressed.
- **Use Case:** An educational researcher wants to determine whether teaching methods impact students' exam scores across multiple classrooms. They collect exam scores from different classrooms and apply ANOVA to assess the significance of teaching methods on exam performance. However, they also need to consider the independence of observations, especially if students within classrooms are related or if there's a hierarchical structure to the data.

### **Mutual Information (MI) for Feature Selection**

Mutual Information (MI) is a statistical measure used in feature selection to quantify the relationship between two variables by measuring the amount of information obtained about one variable through the other variable. In the context of feature selection, MI helps identify the relevance of each feature with respect to the target variable.

#### Formula:
The formula to compute Mutual Information (MI) between two random variables X and Y is as follows:
Mutual+Information+based+Feature+Selection+Method.jpg
![Mutual Information Formula](https://latex.codecogs.com/svg.image?MI(X,Y)&space;=&space;\sum_{x&space;\in&space;X}&space;\sum_{y&space;\in&space;Y}&space;p(x,&space;y)&space;\log&space;\left(&space;\frac{p(x,&space;y)}{p(x)p(y)}&space;\right))

Where:
- \( p(x, y) \) is the joint probability distribution function of variables X and Y.
- \( p(x) \) and \( p(y) \) are the marginal probability distribution functions of variables X and Y, respectively.

#### Assumptions:
1. **Independence**: MI assumes that there is no relationship between variables beyond the one being measured. In other words, MI assumes that the relationship between variables is linear and additive.
2. **Complete Observability**: MI assumes that all possible states of the variables are observed.

#### Steps for Feature Selection using Mutual Information:

1. **Compute MI**: Calculate the MI between each feature and the target variable.
2. **Rank Features**: Rank the features based on their MI scores in descending order. Higher MI scores indicate greater relevance of the feature to the target variable.
3. **Select Features**: Select the top-ranked features with the highest MI scores as the most relevant features for the model.
4. **Model Training**: Train the model using the selected features.

#### How it works:
- MI measures the reduction in uncertainty about one variable (e.g., the target variable) given the knowledge of another variable (e.g., a feature).
- Features with higher MI scores provide more information about the target variable and are considered more relevant for prediction.
- MI considers both linear and non-linear relationships between variables, making it a versatile measure for feature selection.

#### Notable Features/Functionality:
- **Non-linear Relationships**: MI can capture non-linear relationships between variables, unlike simpler methods like correlation coefficients.
- **Feature Ranking**: MI provides a natural way to rank features based on their relevance to the target variable, facilitating feature selection in machine learning tasks.
- **Model Agnostic**: MI can be applied across various machine learning algorithms and is not specific to any particular model.

#### Example Use Case:
Consider a classification task to predict whether an email is spam or not based on various features such as the sender's email address, subject line, and content. By computing MI between each feature and the target variable (spam or not), we can identify which features are most informative for classifying emails. Features with high MI scores, such as specific keywords in the email content or suspicious sender addresses, would be selected for building the spam classification model.

Mutual Information is a powerful tool in feature selection, offering insights into the relevance of features to the target variable and aiding in the creation of efficient and accurate machine learning models.


### Steps:
1. **Compute Mutual Information:** Calculate the mutual information between each feature and the target variable.
   - For discrete variables: MI can be straightforwardly calculated.
   - For numerical variables: MI can be adapted for use with numerical input and categorical output.

2. **Select Features:** Choose the features with the highest mutual information scores as relevant features for the model.


Mutual Information (MI) feature selection can be particularly useful in certain scenarios, but it also has its advantages and disadvantages.

**When to Use Mutual Information Feature Selection:**

1. **Non-linear Relationships:** MI can capture non-linear relationships between features and the target variable, making it suitable for scenarios where linear methods like correlation may not suffice.

2. **Variable Types:** MI can handle both numerical and categorical variables, making it versatile for various types of datasets.

3. **Feature Importance Ranking:** MI provides a quantitative measure of the relevance of each feature to the target variable, allowing for easy ranking and selection of features.

4. **No Assumptions about Data Distribution:** Unlike some other feature selection methods, MI does not assume any specific distribution of data, making it more robust in diverse data environments.

5. **Reduced Overfitting:** By selecting only the most relevant features, MI can help reduce overfitting in machine learning models and improve generalization performance.

**Advantages of Mutual Information Feature Selection:**

1. **Information Gain:** MI measures the amount of information gained about the target variable by knowing the value of a feature, providing a more comprehensive understanding of feature relevance.

2. **Handles Redundancy:** MI can identify and handle redundant features by capturing their shared information content, leading to more concise feature sets.

3. **Interpretability:** MI scores provide interpretable insights into the importance of each feature, aiding in the understanding of the underlying data patterns.

**Disadvantages of Mutual Information Feature Selection:**

1. **Computationally Intensive:** Calculating MI scores for large datasets or a high number of features can be computationally intensive, especially when using non-parametric estimation methods.

2. **Parameter Selection:** MI feature selection may require tuning parameters such as the number of selected features (k), which can impact the final results and may need to be optimized through experimentation.

3. **Assumption of Independence:** MI assumes independence between features, which may not always hold true, especially in cases of highly correlated features or complex interactions.

4. **Feature Engineering Required:** While MI identifies relevant features, it does not perform feature engineering or transformation, which may still be necessary for optimal model performance.

In summary, Mutual Information feature selection is beneficial for capturing non-linear relationships, handling various types of variables, and providing interpretable insights into feature importance. However, it may require careful parameter selection and may not account for all nuances in complex datasets.


## Code Summary: Feature Selection using Correlation

### Description
This code aims to identify and remove highly correlated features from a DataFrame (`df_vt`). It sets a correlation threshold, calculates the correlation matrix, identifies highly correlated features based on the threshold, and removes one of each pair of highly correlated features while considering the "Pass/Fail" column's absolute values for decision making.

### Structure
1. **Set Threshold**: Establishes a correlation threshold of 0.8.
2. **Calculate Correlation Matrix**: Computes the correlation matrix for the DataFrame `df_vt`.
3. **Identify Highly Correlated Features**:
    - Finds indices of highly correlated features based on the threshold.
4. **Initialize Storage**:
    - Sets up a set (`columns_to_remove`) to store columns to be removed.
5. **Iterate Through Correlated Features**:
    - Loops through the indices of highly correlated features.
6. **Check and Determine Columns to Remove**:
    - Compares the absolute values of "Pass/Fail" column at the indices and decides which column to remove.
7. **Remove Identified Columns**:
    - Drops the identified columns from the DataFrame `df_vt`.
8. **Output Shape**:
    - Returns the shape of the resulting DataFrame (`df_corr`).

### Notable Features/Functionality
- Flexible Threshold: Users can adjust the correlation threshold according to their needs.
- Dynamic Column Removal: Removes one of each pair of highly correlated features based on a specified condition regarding the "Pass/Fail" column's absolute values.
- Utilization of Numpy: Makes use of Numpy functions for efficient array operations.



````python
#method1
# Set the correlation threshold
threshold = 0.8

# Calculate the correlation matrix for the DataFrame df_vt
corr_matrix = df_vt.corr()

# Find indices of highly correlated features based on the threshold
high_corr_features = np.where(np.abs(corr_matrix.values[:-1, :-1]) >= threshold)

# Initialize a set to store columns to be removed
columns_to_remove = set()

# Iterate through the indices of highly correlated features
for i, j in list(zip(*high_corr_features)):
    # Check if the indices are not equal
    if i != j:
        # Determine which column to remove based on the absolute value of Pass/Fail column
        columns_to_remove.add(j if np.abs(df_vt["Pass/Fail"]).iloc[i] >= np.abs(df_vt["Pass/Fail"]).iloc[j] else i)

# Drop the identified columns from the DataFrame df_vt
df_corr=df_vt.drop(df_vt.columns[list(columns_to_remove)], axis=1)
df_corr.shape
````


## Method2: Correlation-based Feature Selection

This code snippet performs feature selection based on correlation coefficients among features in a DataFrame (`df_vt`). Here's a breakdown of the steps:

1. **Compute Correlation Matrix**: Calculates the absolute correlation matrix (`corr_matrix`) of the features in the DataFrame.

2. **Create Mask**: Generates a mask (`mask`) to identify the upper triangular portion of the correlation matrix. This portion is chosen to avoid duplicate comparisons and self-correlation.

3. **Apply Mask**: Masks the correlation matrix (`corr_matrix`) using the generated mask, setting the lower triangular portion and diagonal to NaN. This operation results in a DataFrame (`tri_df`) containing only the upper triangular portion of the correlation matrix.

4. **Identify Highly Correlated Features**: Determines features with correlation coefficients greater than a specified threshold. These features are considered candidates for removal (`to_drop`).

5. **Drop Highly Correlated Features**: Removes the highly correlated features identified in the previous step from the original DataFrame (`df_vt`). These features are considered redundant or less informative due to their high correlation with others, and the resulting DataFrame is stored as `df_corr`.

6. **Print DataFrame Shape**: Prints the shape of the DataFrame after removing highly correlated features.

This method helps in identifying and eliminating redundant features, thus improving the efficiency and effectiveness of machine learning models by reducing dimensionality and multicollinearity issues.


````python

# Method2: Correlation-based Feature Selection
# Compute the correlation matrix to measure the correlation between features.
corr_matrix = df_vt.corr().abs()

# Create a mask to identify the upper triangular portion of the correlation matrix.
# The upper triangular portion is chosen to avoid duplicate comparisons and self-correlation.
mask = np.triu(np.ones_like(corr_matrix, dtype=bool), 1)

# Apply the mask to the correlation matrix, setting the lower triangular portion and diagonal to NaN.
tri_df = corr_matrix.where(mask)

# Identify features that have correlation coefficients greater than a specified threshold.
# Features with correlation greater than this threshold are candidates for removal.
to_drop = [c for c in tri_df.columns if any(tri_df[c] > threshold)]

# Drop the highly correlated features from the original DataFrame.
# These features are considered redundant or less informative due to their high correlation with others.
df_corr = df_vt.drop(columns=to_drop, axis=1)

# Print the shape of the DataFrame after removing highly correlated features.
print(df_corr.shape)

````

# ANOVA Feature Selection using scipy
## Description:
This code performs feature selection using ANOVA (Analysis of Variance) test to select columns with significant statistical differences between groups. It utilizes the `f_oneway` function from `scipy.stats` module to calculate the ANOVA statistic and p-values. Features with p-values lower than a specified significance level (`alpha`) are considered significant and retained.

## Structure:
1. Import the necessary library `f_oneway` from `scipy.stats`.
2. Set the significance level (`alpha`) to 0.05.
3. Initialize an empty list `colm` to store significant columns.
4. Iterate through each column in the dataframe (`df_corr`) except the last one:
   - Retrieve the column data.
   - Perform ANOVA test between the current column and the 'Pass/Fail' column.
   - If the p-value is less than or equal to `alpha`, add the column and its p-value to `colm`.
5. Sort `colm` based on p-values in ascending order.
6. Select the top 100 columns with the lowest p-values and store them in `selected_columns`.
7. Create a new dataframe (`df_annova`) containing only the selected columns.
8. Print the shape of the new dataframe.
9. Retrieve the column names of `df_annova`.
10. Prepare training and testing datasets (`train_x` and `test_x`) by excluding the last column from `x_train` and `x_test`, which presumably correspond to the 'Pass/Fail' column.

## Notable Features/Functionality:
- Utilizes ANOVA test for feature selection, which is suitable for comparing means across multiple groups.
- Provides flexibility to adjust the significance level (`alpha`) for controlling the rate of false positives.
- Selects the top 100 significant columns based on p-values for further analysis.



````python
from scipy.stats import f_oneway
alpha=.05
colm=[]
for i in df_corr.columns[:-1]:
  column_data = df_corr[i]
  anova_result=f_oneway(column_data, df_corr['Pass/Fail'])
  #print(f"Column: {i} - ANOVA p-value: {anova_result.pvalue}")
  if anova_result.pvalue <= alpha:
    colm.append((i,anova_result.pvalue))
colm.sort(key=lambda x: x[1])

selected_columns = [column for column, _ in colm[:100]]
df_annova=df_corr[selected_columns]
print(df_annova.shape)
a=df_annova.columns
train_x=x_train[df_corr.columns[:-1]]
test_x=x_test[df_corr.columns[:-1]]
````
