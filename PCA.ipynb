### **Curse of Dimensionality**

The curse of dimensionality refers to the various problems that arise when working with high-dimensional data in spaces with many dimensions. It primarily impacts algorithms and processes that rely on distance metrics, such as nearest neighbor search, clustering, and manifold learning. As the number of dimensions increases, the volume of the space grows exponentially, causing data points to become increasingly sparse, and the relative distances between them to become less meaningful.

#### Definition:

The curse of dimensionality can be defined as follows:
- **In simple terms**: As the number of dimensions (features) in a dataset increases, the amount of data needed to adequately sample that space grows exponentially.
- **Mathematically**: It refers to the phenomenon where the number of samples required to maintain the same level of data density in a space increases exponentially with the dimensionality of the space.

#### Impact on Machine Learning:

The curse of dimensionality has significant implications for machine learning tasks:
1. **Increased computational complexity**: Algorithms that rely on distance calculations become computationally expensive as the number of dimensions grows.
2. **Increased risk of overfitting**: High-dimensional spaces make it easier for models to fit noise in the data, leading to overfitting and poor generalization.
3. **Decreased performance**: Many machine learning algorithms require large amounts of data to accurately learn patterns. In high-dimensional spaces, the amount of data needed to achieve a certain level of performance becomes impractical.
4. **Difficulty in visualization**: Visualizing high-dimensional data becomes challenging, making it harder for humans to interpret and understand the underlying patterns.

#### Methods to Reduce the Impact of Curse of Dimensionality:

Several techniques can help mitigate the curse of dimensionality:
1. **Feature selection**: Choose a subset of the most relevant features, discarding irrelevant or redundant ones. This reduces the dimensionality of the data while preserving most of the information.
2. **Dimensionality reduction**: Techniques like Principal Component Analysis (PCA), t-Distributed Stochastic Neighbor Embedding (t-SNE), and Singular Value Decomposition (SVD) project high-dimensional data onto a lower-dimensional subspace while preserving important characteristics.
3. **Manifold learning**: This approach focuses on preserving the intrinsic dimensionality of the data by learning a low-dimensional representation that captures the underlying structure of the data.
4. **Regularization**: In models like linear regression or neural networks, regularization techniques such as Lasso or Ridge regression can penalize overly complex models, discouraging them from learning from irrelevant features.
5. **Ensemble methods**: Using ensemble methods like Random Forests or Gradient Boosting can help handle high-dimensional data by aggregating predictions from multiple models trained on different subsets of features.

By employing these methods, practitioners can mitigate the adverse effects of the curse of dimensionality and improve the performance of machine learning algorithms on high-dimensional data.


###**Principal Component Analysis (PCA)**

image.png



**Definition:**
Principal Component Analysis (PCA) is a dimensionality reduction technique used to simplify high-dimensional data while preserving most of its important features. It achieves this by transforming the original variables into a new set of orthogonal variables called principal components. These components are ordered by the amount of variance they explain in the data, with the first principal component explaining the most variance, the second explaining the second-most variance, and so on.

**Structure:**
PCA takes a dataset with $( m )$ observations and $( n )$ variables and transforms it into a new dataset with the same number of observations but fewer variables, where the variables are linear combinations of the original variables.

## Mathematical Concept of PCA

The concept of PCA revolves around finding the eigenvectors (principal components) of the covariance matrix of the data.

Let's say we have a dataset $X$ consisting of $n$ observations and $p$ features. The covariance matrix $\Sigma$ of $X$ is given by:

$$
\Sigma = \frac{1}{n-1} (X - \bar{X})^T (X - \bar{X})
$$

where $\bar{X}$ is the mean of the dataset $X$.

The principal components are the eigenvectors of the covariance matrix $\Sigma$. These eigenvectors represent the directions in the feature space along which the data varies the most.

## Formula and Derivation

1. **Compute Covariance Matrix**: Calculate the covariance matrix $\Sigma$ using the formula mentioned above.

2. **Compute Eigenvectors and Eigenvalues**: Compute the eigenvectors $v_i$ and corresponding eigenvalues $\lambda_i$ of the covariance matrix $\Sigma$. This can be done by solving the following equation:

$$
\Sigma v_i = \lambda_i v_i
$$

where $v_i$ is the $i^{th}$ eigenvector and $\lambda_i$ is its corresponding eigenvalue.

3. **Sort Eigenvectors by Eigenvalues**: Sort the eigenvectors in descending order based on their corresponding eigenvalues. This step is crucial because the eigenvectors with higher eigenvalues explain more variance in the data.

4. **Select Principal Components**: Choose the first $k$ eigenvectors corresponding to the $k$ largest eigenvalues to form the principal components. These eigenvectors represent the directions of maximum variance in the data.

The transformation matrix $W$ for PCA is formed by stacking the selected eigenvectors column-wise.

5. **Transform Data**: To transform the original data $X$ into the lower-dimensional space, compute the dot product of the data matrix $X$ with the transformation matrix $W$:

$$
Z = XW
$$

where $Z$ is the transformed data matrix containing the principal components.

This process results in a lower-dimensional representation of the original data while preserving the maximum amount of variance.

**How PCA Works (2nd method):**
1. Standardize the dataset (subtract mean, divide by standard deviation).
2. Compute the covariance matrix of the standardized dataset.
3. Calculate the eigenvectors and eigenvalues of the covariance matrix.
4. Sort the eigenvectors by their corresponding eigenvalues in descending order.
5. Choose the top $( k )$ eigenvectors to form the transformation matrix $( W )$.
6. Transform the original dataset $( X )$ using $( Z = XW )$.

**Use in Machine Learning:**
- **Dimensionality Reduction:** PCA is primarily used for reducing the dimensionality of datasets with many variables, making it easier to visualize and analyze.
- **Noise Reduction:** PCA can also help in reducing noise and redundancy in data, potentially improving the performance of machine learning algorithms.
- **Feature Extraction:** In some cases, the principal components themselves can serve as new features for machine learning models.

**Assumptions:**
PCA assumes that the principal components are linear combinations of the original variables and that the variables are standardized to have zero mean and unit variance.

**Notable Features/Functionality:**
- **Explained Variance:** PCA provides information about the amount of variance explained by each principal component, allowing users to decide how many components to retain based on a desired level of variance explained.
- **Orthogonality:** Principal components are orthogonal to each other, meaning they capture independent sources of variation in the data.
- **Visualization:** PCA can be used to visualize high-dimensional data in lower-dimensional space, making it easier to understand the underlying structure of the data.

**Examples and Use Cases:**
- **Image Compression:** PCA can be used to compress images by reducing the dimensionality of the pixel data while retaining most of the important features.
- **Finance:** In finance, PCA is used for portfolio optimization, risk management, and identifying factors driving asset returns.
- **Genomics:** PCA is used in genomics to analyze gene expression data and identify patterns or clusters of genes with similar expression profiles.

PCA is a versatile technique with applications across various domains, offering a powerful tool for dimensionality reduction and data exploration in machine learning and data analysis.

###**Types of Principal Component Analysis (PCA)**

### **Standard PCA**

**Definition:** Standard PCA, also known as classic PCA, is a dimensionality reduction technique used to reduce the number of variables in a dataset while preserving most of its variance. It works by transforming the original variables into a new set of uncorrelated variables called principal components, which are linear combinations of the original variables. These components are ordered in such a way that the first principal component explains the maximum variance in the data, followed by the second, and so on.

### Mathematical Concept

At its core, PCA seeks to find a set of orthogonal vectors (principal components) that best describe the variance in the data. These principal components are ordered in terms of the amount of variance they explain, with the first PC explaining the most variance, the second PC explaining the second most variance, and so on.

### **Formula and Derivation**

### Step 1: Data Normalization

Before performing PCA, it's essential to normalize the data to have zero mean and unit variance. Let $X$ be the data matrix with $n$ samples and $d$ features. The normalization of each feature $x_i$ is given by:

$$
\bar{x}_i = \frac{x_i - \mu_i}{\sigma_i}
$$

where $\mu_i$ is the mean of feature $i$ and $\sigma_i$ is its standard deviation.

### Step 2: Covariance Matrix

Next, we compute the covariance matrix $C$ for the normalized data:

$$
C = \frac{1}{n-1}X^T X
$$

### Step 3: Eigenvalue Decomposition

We then perform eigenvalue decomposition on the covariance matrix $C$ to find its eigenvectors and eigenvalues. The eigenvectors represent the principal components, and the eigenvalues represent the amount of variance explained by each principal component.

$$
Cv_i = \lambda_i v_i
$$

where $v_i$ is the $i$-th eigenvector and $\lambda_i$ is the corresponding eigenvalue.

### Step 4: Selecting Principal Components

We sort the eigenvectors based on their corresponding eigenvalues in descending order. The eigenvector with the highest eigenvalue represents the direction of maximum variance in the data and is chosen as the first principal component. The subsequent eigenvectors are selected in the same manner.

### Step 5: Projection

Finally, we project the original data onto the new coordinate system defined by the principal components. The transformed data matrix $Y$ is given by:

$$
Y = XV
$$

where $V$ is a matrix whose columns are the selected eigenvectors (principal components).

### Conclusion

In summary, PCA is a powerful technique for dimensionality reduction and data visualization. By finding the principal components that capture the maximum variance in the data, PCA helps in simplifying complex datasets while preserving important information. Through the mathematical concepts and formulas provided above, we gain insights into the inner workings of PCA and how it can be applied effectively in practice.

**When to Use:** Standard PCA is suitable for datasets where linear relationships exist between variables and where preserving most of the variance in the data is crucial. It is commonly used for exploratory data analysis, feature extraction, and data visualization.

### **Incremental PCA**

**Definition:** Incremental PCA (IPCA) is a variant of PCA designed to handle large datasets that cannot fit into memory. Unlike standard PCA, which requires the entire dataset to be loaded into memory for computation, IPCA processes the data in mini-batches, making it suitable for datasets with millions of samples.

**Mathematical Concept of IPCA:**

The goal of PCA is to find a low-dimensional representation of a dataset while preserving as much variance as possible. It does this by identifying the principal components, which are the orthogonal directions in the feature space along which the data varies the most. IPCA follows the same principle, but it updates the principal components incrementally as new data arrives.

**Formula and Derivation:**

Let's denote the dataset matrix as $X \in \mathbb{R}^{n \times d}$, where $n$ is the number of samples and $d$ is the number of features. We assume that the data is centered, meaning the mean of each feature is zero.

1. **Compute Covariance Matrix:**
   The first step in PCA is to compute the covariance matrix of the dataset:
   $$C = \frac{1}{n}X^TX$$

2. **Eigenvalue Decomposition:**
   Next, we perform eigenvalue decomposition on the covariance matrix $C$:
   $$C = VDV^T$$
   where $V$ is a matrix whose columns are the eigenvectors of $C$, and $D$ is a diagonal matrix containing the corresponding eigenvalues.

3. **Update Step:**
   With IPCA, we perform an incremental update to the covariance matrix and its eigenvectors whenever new data arrives.

   Let $X_k$ be the new data batch with size $m \times d$. We update the covariance matrix using the new data:
   $$C_k = \frac{1}{n+m} \left( X^TX + X_k^TX_k \right)$$

4. **Eigenvalue Decomposition of Updated Covariance:**
   We then perform eigenvalue decomposition on the updated covariance matrix $C_k$:
   $$C_k = V_kD_kV_k^T$$

5. **Combining Eigenvectors:**
   We need to ensure that the eigenvectors are consistent across updates. We use the following formula to update the eigenvectors incrementally:
   $$V_{k+1} = (1 - \alpha)V_k + \alpha V_k'$$
   where $V_k'$ is the eigenvectors computed from the updated covariance matrix $C_k$, and $\alpha$ is a learning rate.

6. **Retain Top Principal Components:**
   Finally, we select the top $k$ eigenvectors corresponding to the largest eigenvalues to obtain the principal components of the updated dataset.

This process allows IPCA to adapt to new data while maintaining the low-dimensional representation learned from previous data. It's particularly useful for scenarios where the entire dataset cannot be loaded into memory at once.

**When to Use:** IPCA is useful when working with extremely large datasets that cannot be loaded into memory all at once. It allows you to perform PCA in a memory-efficient manner by processing the data in smaller chunks.

### **Kernel PCA (Kernel Principal Component Analysis)**

Kernel PCA (Principal Component Analysis) is a nonlinear extension of the traditional PCA method. It aims to find nonlinear patterns in data by implicitly mapping the input data into a higher-dimensional space using a kernel function. This higher-dimensional space allows for capturing more complex patterns that cannot be captured by linear methods like traditional PCA.

#### **Mathematical Concept:**

1. **Kernel Trick**:
    The kernel trick is the key concept behind kernel PCA. It allows us to implicitly compute the dot product in the higher-dimensional feature space without explicitly transforming the data into that space. This is computationally efficient and enables us to handle data that are not linearly separable in the original feature space.

2. **Kernel Function**:
    A kernel function $K(x_i, x_j)$ measures the similarity between two data points $x_i$ and $x_j$ in the original feature space. Commonly used kernel functions include:
    - Linear kernel: $K(x_i, x_j) = x_i \cdot x_j$
    - Polynomial kernel: $K(x_i, x_j) = (x_i \cdot x_j + c)^d$
    - Gaussian (RBF) kernel: $K(x_i, x_j) = \exp(-\frac{\|x_i - x_j\|^2}{2\sigma^2})$
    - Sigmoid kernel: $K(x_i, x_j) = \tanh(\alpha x_i \cdot x_j + c)$
    
3. **Kernel PCA Objective**:
    Kernel PCA aims to maximize the variance of the data in the transformed feature space. The objective function is:
    
    $$
    \text{Maximize } \frac{1}{n} \sum_{i=1}^{n} \| \Phi(x_i) - \bar{\Phi} \|^2
    $$
    
    where $\Phi(x_i)$ represents the feature space transformation of data point $x_i$, $\bar{\Phi}$ is the mean of the transformed data points, and $n$ is the number of data points.
    
#### Formula with Derivation:

The kernel PCA algorithm involves the following steps:

1. **Compute Kernel Matrix**:
   Given $N$ data points $x_1, x_2, ..., x_N$, compute the $N \times N$ kernel matrix $K$, where $K_{ij} = K(x_i, x_j)$.

2. **Centering the Kernel Matrix**:
   Center the kernel matrix $K$ by subtracting the mean of each row, column, and the entire matrix:
   
   $$
   \tilde{K} = K - \frac{1}{N} \mathbf{1}_N K - \frac{1}{N} K \mathbf{1}_N + \frac{1}{N^2} \mathbf{1}_N K \mathbf{1}_N
   $$
   
   where $\mathbf{1}_N$ is an $N$-dimensional vector of ones.

3. **Eigen Decomposition**:
   Perform eigen decomposition on the centered kernel matrix $\tilde{K}$. Obtain the eigenvectors $\alpha_1, \alpha_2, ..., \alpha_N$ and corresponding eigenvalues $\lambda_1, \lambda_2, ..., \lambda_N$.

4. **Select Principal Components**:
   Select the $k$ eigenvectors corresponding to the largest $k$ eigenvalues to form the matrix $\mathbf{A} = [\alpha_1, \alpha_2, ..., \alpha_k]$.

5. **Project Data**:
   Project the data onto the first $k$ principal components:
   
   $$
   \mathbf{Z} = \mathbf{A}^T \Phi(X)
   $$
   
   where $\Phi(X)$ is the $N \times d$ matrix containing the original data points as rows, and $\mathbf{Z}$ is the $k \times N$ matrix containing the transformed data points as columns.

Kernel PCA provides a powerful tool for nonlinear dimensionality reduction and pattern discovery in high-dimensional data. It has applications in various fields including image processing, bioinformatics, and finance.

**When to Use:** Kernel PCA is beneficial when dealing with datasets where the relationships between variables are nonlinear. It is commonly used in tasks such as manifold learning, pattern recognition, and nonlinear feature extraction.

###**Sparse PCA (Principal Component Analysis)**

Sparse Principal Component Analysis (Sparse PCA) is a variant of the classical Principal Component Analysis (PCA) that aims to find linear combinations of variables (or features) in a dataset that capture the most variance, while enforcing sparsity in the resulting components. Sparsity implies that the components are mostly comprised of zeros, meaning only a subset of the original features contributes significantly to each component. Sparse PCA is particularly useful for feature selection and dimensionality reduction in high-dimensional datasets, where identifying a small number of informative features is desired.

**Mathematical Concept:**

In Sparse PCA, we seek to find a set of orthogonal vectors $v_1, v_2, ..., v_k$ such that the variance of the projected data onto these vectors is maximized, subject to the constraint that each vector is sparse. This is achieved by solving an optimization problem.

**Formula:**

Let $X$ be the data matrix with dimensions $n \times p$, where $n$ is the number of samples and $p$ is the number of features.

The goal of Sparse PCA is to find the loading vectors $v_1, v_2, ..., v_k$ such that:

1. They are orthogonal to each other: $v_i^Tv_j = 0$ for $i \neq j$.
2. They maximize the variance of the projected data onto these vectors.
3. Each loading vector $v_i$ is sparse, meaning it contains mostly zeros.

Mathematically, the optimization problem for Sparse PCA can be formulated as:

$$
\max_{v_1, v_2, ..., v_k} \sum_{i=1}^k v_i^T X^T X v_i
$$

Subject to the constraints:

1. Orthogonality: $v_i^Tv_j = 0$, for $i \neq j$.
2. Sparsity: $\|v_i\|_0 \leq s$ for $i = 1, 2, ..., k$, where $\|v_i\|_0$ denotes the $L_0$ norm (count of non-zero elements) and $s$ is the desired sparsity level.

**Derivation:**

To solve the optimization problem, we can use the method of Lagrange multipliers. We introduce Lagrange multipliers $\lambda_{ij}$ to enforce orthogonality constraints and $\gamma_i$ to enforce sparsity constraints.

The Lagrangian for the optimization problem becomes:

$$
\mathcal{L}(v_1, v_2, ..., v_k, \lambda_{ij}, \gamma_i) = \sum_{i=1}^k v_i^T X^T X v_i - \sum_{i \neq j} \lambda_{ij} v_i^Tv_j - \sum_{i=1}^k \gamma_i (\|v_i\|_0 - s)
$$

Taking derivatives with respect to $v_i$ and setting them to zero, we get:

$$
\frac{\partial \mathcal{L}}{\partial v_i} = 2X^TXv_i - \sum_{j \neq i} \lambda_{ij}v_j - \gamma_i \cdot \text{sign}(v_i) = 0
$$

where $\text{sign}(v_i)$ is the element-wise sign function of $v_i$. This equation can be solved for each $v_i$.

Once the loading vectors $v_1, v_2, ..., v_k$ are obtained, they can be used to transform the original data into a lower-dimensional space while preserving the most important information.

**When to Use:** Sparse PCA is useful when you want a more interpretable representation of the data, especially when dealing with high-dimensional datasets where most variables are irrelevant or redundant. It is commonly used in fields such as genetics, finance, and neuroscience.

### **Robust PCA (Principal Component Analysis)**

Robust Principal Component Analysis (RPCA) is a technique used for decomposing a given data matrix into two components: a low-rank matrix representing the underlying structure of the data, and a sparse matrix representing the noise or outliers in the data. It is particularly useful when the data is corrupted by outliers or gross errors.

### **Mathematical Concept**

Given a data matrix $M$, RPCA aims to decompose it into two matrices:

1. Low-rank matrix $L$: This represents the underlying structure of the data.
2. Sparse matrix $S$: This captures the noise or outliers present in the data.

Mathematically, the decomposition can be formulated as:

$$
M = L + S
$$

where:
- $M$ is the original data matrix,
- $L$ is the low-rank matrix,
- $S$ is the sparse matrix.

The problem is to find $L$ and $S$ such that the sum of the rank of $L$ and the sparsity of $S$ is minimized.



### Objective Function:

The objective function for RPCA can be defined as:

$$
\min_{L, S} \left \| L \right \|_* + \lambda \left \| S \right \|_1
$$

where:
- $\left \| L \right \|_*$ denotes the nuclear norm (sum of singular values) of $L$,
- $\left \| S \right \|_1$ denotes the $\ell_1$ norm of $S$,
- $\lambda$ is a regularization parameter controlling the trade-off between the low-rank and sparse components.

### Optimization Problem:

The objective function can be optimized using Alternating Direction Method of Multipliers (ADMM) or other optimization techniques.

#### Step 1: Update $L$

We fix $S$ and update $L$ by solving:

$$
\min_{L} \left \| L \right \|_* + \frac{\rho}{2} \left \| L - (M - S + \frac{U}{\rho}) \right \|_F^2
$$

where:
- $\rho$ is the penalty parameter,
- $U$ is the Lagrange multiplier,
- $\left \| \cdot \right \|_F$ denotes the Frobenius norm.

The solution to this subproblem can be obtained by performing Singular Value Thresholding (SVT) on $M - S + \frac{U}{\rho}$.

#### Step 2: Update $S$

We fix $L$ and update $S$ by solving:

$$
\min_{S} \lambda \left \| S \right \|_1 + \frac{\rho}{2} \left \| L + (M - S + \frac{U}{\rho}) \right \|_F^2
$$

This can be solved using soft thresholding on the elements of $L + \frac{U}{\rho}$.

#### Step 3: Update Lagrange Multiplier $U$

Update the Lagrange multiplier:

$$
U = U + \rho \cdot (M - L - S)
$$

#### Step 4: Update Penalty Parameter $\rho$

Update the penalty parameter $\rho$ based on convergence criteria.

### Algorithm:

1. Initialize $L$, $S$, and $U$.
2. Repeat until convergence:
   - Update $L$.
   - Update $S$.
   - Update $U$.
   - Update $\rho$.
3. Output $L$ and $S$.

This algorithm iteratively decomposes the data matrix $M$ into its low-rank and sparse components, effectively separating the underlying structure from the noise or outliers.

**When to Use:** Robust PCA is suitable for datasets that contain outliers or are contaminated with noise. It is commonly used in anomaly detection, image processing, and signal processing tasks where the presence of outliers can significantly affect the results of PCA.

### **Covariance:**

Covariance measures the degree to which two variables change together. It indicates the direction of the linear relationship between two variables. If the covariance is positive, it means that as one variable increases, the other tends to also increase. If the covariance is negative, it means that as one variable increases, the other tends to decrease.

The covariance between two random variables $X$ and $Y$ is denoted as $\text{cov}(X,Y)$ and is calculated as follows:

$$
\text{cov}(X,Y) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})
$$

Where:
- $n$ is the number of data points.
- $X_i$ and $Y_i$ are individual data points.
- $\bar{X}$ and $\bar{Y}$ are the means of $X$ and $Y$, respectively.

### **Covariance Matrix:**

The covariance matrix, often denoted as $\Sigma$, is a square matrix that summarizes the covariances between multiple pairs of random variables. In a covariance matrix, each element represents the covariance between two variables.

For a set of random variables $X_1, X_2, ..., X_n$, the covariance matrix $\Sigma$ is an $n \times n$ matrix, where the $(i, j)$ entry represents the covariance between $X_i$ and $X_j$.

The formula to calculate the covariance matrix is:

$$
\Sigma =
\begin{bmatrix}
\text{cov}(X_1, X_1) & \text{cov}(X_1, X_2) & \cdots & \text{cov}(X_1, X_n) \\
\text{cov}(X_2, X_1) & \text{cov}(X_2, X_2) & \cdots & \text{cov}(X_2, X_n) \\
\vdots & \vdots & \ddots & \vdots \\
\text{cov}(X_n, X_1) & \text{cov}(X_n, X_2) & \cdots & \text{cov}(X_n, X_n) \\
\end{bmatrix}
$$

Where each $\text{cov}(X_i, X_j)$ term is computed using the formula for covariance mentioned earlier.

### Derivation:

Let's derive the formula for covariance between two variables $X$ and $Y$.

Given two random variables $X$ and $Y$, their covariance can be derived as follows:

$$
\text{cov}(X,Y) = E[(X - \mu_X)(Y - \mu_Y)]
$$

Expanding this expression gives:

$$
\text{cov}(X,Y) = E[XY - X\mu_Y - Y\mu_X + \mu_X\mu_Y]
$$

Using linearity of expectation, we get:

$$
\text{cov}(X,Y) = E[XY] - E[X]\mu_Y - E[Y]\mu_X + \mu_X\mu_Y
$$

Since $E[X\mu_Y] = \mu_Y E[X]$ and $E[Y\mu_X] = \mu_X E[Y]$, the expression simplifies to:

$$
\text{cov}(X,Y) = E[XY] - \mu_X\mu_Y - \mu_Y\mu_X + \mu_X\mu_Y
$$

Hence,

$$
\text{cov}(X,Y) = E[XY] - \mu_X\mu_Y
$$

Finally, expressing $E[XY]$ in terms of sample means gives us:

$$
\text{cov}(X,Y) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})
$$

This completes the derivation of the covariance formula. The covariance matrix is then constructed using these covariance values for all pairs of variables.

### **Covariance vs. Correlation in PCA:**

In Principal Component Analysis (PCA), covariance is preferred over correlation for scaling purposes.

- **Covariance**: PCA aims to find the directions (principal components) of maximum variance in the data. Since covariance measures the strength and direction of the linear relationship between two variables, maximizing covariance helps in capturing the directions of maximum variance.

- **Correlation**: While correlation also measures the relationship between variables, it additionally scales the values by the standard deviations of the variables. This scaling makes correlation sensitive to the scale of measurement, which can affect the PCA results. If variables are measured in different units or have different variances, correlation may distort the principal components.

By using covariance instead of correlation, PCA focuses solely on the variances and covariances between variables, making it more robust and invariant to scaling.



### **Orthogonal Matrix**

An orthogonal matrix is a square matrix where its rows and columns are orthogonal unit vectors. In other words, the columns (and rows) of an orthogonal matrix are perpendicular to each other and each column (and row) has a length of 1. This property preserves the lengths and angles of vectors when multiplied by the matrix.

### Mathematical Concept

Let's denote an orthogonal matrix as $Q$. For any orthogonal matrix $Q$, the following property holds:

$$
Q^T \cdot Q = I
$$

where $Q^T$ is the transpose of $Q$, and $I$ is the identity matrix. This property implies that the columns of $Q$ are orthogonal to each other (since the dot product of any two distinct columns is zero) and each column has a length of 1 (since $Q^T \cdot Q = I$ means that the dot product of any column with itself is 1).

### Formula and Derivation

Let's denote the columns of $Q$ as $q_1, q_2, ..., q_n$, where $n$ is the dimension of the matrix.

The $i$-th column of $Q$ is denoted as $q_i$. Since the columns of $Q$ are unit vectors, we have:

$$
\|q_i\| = 1 \quad \text{for } i = 1, 2, ..., n
$$

Now, let's consider the dot product of two distinct columns $q_i$ and $q_j$:

$$
q_i \cdot q_j = 0 \quad \text{for } i \neq j
$$

This is because the columns are orthogonal to each other.

Now, let's consider the product $Q^T \cdot Q$. Its $(i, j)$-th entry is the dot product of the $i$-th row of $Q^T$ and the $j$-th column of $Q$, which is:

$$
(Q^T \cdot Q)_{ij} = q_i^T \cdot q_j
$$

Since $q_i$ and $q_j$ are orthogonal when $i \neq j$, all off-diagonal entries of $Q^T \cdot Q$ are zero. For the diagonal entries ($i = j$), we have:

$$
(Q^T \cdot Q)_{ii} = q_i^T \cdot q_i = \|q_i\|^2 = 1
$$

This is because the length of each column of $Q$ is 1.

Therefore, $Q^T \cdot Q = I$, confirming the definition of an orthogonal matrix.

So, in conclusion, an orthogonal matrix $Q$ satisfies the property $Q^T \cdot Q = I$, where the columns of $Q$ are unit vectors and orthogonal to each other.

### **Explanation of Eigenvalues and Eigenvectors:**

### Eigenvalues and Eigenvectors:

Eigenvalues and eigenvectors are concepts in linear algebra that are closely related to matrices.

- **Eigenvalues**: Eigenvalues ($\lambda$) are scalar values associated with a linear transformation represented by a square matrix. They represent how much the eigenvectors are stretched or compressed by the transformation.

- **Eigenvectors**: Eigenvectors ($\mathbf{v}$) are non-zero vectors that, when transformed by a matrix, only change in magnitude (scale) and not in direction. In other words, after transformation, the eigenvector is proportional to its original form.

### Mathematical Concept and Formula:

For a square matrix $A$, an eigenvector $\mathbf{v}$ and its corresponding eigenvalue $\lambda$ satisfy the equation:

$$A\mathbf{v} = \lambda\mathbf{v}$$

This equation can be rearranged as:

$$(A - \lambda I)\mathbf{v} = \mathbf{0}$$

where $I$ is the identity matrix and $\mathbf{0}$ is the zero vector.

For a matrix $A$, the eigenvalues can be found by solving the characteristic equation:

$$\det(A - \lambda I) = 0$$

Once the eigenvalues are found, the corresponding eigenvectors can be determined by solving the equation $(A - \lambda I)\mathbf{v} = \mathbf{0}$.

### Derivation:

Let's denote an eigenvector as $\mathbf{v}$ and its corresponding eigenvalue as $\lambda$. Then, from the definition of eigenvalues and eigenvectors, we have:

$$A\mathbf{v} = \lambda\mathbf{v}$$

Rearranging terms, we get:

$$(A - \lambda I)\mathbf{v} = \mathbf{0}$$

This equation implies that the matrix $(A - \lambda I)$ transforms the eigenvector $\mathbf{v}$ into the zero vector. For this to hold, the matrix $(A - \lambda I)$ must be singular, meaning its determinant is zero:

$$\det(A - \lambda I) = 0$$

This characteristic equation provides the eigenvalues $\lambda$ for the matrix $A$. Once the eigenvalues are found, the corresponding eigenvectors can be determined by solving the equation $(A - \lambda I)\mathbf{v} = \mathbf{0}$.

### Use in PCA (Principal Component Analysis):

In Principal Component Analysis (PCA), eigenvectors and eigenvalues play a crucial role. PCA aims to reduce the dimensionality of data while preserving its variance. The eigenvectors of the covariance matrix of the data represent the directions of maximum variance, and the corresponding eigenvalues represent the magnitude of that variance along those directions.

1. **Eigenvalues**: They indicate the amount of variance captured by each principal component. Higher eigenvalues correspond to more significant variability in the data along the direction of the corresponding eigenvector.

2. **Eigenvectors**: These represent the principal components themselves. The directions of these eigenvectors denote the directions of maximum variance in the data.

In PCA, one typically sorts the eigenvalues in descending order and selects the top eigenvectors (principal components) corresponding to the largest eigenvalues. These principal components are used to transform the original data into a new, lower-dimensional space while retaining as much variance as possible. This transformation aids in visualization, data compression, and feature extraction.

### **Finding Eigenvalues and Eigenvectors**

Given a square matrix $A$, to find its eigenvalues and eigenvectors:

1. **Setup**: Begin by writing down the matrix equation for finding eigenvalues and eigenvectors: $A\mathbf{v} = \lambda \mathbf{v}$, where:
   - $A$ is the given square matrix,
   - $\mathbf{v}$ is the eigenvector,
   - $\lambda$ is the eigenvalue associated with that eigenvector.

2. **Characteristic Equation**: Rearrange the equation $A\mathbf{v} = \lambda \mathbf{v}$ to $(A - \lambda I)\mathbf{v} = \mathbf{0}$, where $I$ is the identity matrix.
   - This becomes the characteristic equation $|A - \lambda I| = 0$, where $| \cdot |$ denotes the determinant.

3. **Calculate Determinant**: Compute the determinant of the matrix $A - \lambda I$. This determinant is a polynomial in $\lambda$ called the characteristic polynomial.

4. **Solve the Characteristic Polynomial**: Set the characteristic polynomial equal to zero and solve for $\lambda$. The roots of this polynomial are the eigenvalues of the matrix.

5. **Find Eigenvectors**: For each eigenvalue $\lambda_i$, substitute it back into the equation $(A - \lambda_i I)\mathbf{v} = \mathbf{0}$ and solve for the vector $\mathbf{v}$. These solutions are the corresponding eigenvectors.

6. **Normalization**: Normalize each eigenvector by dividing it by its magnitude (norm) to obtain a unit eigenvector.

### Example

Let's consider a simple 2x2 matrix $A$:

$$
A = \begin{pmatrix} 2 & 1 \\ 1 & 3 \end{pmatrix}
$$

1. **Setup**: We want to find $\lambda$ and $\mathbf{v}$ such that $A\mathbf{v} = \lambda \mathbf{v}$.

2. **Characteristic Equation**: We have $(A - \lambda I)\mathbf{v} = \mathbf{0}$.
   - So, $|A - \lambda I| = 0$ gives us the characteristic polynomial.

3. **Calculate Determinant**:
   - $A - \lambda I = \begin{pmatrix} 2-\lambda & 1 \\ 1 & 3-\lambda \end{pmatrix}$
   - $|A - \lambda I| = (2-\lambda)(3-\lambda) - (1 \cdot 1) = \lambda^2 - 5\lambda + 5$

4. **Solve the Characteristic Polynomial**: Setting $|A - \lambda I| = 0$ gives us $\lambda^2 - 5\lambda + 5 = 0$.
   - Solving this quadratic equation gives us the eigenvalues: $\lambda_1 = 2 + i$ and $\lambda_2 = 2 - i$, where $i$ is the imaginary unit.
   
  **use the below quadratic equation formula to find the eigen value;**

     $$x = \frac{ -b \pm \sqrt{b^2 - 4ac}}{ 2a }$$

5. **Find Eigenvectors**:
   - For $\lambda_1 = 2 + i$:
     - Substitute $\lambda_1$ into $(A - \lambda_1 I)\mathbf{v} = \mathbf{0}$.
     - Solve the resulting system of equations to find the eigenvector $\mathbf{v}_1$.
   - For $\lambda_2 = 2 - i$:
     - Substitute $\lambda_2$ into $(A - \lambda_2 I)\mathbf{v} = \mathbf{0}$.
     - Solve the resulting system of equations to find the eigenvector $\mathbf{v}_2$.



6. **Normalization**: Normalize each eigenvector to obtain unit eigenvectors.

This process can be generalized to larger matrices as well, following the same steps.

### **Square Matrix**

A square matrix $ A $ is a matrix with the same number of rows and columns. Here are some key properties of a square matrix:

1. **Determinant**:
   - The determinant of a square matrix $ A $, denoted as $ |A| $, is a scalar value that provides important information about the matrix. It determines whether the matrix is invertible or singular.
   
2. **Eigenvalues and Eigenvectors**:
   - Square matrices have eigenvalues and corresponding eigenvectors, which are crucial in various mathematical and computational tasks such as solving systems of linear equations, dimensionality reduction, and stability analysis.

3. **Diagonal Matrix**:
   - A square matrix is called diagonal if all its non-diagonal elements are zero, i.e., $ a_{ij} = 0 $ for $ i \neq j $.

4. **Symmetric Matrix**:
   - A square matrix $ A $ is symmetric if it is equal to its transpose, i.e., $ A = A^T $.

5. **Orthogonal Matrix**:
   - A square matrix $ Q $ is orthogonal if its transpose is equal to its inverse, i.e., $ Q^T = Q^{-1} $.

6. **Idempotent Matrix**:
   - A square matrix $ A $ is idempotent if $ A^2 = A $.

7. **Trace**:
   - The trace of a square matrix $ A $, denoted as $ \text{Tr}(A) $, is the sum of its diagonal elements.


##**Symmetric matrix**
A **symmetric matrix** is a square matrix that is equal to its transpose. In mathematical terms, for a matrix $A$ to be symmetric:

$$ A = A^T $$

This means that the $i$th row and $j$th column element of matrix $A$ is equal to the $j$th row and $i$th column element. In other words, the matrix is symmetric across its main diagonal.

**Properties of symmetric matrices** include:

1. **Equal Eigenvalues**: Symmetric matrices always have real eigenvalues. Additionally, if a symmetric matrix is positive definite, all its eigenvalues are positive.

2. **Orthogonal Eigenvectors**: Symmetric matrices have orthogonal eigenvectors corresponding to distinct eigenvalues. This property is important in diagonalizing symmetric matrices.

3. **Real Spectral Decomposition**: Symmetric matrices can be decomposed into the product of an orthogonal matrix and a diagonal matrix containing the eigenvalues. This decomposition is known as the spectral decomposition or eigenvalue decomposition.

4. **Positive Definiteness**: Symmetric matrices have a special property called positive definiteness. A symmetric matrix $A$ is positive definite if $x^T A x > 0$ for all non-zero vectors $x$. Positive definiteness is related to convexity, optimization, and other mathematical properties.

5. **Efficient Solving of Linear Systems**: Due to their special structure, symmetric matrices can often be solved more efficiently than general matrices. Algorithms such as the Cholesky decomposition and conjugate gradient methods are commonly used for solving linear systems involving symmetric positive definite matrices.

6. **Hermitian Property (in Complex Matrices)**: In the context of complex matrices, symmetric matrices are equivalent to Hermitian matrices, where $A = A^\dagger$, and $A^\dagger$ denotes the conjugate transpose of $A$.




##**Orthogonal matrix**
An orthogonal matrix is a square matrix whose rows and columns are orthogonal unit vectors. Here are some key properties of an orthogonal matrix:

1. **Orthogonality**: The columns (or rows) of an orthogonal matrix are orthogonal to each other. This means that their dot product is zero:
   $$ \mathbf{A}^T \mathbf{A} = \mathbf{I} $$
   Where:
   - $( \mathbf{A}^T )$ is the transpose of the matrix $( \mathbf{A} )$.
   - $( \mathbf{I} )$ is the identity matrix.

2. **Orthonormality**: In addition to being orthogonal, the columns (or rows) of an orthogonal matrix are also unit vectors, meaning they have a length of 1. Consequently, the inverse of an orthogonal matrix is equal to its transpose:
   $$ \mathbf{A}^{-1} = \mathbf{A}^T $$

3. **Preservation of Length**: Multiplying a vector by an orthogonal matrix preserves its length, i.e., the Euclidean norm. If $( \mathbf{x} )$ is a vector and $( \mathbf{Q} )$ is an orthogonal matrix, then:
   $$ \| \mathbf{Qx} \| = \| \mathbf{x} \| $$

4. **Preservation of Angles**: Orthogonal matrices also preserve angles between vectors. If $( \mathbf{x} )$ and $( \mathbf{y} )$ are two vectors, then the angle between them is the same as the angle between $( \mathbf{Qx} )$ and $( \mathbf{Qy} )$.

5. **Rotation**: Orthogonal matrices often represent rotations, reflections, or combinations of these transformations in Euclidean space.

6. **Determinant**: The determinant of an orthogonal matrix is either +1 or -1. If it's +1, it represents a proper (rotation) orthogonal matrix. If it's -1, it represents an improper (reflection) orthogonal matrix.

7. **Eigenvalues**: The eigenvalues of an orthogonal matrix have a magnitude of 1.




## **Diagonal Matrix**

A diagonal matrix is a square matrix in which all the elements outside the main diagonal are zero.

1. **Main Diagonal**: The elements of a diagonal matrix are non-zero only along the main diagonal from the top left to the bottom right.

2. **Representation**: If \( D \) is a diagonal matrix, its representation is:
   \[ D = \begin{bmatrix} d_{11} & 0 & \cdots & 0 \\
                          0 & d_{22} & \cdots & 0 \\
                          \vdots & \vdots & \ddots & \vdots \\
                          0 & 0 & \cdots & d_{nn} \end{bmatrix} \]
   where \( d_{11}, d_{22}, \ldots, d_{nn} \) are the diagonal elements.

3. **Eigenvalues**: The eigenvalues of a diagonal matrix are its diagonal elements, i.e., \( \lambda_i = d_{ii} \) for \( i = 1, 2, \ldots, n \).

4. **Determinant**: The determinant of a diagonal matrix is the product of its diagonal elements:
   \[ \text{det}(D) = \prod_{i=1}^n d_{ii} \]

5. **Trace**: The trace of a diagonal matrix is the sum of its diagonal elements:
   \[ \text{tr}(D) = \sum_{i=1}^n d_{ii} \]

6. **Inverse**: A diagonal matrix is invertible if and only if all its diagonal elements are non-zero. If \( D \) is invertible, then its inverse \( D^{-1} \) is obtained by taking the reciprocal of each diagonal element.

7. **Product with Scalar**: Multiplying a diagonal matrix \( D \) by a scalar \( c \) results in a diagonal matrix where each diagonal element is multiplied by \( c \).



## **Eigen Decomposition**

Eigen decomposition is a fundamental concept in linear algebra that allows us to decompose a square matrix into its constituent parts. At its core, it reveals the underlying structure of the matrix by expressing it in terms of its eigenvalues and eigenvectors.

### Mathematical Formula

Given a square matrix $A$, eigen decomposition expresses it as:

$$ A = Q \Lambda Q^{-1} $$

Where:
- $Q$ is a matrix whose columns are the eigenvectors of $A$.
- $\Lambda$ is a diagonal matrix containing the eigenvalues of $A$.

### Computation Steps

1. **Compute Eigenvalues and Eigenvectors**:
    - Solve the characteristic equation $|A - \lambda I| = 0$ to find the eigenvalues $\lambda$.
    - For each eigenvalue, solve the equation $(A - \lambda I)v = 0$ to find the corresponding eigenvectors $v$.

2. **Construct Matrix $Q$**:
    - Arrange the eigenvectors as columns of matrix $Q$.

3. **Construct Diagonal Matrix $\Lambda$**:
    - Place the eigenvalues along the diagonal of $\Lambda$, with other entries being zero.

4. **Verify**:
    - Ensure that $A = Q \Lambda Q^{-1}$ holds true.

### Importance

- **Spectral Analysis**: Eigen decomposition helps in understanding the behavior of linear transformations and systems.
- **Dimensionality Reduction**: Techniques like Principal Component Analysis (PCA) rely on eigen decomposition.
- **Solving Differential Equations**: Eigen decomposition simplifies the solution of systems of ordinary differential equations.

Eigen decomposition is a powerful tool with applications in various fields such as physics, engineering, computer science, and more.



### **Property of Eigen Decomposition**

Given a square matrix $\mathbf{A}$, its eigen decomposition is represented as:

$$ \mathbf{A} = \mathbf{Q} \mathbf{\Lambda} \mathbf{Q}^{-1} $$

where:
- $\mathbf{Q}$ is the matrix whose columns are the eigenvectors of $\mathbf{A}$,
- $\mathbf{\Lambda}$ is a diagonal matrix containing the eigenvalues of $\mathbf{A}$.

**Properties**:

1. **Existence**: If $\mathbf{A}$ has $n$ linearly independent eigenvectors, it can be decomposed into $\mathbf{Q}$ and $\mathbf{\Lambda}$.
   
2. **Unique up to Ordering**: The eigenvalues in $\mathbf{\Lambda}$ are unique but may be ordered differently.

3. **Inverse and Transpose**: The inverse and transpose of $\mathbf{A}$ can be expressed using the eigenvectors and eigenvalues as follows:
   - Inverse: $\mathbf{A}^{-1} = \mathbf{Q} \mathbf{\Lambda}^{-1} \mathbf{Q}^{-1}$
   - Transpose: $\mathbf{A}^T = \mathbf{Q} \mathbf{\Lambda} \mathbf{Q}^T$

4. **Matrix Power**: For any non-negative integer $k$, $\mathbf{A}^k = \mathbf{Q} \mathbf{\Lambda}^k \mathbf{Q}^{-1}$.

5. **Trace and Determinant**: The trace and determinant of $\mathbf{A}$ can be computed using its eigenvalues:
   - Trace: $ \text{Tr}(\mathbf{A}) = \sum_{i=1}^{n} \lambda_i $
   - Determinant: $ |\mathbf{A}| = \prod_{i=1}^{n} \lambda_i $

6. **Powers of $\mathbf{A}$**: $\mathbf{A}^k$ can be efficiently computed using eigen decomposition.

7. **Condition Number**: The condition number of $\mathbf{A}$ is the ratio of the largest to the smallest eigenvalue.

Eigen decomposition is fundamental in various numerical algorithms, including principal component analysis (PCA), diagonalization of matrices, solving differential equations, and many others.




### **Singular Value Decomposition (SVD):**

SVD is a matrix factorization method widely used in various fields including statistics, signal processing, and machine learning. It decomposes any given matrix $A$ into three other matrices, typically represented as:

$$
A = U \Sigma V^T
$$

Where:
- $A$ is the original matrix of size $m \times n$.
- $U$ is an orthogonal matrix of size $m \times m$.
- $\Sigma$ is a diagonal matrix of size $m \times n$, containing the singular values of $A$.
- $V^T$ is the transpose of an orthogonal matrix of size $n \times n$.

### **Mathematical Concept and Formula with Derivation**

Given a matrix $A$, the goal of SVD is to find orthogonal bases in the input and output spaces such that the transformation of $A$ can be represented in a diagonalized form. The mathematical concept involves finding the eigenvalues and eigenvectors of $A^TA$ and $AA^T$.

1. **Derivation of $U$**: $U$ is formed from the eigenvectors of $AA^T$. Let $AA^T = U \Sigma U^T$, where $U$ contains the eigenvectors of $AA^T$. The columns of $U$ are the left singular vectors of $A$.

2. **Derivation of $V$**: $V$ is formed from the eigenvectors of $A^TA$. Let $A^TA = V \Sigma^T V^T$, where $V$ contains the eigenvectors of $A^TA$. The columns of $V$ are the right singular vectors of $A$.

3. **Derivation of $\Sigma$**: The singular values, represented by the diagonal elements of $\Sigma$, are the square roots of the eigenvalues of both $AA^T$ and $A^TA$. They are ordered in descending order.

### Use in PCA

Principal Component Analysis (PCA) is a technique used for dimensionality reduction and data visualization. SVD plays a crucial role in PCA:

1. **Data Matrix**: Given a data matrix $X$ of size $m \times n$, where $m$ is the number of samples and $n$ is the number of features, SVD is applied to $X$ to decompose it into three matrices: $X = U \Sigma V^T$.

2. **Dimensionality Reduction**: PCA aims to find the principal components, which are the directions of maximum variance in the data. These principal components are equivalent to the columns of $V$. By selecting the top $k$ columns of $V$, where $k$ is the desired reduced dimensionality, we effectively reduce the dimensionality of the data.

3. **Data Reconstruction**: The original data can be approximately reconstructed from the reduced representation by multiplying the selected principal components with the corresponding rows of $\Sigma$ and $U^T$.

In summary, SVD provides a powerful tool for decomposing matrices, extracting important features, and performing dimensionality reduction, making it a fundamental component of techniques like PCA in data analysis and machine learning.

In Singular Value Decomposition (SVD), $V$ represents the matrix of right singular vectors.

Given a matrix $A$ of size $m \times n$, the SVD of $A$ is represented as:

$$
A = U \Sigma V^T
$$

Where:
- $U$ is an $m \times m$ orthogonal matrix containing the left singular vectors.
- $\Sigma$ is an $m \times n$ diagonal matrix containing the singular values of $A$.
- $V^T$ (or $V$ transposed) is an $n \times n$ orthogonal matrix containing the right singular vectors.

The matrix $V$ contains the orthonormal basis for the column space of $A$, and it provides information about the relationships among the columns of $A$. Each column of $V$ corresponds to a right singular vector of $A$, and these vectors form a basis for the space spanned by the columns of $A$.

In applications like dimensionality reduction and data compression, the matrix $V$ is often used to represent the principal directions or components of the data. These components capture the directions of maximum variance in the data, making them useful for various analytical tasks.


### **Properties of Singular Value Decomposition (SVD)**

Singular Value Decomposition (SVD) is a factorization of a matrix $A$ into three matrices $U$, $\Sigma$, and $V^T$, where $U$ and $V$ are orthogonal matrices and $\Sigma$ is a diagonal matrix containing the singular values of $A$. The SVD of a matrix $A$ is given by:

$$ A = U \Sigma V^T $$

#### Properties:

1. **Existence**: Every $m \times n$ matrix $A$ has an SVD.

2. **Uniqueness**: The SVD of a matrix $A$ is unique.

3. **Orthogonality**:
   - $U$ and $V$ are orthogonal matrices:
     - $UU^T = I$, $U^TU = I$
     - $VV^T = I$, $V^TV = I$

4. **Diagonal Singular Values**: $\Sigma$ is a diagonal matrix with non-negative singular values sorted in descending order along the diagonal.

5. **Rank**: The rank of $A$ is equal to the number of non-zero singular values in $\Sigma$.

6. **Spectral Norm**: The spectral norm of $A$ is equal to the largest singular value.

7. **Condition Number**: The condition number of $A$ is the ratio of the largest singular value to the smallest non-zero singular value.

8. **Low-rank Approximation**: SVD provides an optimal low-rank approximation of the original matrix $A$.

9. **Inverse**: If $A$ has full column rank, then $A^+$, the Moore-Penrose pseudoinverse of $A$, is given by $A^+ = V \Sigma^+ U^T$, where $\Sigma^+$ is the pseudoinverse of $\Sigma$.

10. **Applications**: SVD is widely used in data compression, noise reduction, dimensionality reduction, and solving linear least squares problems.


**To find eigenvalues and eigenvectors of a non-square matrix, you would typically use a method called singular value decomposition (SVD). Here's a step-by-step explanation:**

### Singular Value Decomposition (SVD)

1. **Given a non-square matrix $A$**, typically with dimensions $m \times n$ where $m \neq n$, we want to find its eigenvalues and eigenvectors.

2. **Form the product of $A$ and its transpose**:  
   Calculate $A^T \cdot A$. This will give you a square matrix $B$ of size $n \times n$.

3. **Calculate the eigenvalues and eigenvectors of $B$**:  
   Solve the characteristic equation $|B - \lambda I| = 0$ to find the eigenvalues $\lambda_i$. Then for each eigenvalue $\lambda_i$, solve the equation $(B - \lambda_i I)x = 0$ to find the corresponding eigenvector $x_i$.

4. **Compute the singular values and singular vectors**:  
   The singular values $\sigma_i$ of $A$ are the square roots of the eigenvalues of $B$. Sort them in descending order, usually denoted as $\sigma_1, \sigma_2, \ldots, \sigma_k$, where $k$ is the rank of $A$.

5. **Find the left singular vectors**:  
   The left singular vectors $u_i$ of $A$ are the eigenvectors of $AA^T$. These correspond to the eigenvalues of $B$.

6. **Find the right singular vectors**:  
   The right singular vectors $v_i$ of $A$ are the eigenvectors of $A^TA$. These are also the eigenvectors of $B$.

7. **Forming eigenvectors from singular vectors**:  
   If $A$ is of size $m \times n$ and rank $k$, the eigenvectors of $A$ can be formed as linear combinations of $u_i$ and $v_i$. Specifically, for each $i$, the eigenvector corresponding to the $i$-th largest singular value $\sigma_i$ is given by $Av_i = \sigma_i u_i$.

### Example
Let's say we have a non-square matrix $A$ of size $3 \times 2$:
$$
A = \begin{bmatrix} 1 & 2 \\ 3 & 4 \\ 5 & 6 \end{bmatrix}
$$

1. Calculate $A^TA$ and $AA^T$.
2. Find the eigenvalues and eigenvectors of $A^TA$ and $AA^T$.
3. Compute the singular values and singular vectors.
4. Form eigenvectors from singular vectors using $Av_i = \sigma_i u_i$.

This procedure allows you to find the eigenvalues and eigenvectors of non-square matrices through singular value decomposition.



## Simple Eigenvector Calculation vs. Rayleigh Quotients

### Simple Eigenvector Calculation

In simple eigenvector calculation, we deal with square matrices $A$ and seek vectors $v$ such that:

$$
A v = \lambda v
$$

Where $A$ is the matrix, $v$ is the eigenvector, and $\lambda$ is the eigenvalue associated with that eigenvector.

To find eigenvectors and eigenvalues, we usually solve the characteristic equation:

$$
\det(A - \lambda I) = 0
$$

Where $I$ is the identity matrix.

### Rayleigh Quotients

Rayleigh quotients are a method to approximate eigenvalues and eigenvectors iteratively. The Rayleigh quotient $R(v)$ for a vector $v$ is defined as:

$$
R(v) = \frac{{v^T Av}}{{v^T v}}
$$

Where $v^T$ denotes the transpose of $v$.

The key property of Rayleigh quotients is that for the eigenvector $v$ corresponding to the eigenvalue $\lambda$, $R(v)$ equals $\lambda$. In other words:

$$
R(v) = \lambda
$$

This property forms the basis of the iterative algorithm to find eigenvalues and eigenvectors. By iteratively refining $v$ based on the Rayleigh quotient, we can converge towards the true eigenvectors and eigenvalues of the matrix $A$.

### Difference

The main difference lies in their computational approach. Simple eigenvector calculation directly solves for eigenvalues and eigenvectors through the characteristic equation, which can be computationally intensive, especially for large matrices.

On the other hand, Rayleigh quotients provide an iterative approach that can converge to eigenvalues and eigenvectors with fewer computational resources, making them particularly useful for large-scale problems.

In summary, while simple eigenvector calculation provides exact solutions, Rayleigh quotients offer an efficient iterative method for approximating eigenvectors and eigenvalues.


### Which One is Best for PCA?

Principal Component Analysis (PCA) is a technique used for dimensionality reduction and data analysis. In PCA, you're interested in finding the eigenvectors (principal components) of the covariance matrix of your data.

For PCA, using simple eigen vector calculation is the most straightforward approach. You compute the eigenvectors and eigenvalues of the covariance matrix directly, and these eigenvectors represent the directions of maximum variance in the data. The eigenvalues indicate the amount of variance explained by each eigenvector.

While Rayleigh quotients can be useful for certain iterative methods and numerical approximations, they're not typically used directly in PCA. PCA primarily relies on the eigenvectors obtained through direct computation, as they provide a clear interpretation of the principal components of the data.





### Rayleigh Quotient for Eigenvalues

The Rayleigh quotient is a useful concept in linear algebra, particularly in the context of eigenvalues and eigenvectors. Given a square matrix $A$ and a non-zero vector $x$, the Rayleigh quotient is defined as:

$$
R(x) = \frac{x^T A x}{x^T x}
$$

Where:
- $x^T$ denotes the transpose of vector $x$.
- $A$ is the square matrix.
- $x$ is the non-zero vector.

The Rayleigh quotient provides a way to approximate the eigenvalues of a matrix. In particular, if $x$ is an eigenvector of $A$, then $R(x)$ will be equal to the corresponding eigenvalue. This property makes it useful for iterative methods for computing eigenvalues.

#### Steps to Compute:

1. **Choose a Non-zero Vector:** Start by selecting a non-zero vector $x$. This vector can be chosen randomly or through some heuristic if a specific eigenvector is sought.

2. **Compute the Rayleigh Quotient:** Calculate the Rayleigh quotient using the formula provided above. This involves multiplying $A$ with $x$, then taking the dot product of the resulting vector with $x$, and dividing by the dot product of $x$ with itself.

3. **Repeat or Refine:** Depending on the context, you may need to refine your vector $x$ or iterate the process with different vectors to get a better approximation of the eigenvalue.

#### Example Computation:

Let's say we have a matrix $A$:

$$
A = \begin{pmatrix} 2 & 1 \\ 1 & 3 \end{pmatrix}
$$

And we choose a non-zero vector:

$$
x = \begin{pmatrix} 1 \\ 1 \end{pmatrix}
$$

We can compute the Rayleigh quotient as follows:

$$
R(x) = \frac{x^T A x}{x^T x} = \frac{\begin{pmatrix} 1 & 1 \end{pmatrix} \begin{pmatrix} 2 & 1 \\ 1 & 3 \end{pmatrix} \begin{pmatrix} 1 \\ 1 \end{pmatrix}}{\begin{pmatrix} 1 & 1 \end{pmatrix} \begin{pmatrix} 1 \\ 1 \end{pmatrix}}
$$

After performing the matrix multiplications and dot products, we would obtain the Rayleigh quotient $R(x)$, which would be an approximation of one of the eigenvalues of matrix $A$.
