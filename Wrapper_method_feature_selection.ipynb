### **Wrapper Methods for Feature Selection**

Wrapper methods for feature selection are a type of feature selection technique that employs a predictive model to evaluate and score different combinations of features. They are termed "wrapper" methods because they encase the feature selection process within the evaluation of a predictive model.

#### How Wrapper Methods Work:

1. **Subset Generation:**
   - Wrapper methods start by generating subsets of features. This can be accomplished through various strategies such as:
     - **Forward Selection:** Start with an empty set of features and iteratively add features one at a time based on their performance.
     - **Backward Elimination:** Begin with all features and iteratively remove the least significant features.
     - **Random Subset Generation:** Create subsets of features randomly.
   - The choice of subset generation method depends on the specific wrapper method being used.

2. **Subset Evaluation:**
   - Once a subset of features is generated, a predictive model is trained on this subset.
   - The model's performance is then evaluated, typically through techniques like cross-validation.
   - The performance of the model serves as an estimate of the quality of the features in the subset.

3. **Stopping Criterion:**
   - The process of generating and evaluating subsets of features is repeated until a stopping criterion is met.
   - Common stopping criteria include:
     - A predefined number of subsets evaluated.
     - A specific amount of time elapsed.
     - No improvement in model performance after a certain number of iterations.

#### Notable Features and Functionality:

- **Model-Based Evaluation:** Wrapper methods utilize the performance of a predictive model as the guiding criterion for feature selection.
- **Iterative Process:** The process involves iteratively generating subsets of features and evaluating them, potentially leading to an exhaustive search through the feature space.
- **Flexible Subset Generation:** Different methods can be employed to generate subsets, providing flexibility in exploring various combinations of features.
- **Stopping Criteria:** Allows for control over the termination of the feature selection process, preventing exhaustive search in large feature spaces.

#### Examples and Use Cases:

- **Predictive Modeling:** Wrapper methods are commonly used in machine learning tasks where predictive modeling is essential, such as classification or regression.
- **Dimensionality Reduction:** These methods are valuable in reducing the dimensionality of data by selecting the most relevant features, thus improving model performance and interpretability.
- **Feature Engineering:** Wrapper methods aid in identifying the subset of features that contribute the most to the predictive performance of a model, facilitating better feature engineering decisions.

By employing wrapper methods for feature selection, practitioners can enhance the performance and interpretability of their predictive models by focusing on the most informative features.

## Types of Wrapper Methods

Wrapper methods are feature selection techniques that evaluate different subsets of features by training models and selecting the subset that yields the best performance according to a specified evaluation criterion. There are several types of wrapper methods, each with its own approach to selecting feature subsets.

### 1. Forward Selection:
In forward selection, the algorithm starts with an empty set of features and iteratively adds one feature at a time. At each iteration, the algorithm evaluates the performance of the model with the added feature and selects the feature that maximizes performance. This process continues until a stopping criterion is met.

### 2. Backward Elimination:
Backward elimination begins with the full set of features and iteratively removes one feature at a time. At each iteration, the algorithm evaluates the performance of the model after removing a feature and selects the subset of features that maximizes performance. This process continues until a stopping criterion is met.

### 3. Recursive Feature Elimination (RFE):
RFE is an iterative feature selection technique that starts with the full set of features and progressively eliminates features based on their importance or contribution to the model. The algorithm trains a model on the full feature set, ranks the features based on their importance, and removes the least important feature. This process is repeated until the desired number of features is reached or until performance stops improving.

### 4. Bidirectional Search:
Bidirectional search combines forward selection and backward elimination by simultaneously adding and removing features from the model. At each iteration, the algorithm evaluates the performance of both adding and removing a feature and selects the subset of features that maximizes performance.

### 5. Exhaustive Search:
Exhaustive search evaluates all possible subsets of features and selects the subset that yields the best performance according to a specified evaluation criterion. This approach involves training and evaluating a model for every possible combination of features, making it computationally expensive for datasets with a large number of features.

Each type of wrapper method has its own advantages and disadvantages, and the choice of method depends on factors such as the size of the dataset, the computational resources available, and the specific goals of the analysis.


------
## Advantages of Wrapper Methods:
- **Accuracy:** Wrapper methods typically yield the best performing feature subset for a given machine learning algorithm. They leverage the predictive power of the algorithm itself during feature selection.
- **Interaction of Features:** These methods consider feature interactions. Unlike filter methods that assess features independently, wrapper methods evaluate subsets of features together. This enables the discovery of feature groups that collectively enhance model performance, even if individual features are weak predictors.

## Disadvantages of Wrapper Methods:
- **Computational Complexity:** Wrapper methods incur high computational costs. They generate and evaluate numerous feature subsets, making them time-consuming, especially for datasets with many features.
- **Risk of Overfitting:** There's a risk of overfitting as wrapper methods optimize the feature subset to maximize the performance of a specific machine learning model. This may lead to selecting a subset that performs well on training data but poorly on unseen data.
- **Model Specific:** The selected feature subset is optimized for a specific model used in the feature selection process. Consequently, it may not perform as effectively with a different type of model.

### Key Considerations:
- Wrapper methods offer high predictive accuracy and consider feature interactions but at the expense of increased computational complexity and potential overfitting. Additionally, the selected feature subset may be tailored to a specific model.

### **Forward and Backward Sequential Feature Selection**

Forward and backward sequential feature selection are techniques used in feature selection, a process where you choose a subset of relevant features to use in your model.

#### **Forward Sequential Feature Selection:**

**When to Use:** Use forward sequential feature selection when you have a large number of features and want to iteratively add the most informative features to your model.

**How it Works:**
- Start with an empty set of features.
- Iteratively add one feature at a time, selecting the feature that improves the model performance the most.
- Continue adding features until a stopping criterion is met.

**Pros:**
- Simple and easy to implement.
- Can quickly identify the most important features.

**Cons:**
- May not find the global optimal subset of features.
- Prone to overfitting if not careful with the stopping criterion.

#### B**ackward Sequential Feature Selection:**

**When to Use:** Use backward sequential feature selection when you have a large number of features and want to iteratively remove the least informative features from your model.

**How it Works:**
- Start with all features.
- Iteratively remove one feature at a time, selecting the feature whose removal improves the model performance the least.
- Continue removing features until a stopping criterion is met.

**Pros:**
- May find a more optimal subset of features compared to forward selection.
- Less prone to overfitting as features are removed iteratively.

**Cons:**
- Computationally more expensive than forward selection.
- May still not find the global optimal subset of features.

#### When to Choose:

- Choose forward sequential feature selection for faster selection when suspecting a small subset of features is important.
- Choose backward sequential feature selection for a more exhaustive search for the optimal subset, even if it's computationally expensive.

In practice, it's beneficial to try both methods and compare the resulting subsets and model performance. Additionally, consider more advanced techniques like recursive feature elimination (RFE) for feature selection.

### **Recursive Feature Elimination (RFE) for Feature Selection**

### Overview
- Recursive Feature Elimination (RFE) is a feature selection technique that recursively removes less important features from the dataset until the desired number of features is reached.
- It works by training the model on the full feature set, ranking the features based on their importance, and then removing the least important feature. This process is repeated iteratively until the desired number of features is achieved.

### Steps:
1. **Initialization:** Start with the full feature set.
2. **Model Training:** Train a predictive model (e.g., linear regression, support vector machine) using the full feature set.
3. **Feature Ranking:** Retrieve the feature importance scores or coefficients from the trained model.
4. **Feature Elimination:** Identify the least important feature based on the feature rankings.
5. **Remove Feature:** Remove the least important feature from the dataset.
6. **Check Stopping Criteria:** If the desired number of features is reached or a stopping criterion is met (e.g., a predefined threshold for performance), stop the process. Otherwise, go to the next step.
7. **Repeat:** Repeat steps 2-6 with the reduced feature set.
8. **Finalize:** Once the desired number of features is reached, finalize the feature selection process.

### Formula:
The formula for Recursive Feature Elimination (RFE) involves iteratively training the model and selecting the least important feature until the desired number of features is achieved.

### Example:
For example, in the case of linear regression, RFE works by:
- Training the linear regression model on the full feature set.
- Retrieving the coefficients of each feature.
- Removing the feature with the smallest coefficient (least importance).
- Repeating this process until the desired number of features is obtained.

### Advantages:
- Automatically selects the most relevant features.
- Can improve model performance by reducing overfitting and increasing interpretability.

### Disadvantages:
- Computationally expensive, especially for large datasets with a large number of features.
- Performance may vary depending on the choice of the underlying predictive model and the stopping criteria.

**Key Points:**
- **Iterative Process:** Gradually eliminates the least important features.
- **Feature Importance:** Relies on the notion of feature importance determined by the trained model.
- **Flexibility:** Can be used with various machine learning algorithms.
- **Hyperparameters:** Include the number of features to remove in each iteration and the desired number of final features.
- **Evaluation Metric:** Choice of evaluation metric is crucial.
- **Computational Complexity:** Can be computationally expensive for large datasets.

RFE is a powerful technique for improving model interpretability, reducing overfitting, and enhancing computational efficiency by selecting the most relevant subset of features.

### **Exhaustive Wrapper Method for Feature Selection**

The **exhaustive wrapper method** for feature selection is a technique employed in machine learning to identify the optimal subset of features for predictive modeling. Unlike other methods that rely on heuristics or statistical measures, exhaustive wrapper methods systematically evaluate all possible combinations of features to select the subset that maximizes model performance according to a specified evaluation criterion.

#### How It Works:

1. **Generate Feature Subsets**: Begin by generating all possible subsets of features from the original feature set. This process involves creating combinations of features of varying sizes, ranging from individual features to combinations of multiple features.

2. **Train and Evaluate Models**: For each feature subset generated, train a predictive model (e.g., a machine learning algorithm) using only the features within that subset. Subsequently, evaluate the performance of each trained model using a predetermined evaluation metric, such as accuracy, precision, recall, or F1-score.

3. **Select Optimal Subset**: Compare the performance of models trained on different feature subsets and identify the subset that yields the highest performance according to the evaluation criterion. This subset is deemed the optimal subset of features for the predictive model.

4. **Optional Refinement**: Depending on computational resources and time constraints, additional steps may be taken to further refine the selected subset. This could involve retraining the model on a smaller set of promising feature subsets or applying additional filtering criteria to eliminate redundant or irrelevant features.

5. **Final Model Training**: Once the optimal feature subset is determined, train the final predictive model using the selected features. This model is then ready for deployment and can be used to make predictions on new data.

#### Advantages:

- **Guaranteed Optimal Subset**: Exhaustive wrapper methods ensure that the best possible feature subset is identified according to the specified evaluation criterion.
  
- **Flexibility in Evaluation Criterion**: Users can define their own evaluation metric based on the task at hand, allowing for customization and adaptability.

#### Considerations:

- **Computational Expense**: This approach can be computationally expensive, particularly for datasets with a large number of features, as it involves training and evaluating a substantial number of models.
  
- **Resource and Time Constraints**: Users must consider the available computational resources and time constraints when employing exhaustive wrapper methods, as they may require significant time and computational power.

#### Common Implementations:

- **Recursive Feature Elimination (RFE)**: This algorithm iteratively removes the least significant features from the model until the optimal subset is identified.
  
- **Forward Selection**: In this approach, features are added to the model one at a time based on their impact on performance until the optimal subset is found.

#### Use Cases:

- **High-Dimensional Data**: Exhaustive wrapper methods are particularly useful when dealing with datasets containing a large number of features, where heuristic or filter-based methods may be less effective.
  
- **Custom Evaluation Metrics**: In scenarios where standard evaluation metrics may not fully capture the task's requirements, exhaustive wrapper methods allow for the use of customized evaluation criteria tailored to specific needs.

````python
df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
df.head()
````

````python
# Importing necessary modules for feature selection
from mlxtend.feature_selection import ExhaustiveFeatureSelector, SequentialFeatureSelector
# Importing Linear Regression model from scikit-learn
from sklearn.linear_model import LinearRegression

# Instantiating a Linear Regression model
lr = LinearRegression()

# Creating an Exhaustive Feature Selector object
# Exhaustive feature selection evaluates all possible feature combinations.
# Parameters:
# - model: the estimator to use for feature selection, here it's a Linear Regression model
# - max_features: maximum number of features to select
# - scoring: the scoring metric to use for evaluating feature subsets, here it's 'r2' (coefficient of determination)
# - cv: cross-validation strategy, here it's 10-fold cross-validation
# - print_progress: whether to print progress messages, here set to True
# - n_jobs: number of CPU cores to use for computation, here it's set to -1 to utilize all available cores
exh = ExhaustiveFeatureSelector(lr, max_features=13, scoring='r2', cv=10, print_progress=True, n_jobs=-1)

# Fitting the feature selector to the training data
sel = exh.fit(X_train, y_train)
````

##Adjusted r2_score
The provided markdown explanation breaks down a mathematical formula used to compute the adjusted R-squared score in the context of linear regression model evaluation. Let's summarize it:

### Description of the Formula
The formula for adjusted R-squared is:

```markdown
1 - (((1 - r2) * (n - 1)) / (n - p - 1))
```

Here's what each component likely represents:

- `r2`: Coefficient of determination or R-squared value.
- `n`: Number of observations or data points.
- `p`: Number of predictor variables in the regression model.

### Structure and Functionality
1. `(1 - r2)`: Subtracts the R-squared value from 1.
2. `* (n - 1)`: Multiplies the result by (n - 1), potentially related to degrees of freedom or adjustments in statistical computations.
3. `/ (n - p - 1)`: Divides the previous result by (n - p - 1), where (n - p - 1) represents the degrees of freedom adjustment.

### Interpretation
The adjusted R-squared score accounts for the number of predictor variables in the model, providing a more accurate assessment of model fit compared to the regular R-squared value, especially when dealing with multiple predictors.

### Use Cases
- Evaluating the goodness of fit of a linear regression model.
- Comparing different models with varying numbers of predictors.
- Adjusting for the complexity of the model when interpreting the R-squared value.

Overall, the adjusted R-squared score is a valuable metric for understanding the performance of regression models, particularly in situations involving multiple predictors.



````python

def adjusted_r2(r2, n, p):
    #Calculate adjusted R-squared value.
    return 1 - (((1 - r2) * (n - 1)) / (n - p - 1))
````

````python
# Extract the average score ('r2') from the metric dictionary.
r2 = metric["avg_score"]

# Get the number of samples in the training set ('n').
n = X_train.shape[0]

# Calculate the number of predictors excluding the intercept term ('p').
p = X_train.shape[1] - 1

# Compute the adjusted R-squared ('adjusted_r2') using the extracted average score,
# number of samples, and number of predictors.
metric["adjusted_r2"] = adjusted_r2(r2, n, p)

# Calculate the number of features selected ('no of feature') by applying a function
# to count the length of each feature index list in the "feature_idx" column.
metric["no of feature"] = metric["feature_idx"].apply(lambda x: len(x))

# Sort the metric dictionary based on the adjusted R-squared values in descending order
# to prioritize models with higher adjusted R-squared values.
metric.sort_values(by="adjusted_r2", ascending=False)
````

## Sequential Backward Selection (SBS) / Sequential Backward Elimination (SBE)

Sequential Backward Selection (SBS), also known as Sequential Backward Elimination (SBE), is a feature selection technique used to iteratively remove features from a full feature set until the optimal subset of features is obtained. Here's a summary of how it works:

1. **Start with All Features:**
   - Begin with a model that includes all features available in the dataset.

2. **Iterative Feature Removal:**
   - Remove one feature at a time from the current feature set.
   - Select the feature whose removal results in the smallest decrease in model performance.

3. **Evaluate Model Performance:**
   - After removing a feature, retrain the model using the reduced feature set.
   - Evaluate the model's performance using a chosen evaluation metric.

4. **Stopping Criterion:**
   - Repeat the iterative process until a stopping criterion is met.
   - Common stopping criteria include a predefined number of features to select or no improvement in model performance.

5. **Select Optimal Subset:**
   - Once the stopping criterion is met, select the subset of features that resulted in the best model performance.

**Pros of Sequential Backward Selection:**
- May find a more optimal subset of features compared to other feature selection methods.
- Less prone to overfitting as features are removed iteratively.
- Can be combined with any machine learning algorithm and evaluation metric.

**Cons of Sequential Backward Selection:**
- Computationally more expensive than some other feature selection methods, especially for datasets with a large number of features.
- Requires retraining the model multiple times, which can be time-consuming for complex models or large datasets.

**Example:**
Suppose you have a dataset with 10 features and want to use sequential backward selection to select the most relevant subset of features for a classification task. You start with all 10 features, iteratively remove one feature at a time based on the decrease in model performance, and continue until the optimal subset of features is obtained, satisfying a predefined stopping criterion. The final subset of features selected by SBS is then used to train the final model.

## Recursive Feature Elimination (RFE) for Feature Selection

### Overview
- Recursive Feature Elimination (RFE) is a feature selection technique that recursively removes less important features from the dataset until the desired number of features is reached.
- It works by training the model on the full feature set, ranking the features based on their importance, and then removing the least important feature. This process is repeated iteratively until the desired number of features is achieved.

**Key Points:**
- **Iterative Process:** Gradually eliminates the least important features.
- **Feature Importance:** Relies on the notion of feature importance determined by the trained model.
- **Flexibility:** Can be used with various machine learning algorithms.
- **Hyperparameters:** Include the number of features to remove in each iteration and the desired number of final features.
- **Evaluation Metric:** Choice of evaluation metric is crucial.
- **Computational Complexity:** Can be computationally expensive for large datasets.

RFE is a powerful technique for improving model interpretability, reducing overfitting, and enhancing computational efficiency by selecting the most relevant subset of features.

````python
df = pd.read_csv("https://raw.githubusercontent.com/selva86/datasets/master/BostonHousing.csv")
df.head()
````

````python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(df.iloc[:,:-1],df["medv"],random_state=2,test_size=.2)
````

````python
from sklearn.preprocessing import StandardScaler,OneHotEncoder
sc=StandardScaler()
xtrain=sc.fit_transform(xtrain)
xtest=sc.transform(xtest)
````

````python
# Importing the Recursive Feature Elimination (RFE) module from the scikit-learn library,
# which helps in selecting features by recursively considering smaller and smaller sets of features.
from sklearn.feature_selection import RFE

# This model will be used as an estimator within the RFE process.
lr = LinearRegression()

# Configuring the RFE instance with the specified parameters:

rf = RFE(estimator=lr, n_features_to_select=10, step=1)

# Fitting the RFE model to the training data.
# This process involves repeatedly fitting the model and eliminating the least important features
# until the desired number of features is selected.
rf.fit(xtrain, ytrain)
````
