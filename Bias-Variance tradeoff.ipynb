## **Bias and Variance in Machine Learning**

In machine learning, bias and variance are two fundamental concepts that help in understanding the performance and behavior of a predictive model. Here's a brief explanation of each:

### 1. Bias
- Bias refers to the error introduced by approximating a real-world problem with a simplified model. It represents the difference between the expected prediction of the model and the true value.
- A high bias model tends to underfit the training data, meaning it fails to capture the underlying patterns and relationships in the data.
- Models with high bias have a tendency to oversimplify the problem and make strong assumptions about the data, leading to poor performance on both the training and unseen data.

### 2. Variance
- Variance refers to the variability of model predictions for a given data point, or in other words, how much the predictions for a given point differ across different realizations of the model.
- A high variance model tends to overfit the training data, meaning it captures noise and random fluctuations in the training data as if they were genuine patterns.
- Models with high variance are highly flexible and can capture complex patterns in the training data, but they may fail to generalize well to unseen data due to their sensitivity to small fluctuations in the training data.

In summary, bias and variance are two sources of error that contribute to the overall prediction error of a model. The goal in machine learning is to find a balance between bias and variance, known as the bias-variance tradeoff, by selecting models that generalize well to unseen data without underfitting or overfitting.


### **Understanding Bias in Machine Learning**

In machine learning, Bias refers to the systematic error introduced by the model's assumptions during training, leading to discrepancies between predicted and expected values. Let's delve into what Bias is and how it influences model performance.

#### Definition
Bias can be succinctly defined as the disparity between the Predicted Value (output of the model) and the Expected Value (true outcome). When a model is trained, it forms assumptions about the data. However, these assumptions might not hold true when the model encounters unseen testing or validation data.

#### Influence on Model Behavior
Consider a scenario where a model, such as one employing the k-nearest neighbors algorithm, is trained with a large number of neighbors. This model might oversimplify the relationship between input features and the target variable. For instance, in predicting diabetes, it might only consider glucose level and blood pressure, neglecting other potential influential factors. This oversimplification leads to a high bias error, resulting in a model that fails to capture the nuances and variations present in the data.

## Mathematical Derivation of Bias in Machine Learning

Bias, in the context of machine learning, refers to the error introduced by approximating a real-world problem with a simplified model. Mathematically, bias is defined as the difference between the expected (average) prediction of our model and the true value we are trying to predict.

Let's denote the true target value as \( y \) and the predicted value by our model as $ ( \hat{f}(x) ) $. The bias can be expressed as:
$$
[
\text{Bias}[\hat{f}(x)] = E[\hat{f}(x)] - f(x)
]
$$

Where:
- $( E[\hat{f}(x)] ) $ represents the expected value (average) of the predictions made by our model over all possible training datasets.
- \( f(x) \) is the true (unknown) function that we are trying to approximate.

In simpler terms, bias measures how far off, on average, the predictions of our model are from the true values. A high bias indicates that the model is too simplistic and fails to capture the underlying patterns in the data, leading to underfitting.

Finding the right balance between bias and variance is crucial for building a machine learning model that generalizes well to new, unseen data.

#### Implications: Underfitting
High bias error leads to underfitting, where the model fails to grasp the intricacies of the training data. This results in overly simplistic predictions that do not adequately reflect the complexity of the underlying relationship between features and target variable. The model essentially overlooks relevant information, leading to suboptimal performance on unseen data.

#### Conclusion
Understanding bias is crucial in evaluating and improving machine learning models. It highlights the importance of mitigating oversimplification and ensuring models capture the nuances present in the data. By addressing bias, practitioners can enhance model accuracy and robustness, thereby improving decision-making processes in various domains.

#### Examples and Use Cases
- Predictive modeling in healthcare: Ensuring medical diagnosis models consider all relevant patient parameters to avoid oversimplified predictions.
- Financial forecasting: Incorporating diverse economic indicators to build accurate models that reflect market complexities and trends.

By considering bias in model development, practitioners can build more reliable and effective machine learning solutions across diverse domains.

### **Understanding Variance in Machine Learning**

Variance in the context of modeling refers to the sensitivity of a model to the fluctuations or noise in the training data. Unlike bias, which represents the model's simplifying assumptions about the underlying data, variance captures how much the model's predictions vary as new data is introduced.

### High Variance and its Implications

When a model exhibits high variance, it means that it is overly sensitive to the training data and captures not only the underlying patterns but also the noise present in the data. Consequently, when confronted with new, unseen data (testing data), the model may fail to generalize well and provide inaccurate predictions.

### Summary of Bias and Variance Errors

- **High Bias Error**: A model with high bias error tends to underfit the data, meaning it makes overly simplistic assumptions about the underlying patterns. It fails to capture the complexity of the data.
  
- **High Variance Error**: On the other hand, a model with high variance error tends to overfit the data. It learns too much from the training data, including the noise, and fails to generalize well to new data.
  
- **Balanced Model**: A good model strikes a balance between bias and variance errors. It captures the underlying patterns in the data without being overly sensitive to noise, resulting in accurate predictions on both training and testing data.

### Examples and Use Cases

- **Example**: Consider a polynomial regression model trained on a dataset with a few data points. If the model's degree is too high, it might fit the training data perfectly (low bias) but capture every fluctuation and noise in the data (high variance). As a result, it would fail to generalize well to new data.
  
- **Use Case**: In medical diagnosis, a machine learning model might be trained on patient data to predict the likelihood of a certain disease. If the model overfits the training data by capturing noise, it might lead to incorrect diagnoses for new patients, potentially harming their treatment outcomes. Therefore, it's crucial to ensure a balance between bias and variance errors in such critical applications.

## Mathematical Derivation of Variance in Machine Learning

Variance ($\text{Var}$) measures the variability of model predictions for a given data point. It quantifies how much the predictions differ from the mean prediction.

The formula for variance is given by:

$$
\text{Var}[X] = E[(X - E[X])^2]
$$

Where:
- $X$ is a random variable.
- $E[X]$ is the expected value (mean) of $X$.

### Derivation:

1. **Squared Deviation from the Mean**:
   We start by calculating the squared deviation of each observation from the mean:

   $$
   (X - E[X])^2
   $$

2. **Expectation of Squared Deviation**:
   Taking the expected value (mean) of the squared deviation gives us the variance:

   $$
   \text{Var}[X] = E[(X - E[X])^2]
   $$

This derivation shows that variance represents the average squared deviation of each observation from the mean, providing a measure of the spread or variability of the data.

In machine learning, variance is used to evaluate the model's sensitivity to fluctuations in the training data. High variance can indicate overfitting, where the model captures noise in the training data rather than the underlying patterns.


### **Bias-Variance Tradeoff in Machine Learning**

image.png

The Bias-Variance tradeoff is a fundamental concept in machine learning that deals with the balance between model simplicity and its ability to capture the complexity of the underlying data. It aims to find the optimal model complexity to minimize both bias and variance, which are two sources of error in predictive modeling.

#### Structure:

1. **Bias**: Bias refers to the error introduced by approximating a real-world problem with a simplified model. A high bias model tends to oversimplify the data and may fail to capture underlying patterns, resulting in underfitting. Models with high bias are typically too rigid and unable to learn from the training data effectively.

2. **Variance**: Variance measures the model's sensitivity to fluctuations in the training dataset. High variance models are overly complex and tend to capture noise in the training data, leading to overfitting. Overfitting occurs when the model performs well on the training data but fails to generalize to unseen data.

#### How it works:

- **Underfitting**: When a model has high bias and low variance, it underfits the data. This means the model is too simplistic to capture the underlying structure of the data, resulting in poor performance both on the training and test datasets.

- **Overfitting**: On the other hand, when a model has low bias and high variance, it overfits the data. This means the model is too complex and captures noise or random fluctuations in the training data, leading to excellent performance on the training set but poor performance on unseen data.

- **Tradeoff**: The goal of machine learning is to find the right balance between bias and variance to achieve optimal model performance. This involves selecting a model with an appropriate level of complexity that generalizes well to unseen data while capturing the underlying patterns present in the training data.

#### Notable Features/Functionality:

- **Cross-Validation**: Cross-validation techniques such as k-fold cross-validation help in assessing the bias-variance tradeoff by providing estimates of the model's performance on unseen data.

- **Regularization**: Techniques like Lasso and Ridge regression introduce penalties to the model's complexity, helping to reduce overfitting and control variance.

#### Examples and Use Cases:

- **Polynomial Regression**: In polynomial regression, the choice of polynomial degree represents a tradeoff between bias and variance. Lower degree polynomials have high bias but low variance, while higher degree polynomials have low bias but high variance.

- **Decision Trees**: Decision trees can easily overfit the training data if they are allowed to grow too deep. Pruning techniques and limiting the depth of the tree are ways to control overfitting and balance bias and variance.

- **Support Vector Machines (SVM)**: SVM models with a large margin may have low bias but high variance, while models with a smaller margin may have high bias but low variance. Adjusting the regularization parameter helps in finding the right balance.

In summary, the bias-variance tradeoff is a crucial concept in machine learning, guiding model selection and parameter tuning to achieve optimal predictive performance. It emphasizes the importance of finding the right balance between model simplicity and complexity to generalize well to unseen data.

##**bias-variance decomposition**
The bias-variance decomposition is a fundamental concept in machine learning that decomposes the expected prediction error into three components: bias, variance, and irreducible error. Let's denote the true function as $f(x)$, the estimated function as $\hat{f}(x)$, and the noise as $\varepsilon$. The mean squared error (MSE) of the model is defined as:

$$
\text{MSE} = \mathbb{E}[(y - \hat{f}(x))^2]
$$

Expanding the square term yields:

$$
\text{MSE} = \mathbb{E}[y^2 - 2y\hat{f}(x) + \hat{f}^2(x)]
$$

Using linearity of expectation, this can be written as:

$$
\text{MSE} = \mathbb{E}[y^2] - 2\mathbb{E}[y\hat{f}(x)] + \mathbb{E}[\hat{f}^2(x)]
$$

Since $y = f(x) + \varepsilon$, we have:

$$
\mathbb{E}[y^2] = \mathbb{E}[(f(x) + \varepsilon)^2] = \mathbb{E}[f^2(x)] + 2f(x)\mathbb{E}[\varepsilon] + \mathbb{E}[\varepsilon^2]
$$

Since $\varepsilon$ has zero mean ($\mathbb{E}[\varepsilon] = 0$) and variance $\sigma^2$, we get:

$$
\mathbb{E}[y^2] = f^2(x) + \sigma^2
$$

Similarly, we have:

$$
\mathbb{E}[y\hat{f}(x)] = \mathbb{E}[(f(x) + \varepsilon)\hat{f}(x)] = \mathbb{E}[f(x)\hat{f}(x)] + \mathbb{E}[\varepsilon\hat{f}(x)]
$$

Since $\varepsilon$ is independent of $\hat{f}(x)$, we have $\mathbb{E}[\varepsilon\hat{f}(x)] = 0$. Thus:

$$
\mathbb{E}[y\hat{f}(x)] = f(x)\mathbb{E}[\hat{f}(x)]
$$

Finally, we have:

$$
\mathbb{E}[\hat{f}^2(x)] = \text{Var}[\hat{f}(x)] + (\mathbb{E}[\hat{f}(x)])^2
$$

Combining these results, the MSE can be expressed as:

$$
\text{MSE} = (f(x) - \mathbb{E}[\hat{f}(x)])^2 + \text{Var}[\hat{f}(x)] + \sigma^2
$$

This decomposition shows that the MSE is comprised of three terms: the squared bias, the variance, and the irreducible error. The bias term measures how much the estimated function $\hat{f}(x)$ differs from the true function $f(x)$. The variance term measures how much the estimated function $\hat{f}(x)$ varies across different training datasets. The irreducible error represents the noise inherent in the data that cannot be reduced by any model.


## Variance

Variance captures how much your classifier changes if you train on a different training set. It measures how "over-specialized" your classifier is to a particular training set, indicating the degree of overfitting. Essentially, it quantifies how much the performance of your model varies across different datasets. If you have the best possible model for your training data, variance tells you how far off you are from the average classifier.

## Bias

Bias refers to the inherent error that you obtain from your classifier even with infinite training data. This error arises because your classifier is "biased" to a particular kind of solution, such as a linear classifier. In other words, bias is a characteristic of your model that leads it to consistently underpredict or overpredict the true values. It represents the discrepancy between the predictions made by your model and the true underlying values.

## Noise

Noise quantifies how big the data-intrinsic noise is. This error measures the ambiguity inherent in your data distribution and feature representation. Noise reflects the inherent uncertainty or randomness in the data that cannot be eliminated by any model. It is an essential aspect of the data that cannot be overcome, and models must learn to deal with it effectively.

```python
from mlxtend.evaluate import bias_variance_decomp
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from mlxtend.data import boston_housing_data
from sklearn.model_selection import train_test_split


X, y = boston_housing_data()
X_train, X_test, y_train, y_test = train_test_split(X, y,
                                                    test_size=0.3,
                                                    random_state=123,
                                                    shuffle=True)



```

```python
lr = LinearRegression()

avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        lr, X_train, y_train, X_test, y_test,
        loss='mse',
        random_seed=123)

print('Average expected loss: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```

```python
dt = DecisionTreeRegressor(random_state=123)

avg_expected_loss, avg_bias, avg_var = bias_variance_decomp(
        dt, X_train, y_train, X_test, y_test,
        loss='mse',
        random_seed=123)

print('Average expected loss: %.3f' % avg_expected_loss)
print('Average bias: %.3f' % avg_bias)
print('Average variance: %.3f' % avg_var)
```
