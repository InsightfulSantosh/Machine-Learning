## **What is Unsupervised Learning?**

Unsupervised learning is a type of machine learning where the algorithm learns patterns from data without being explicitly told what to look for or what the patterns mean. Think of it like a child sorting a box of toys.

Imagine a kid has a box full of different toys: there are cars, dolls, balls, and blocks. No one tells the kid how to sort them, but the kid starts to see patterns. She might notice that some toys have wheels and others don't, so she puts cars in one pile and everything else in another. Or, she might see that some toys are soft (like dolls) and others are hard (like blocks), and she sorts them this way.

In this example, the kid is the "algorithm" and the toys are the "data." Just like the kid figures out ways to sort the toys into groups based on what she observes (wheels, softness), unsupervised learning algorithms sort data into groups based on patterns they find, without being told what these patterns represent. The key point is that the kid (or the algorithm) is figuring out the patterns all by herself, without any guidance or instructions on how to sort the toys (or the data).

## <b><u> Introduction to k-means clustering </u></b>

Clustering algorithms seek to learn, from the properties of the data, an optimal division or discrete labeling of groups of points.Many clustering algorithms are available in Scikit-Learn and elsewhere, but perhaps the simplest to understand is an algorithm known as *k-means clustering*, which is implemented in ``sklearn.cluster.KMeans``.

The *k*-means algorithm searches for a pre-determined number of clusters within an unlabeled multidimensional dataset. It accomplishes this using a simple conception of what the optimal clustering looks like:

- The "cluster center" is the arithmetic mean of all the points belonging to the cluster.
- Each point is closer to its own cluster center than to other cluster centers.

These two assumptions are the basis of the *k*-means model.

First, let's generate a two-dimensional dataset containing four distinct blobs.To emphasize that this is an unsupervised algorithm, we will leave the labels out of the visualization.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()  # for plot styling
import numpy as np
```

```python
from sklearn.datasets import make_blobs
X, y_true = make_blobs(n_samples=300, centers=4,
                       cluster_std=0.60, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50)
```

By eye, it is relatively easy to pick out the four clusters.The *k*-means algorithm does this automatically:

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
```

Let's visualize the results by plotting the data colored by these labels. We will also plot the cluster centers as determined by the *k*-means estimator:

```python
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
```

```python
kmeans.cluster_centers_
```

The good news is that the *k*-means algorithm (at least in this simple case) assigns the points to clusters very similarly to how we might assign them by eye.But you might wonder how this algorithm finds these clusters so quickly! After all, the number of possible combinations of cluster assignments is exponential in the number of data pointsan exhaustive search would be very, very costly. Fortunately for us, such an exhaustive search is not necessary: instead, the typical approach to *k*-means involves an intuitive iterative approach known as *expectationmaximization*.

### <b><u> k-Means Algorithm: ExpectationMaximization </u></b>

Expectationmaximization (EM) is a powerful algorithm that comes up in a variety of contexts within data science. *k*-means is a particularly simple and easy-to-understand application of the algorithm.

In short, the expectationmaximization approach here consists of the following procedure:

1. Guess some cluster centers
2. Repeat until converged
   1. *E-Step*: assign points to the nearest cluster center
   2. *M-Step*: set the cluster centers to the mean

Here the "E-step" or "Expectation step" is so-named because it involves updating our expectation of which cluster each point belongs to.

The "M-step" or "Maximization step" is so-named because it involves maximizing some fitness function that defines the location of the cluster centers  in this case, that maximization is accomplished by taking a simple mean of the data in each cluster.

The literature about this algorithm is vast, but can be summarized as follows: under typical circumstances, each repetition of the E-step and M-step will always result in a better estimate of the cluster characteristics.

### **Silhouette Analysis:**

#### <b><u> Selecting the number of clusters with silhouette analysis on KMeans clustering </u></b>

Silhouette analysis can be used to study the separation distance between the resulting clusters. The silhouette plot displays a measure of how close each point in one cluster is to points in the neighboring clusters and thus provides a way to assess parameters like number of clusters visually. This measure has a range of [-1, 1].

Silhouette coefficients (as these values are referred to as) near +1 indicate that the sample is far away from the neighboring clusters. A value of 0 indicates that the sample is on or very close to the decision boundary between two neighboring clusters and negative values indicate that those samples might have been assigned to the wrong cluster.

### <b> Calculation of Silhouette score </b>

#### The Silhouette Coefficient for a sample is $ S = \frac{(b - a)}{max(a, b)}$.

Silhouette score is used to evaluate the quality of clusters created using clustering algorithms such as K-Means in terms of how well samples are clustered with other samples that are similar to each other. The Silhouette score is calculated for each sample of different clusters. To calculate the Silhouette score for each observation/data point, the following distances need to be found out for each observations belonging to all the clusters:

* Mean distance between the observation and all other data points in the same cluster. This distance can also be called a mean intra-cluster distance. The mean distance is denoted by a.
* Mean distance between the observation and all other data points of the next nearest cluster. This distance can also be called a mean nearest-cluster distance. The mean distance is denoted by b.

**Example:**

Suppose we have a dataset of eight points, and we've clustered them into two clusters:

- **Cluster 1:** Points A, B, C, D
- **Cluster 2:** Points E, F, G, H

We'll calculate the silhouette score for one of these points, say point A, step by step:

1. **Calculate the Average Distance to Points in the Same Cluster (a):**
   - First, we calculate the average distance of point A to the other points in Cluster 1 (B, C, D). This distance is a measure of how well A is grouped with these points. Suppose this average distance is 2 units.

2. **Calculate the Average Distance to Points in the Nearest Cluster (b):**
   - Next, we calculate the average distance of point A to the points in Cluster 2 (E, F, G, H). This distance measures how different A is from the points in the nearest other cluster. Let's say this average distance is 5 units.

3. **Calculate the Silhouette Score for Point A:**
   - The silhouette score \( S \) for a point is calculated as $( S = \frac{(b - a)}{\max(a, b)})$.
   - For point A, \( a = 2 \) and \( b = 5 \), so $( S = \frac{(5 - 2)}{\max(2, 5)} = \frac{3}{5} = 0.6)$.

4. **Interpretation:**
   - Since the silhouette score for A is 0.6, which is closer to 1 than to 0, it suggests that A is well matched to its own cluster and quite distinct from the nearest cluster. This is a sign of good clustering for point A.

5. **Calculate Silhouette Scores for All Points and Find the Average:**
   - We would repeat this process for each point in the dataset. After calculating the individual scores, we would average them to understand the overall quality of the clustering.

6. **Overall Clustering Interpretation:**
   - If the average silhouette score is high (close to 1), it indicates that, on average, points are well matched to their own cluster and distinct from other clusters.
   - If the average score is around 0 or negative, it suggests overlapping clusters or points being assigned to incorrect clusters.

This example simplifies the process for clarity. In practice, especially with larger and more complex datasets, these calculations can become more intricate and are usually performed using programming libraries.

### **Elbow Method:**

The Elbow Method is a popular technique used to determine the optimal number of clusters in k-means clustering. The method plots a graph that helps you visually identify the best number of clusters for your data. Here's how it works:

1. **Plotting Criteria**: In the Elbow Method, you plot the number of clusters on the x-axis and a measure of clustering quality on the y-axis. The most common measure for the y-axis is the Within-Cluster Sum of Squares (WCSS), which is the sum of the squared distances between each data point and the centroid of its assigned cluster. The idea is to measure how compact the clusters are, with lower values of WCSS indicating more tightly grouped data.

2. **Running the Algorithm**: You run the k-means clustering algorithm on your dataset for a range of cluster numbers (e.g., 1 to 10). For each number of clusters, you calculate the WCSS.

3. **Creating the Elbow Plot**: Plot these WCSS values against the number of clusters on a graph.

4. **Interpreting the Plot**: The plot typically shows a rapid decrease in WCSS as the number of clusters increases, but after a certain point, the rate of decrease sharply slows down, creating an "elbow" in the graph. The number of clusters at this "elbow" point is considered to be the optimal number. This is because adding more clusters beyond this point doesn't provide much better modeling of the data.

   - **Before the Elbow**: Large decreases in WCSS signify that adding more clusters significantly improves the fit of the model to the data.
   - **At the Elbow**: The elbow point indicates that the benefit of adding an additional cluster starts to diminish.
   - **After the Elbow**: Further increases in the number of clusters result in smaller reductions in WCSS, suggesting that these additional clusters may not be capturing substantial structures in the data.

5. **Example**: Suppose you plot the WCSS for cluster numbers 1 through 10. You may notice that from 1 to 4, the WCSS decreases quickly. But from 4 to 5, the decrease in WCSS is less sharp, and beyond 5, changes in WCSS are minimal. In this case, the "elbow" is at 4, suggesting that 4 is the optimal number of clusters.

The Elbow Method is a heuristic and doesn't always yield a clear-cut elbow, but it's a useful starting point for identifying a reasonable number of clusters in many datasets.

The Within-Cluster Sum of Squares (WCSS) is a measure used to evaluate the performance of a clustering algorithm, like k-means. It calculates the sum of the squared distances between each data point in a cluster and the centroid of that cluster. The formula for WCSS can be expressed as follows:

$$[ \text{WCSS} = \sum_{i=1}^{k} \sum_{x \in C_i} ||x - \mu_i||^2 ]$$

Where:
- \( k \) is the number of clusters.
- $( C_i )$ represents the set of all points in cluster \( i \).
- \( x \) is an individual data point in cluster $( C_i )$.
- $( \mu_i )$ is the centroid of cluster \( i \).
- $( ||x - \mu_i||^2 )$ is the squared Euclidean distance between the data point \( x \) and the centroid $( \mu_i )$ of the cluster to which \( x \) belongs.

Let's break down this formula:

1. **Distance Calculation**: For each data point in a cluster, calculate the Euclidean distance between the point and the centroid of its cluster. This distance is then squared.

2. **Summation within a Cluster**: Sum these squared distances for all the points within a cluster. This gives you a measure of how 'tight' the cluster is - lower values mean the points are closer to the centroid, and thus, the cluster is more compact.

3. **Summation across Clusters**: Finally, sum these values across all clusters. This total sum is the WCSS.

The goal in many clustering algorithms, like k-means, is to minimize the WCSS, indicating that the clusters are as compact and distinct as possible. When using the Elbow Method to determine the optimal number of clusters, you're looking for the point where increases in the number of clusters don't lead to significant decreases in WCSS, suggesting that adding more clusters isn't improving the model meaningfully.

**What is kmeans.inertia?**

In the context of the k-means clustering algorithm, the attribute inertia_ from a k-means model in libraries such as scikit-learn refers to the Within-Cluster Sum of Squares (WCSS).

When you run the k-means algorithm on a dataset and fit the data, the inertia_ attribute gives you the sum of the squared distances of each data point to its nearest cluster center (centroid).

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
import warnings
warnings.filterwarnings('ignore')

# Generate some sample data for demonstration
np.random.seed(42)
data, _ = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Applying k-means with different number of clusters
inertia = []  # List to store the inertia for each k
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(data)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Method Graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.xticks(range(1, 11))
plt.show()
```

The plot above is an example of the Elbow Method using a generated dataset. Here's what it shows:

- The x-axis represents the number of clusters (like the number of jars in the candy example).
- The y-axis represents the inertia, which is a measure of how well the data is clustered. Lower inertia means the data points in each cluster are closer to each other, indicating better clustering.

As you can see from the plot, as we increase the number of clusters from 1 to 10, the inertia decreases. However, after a certain point, the rate of decrease in inertia slows down significantly. This point is like the 'elbow' of the plot.

In this example, the elbow seems to occur around 4 clusters. This suggests that 4 is the optimal number of clusters for this particular dataset. Beyond this point, increasing the number of clusters doesn't provide as significant an improvement in the clustering.

This method is useful in deciding the most appropriate number of clusters to use in k-means clustering, as it provides a good balance between the number of clusters and the compactness of the clusters.

```python
# importing the libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# loading the dataset
credit_df = pd.read_csv('Mall_Customers.csv')
credit_df.head()
```

```python
# check for missing values
credit_df.isnull().sum()
```

```python
# normalize the features
from sklearn.preprocessing import StandardScaler, normalize
X = credit_df.iloc[:, 3:5].values # Columns of interest

X = StandardScaler().fit_transform(X)
X = normalize(X)
print(X)
```

#### **Train the Model:**

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import SilhouetteVisualizer
from sklearn.metrics import silhouette_score

def silhouette_analysis(n):
  for n_clusters in range(2,n):
    km = KMeans(n_clusters=n_clusters)
    preds = km.fit_predict(X)
    centers = km.cluster_centers_

    score = silhouette_score(X, preds, metric='euclidean')
    print('For n_clusters = {}, silhouette score is {}'.format(n_clusters, score))

    visualizer = SilhouetteVisualizer(km)

    visualizer.fit(X)
    visualizer.poof()
```

```python
silhouette_analysis(10)
```

**Elbow Method:**

```python
# Applying k-means with different number of clusters
inertia = []  # List to store the inertia for each k
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=0).fit(X)
    inertia.append(kmeans.inertia_)

# Plotting the Elbow Method Graph
plt.figure(figsize=(8, 6))
plt.plot(range(1, 10), inertia, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of clusters")
plt.ylabel("Inertia")
plt.xticks(range(1, 10))
plt.show()
```

```python
# Apply k-means with the chosen number of clusters (assuming k=4)
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
cluster_labels = kmeans.fit_predict(X)

# Add cluster labels to the dataframe
credit_df['Cluster'] = cluster_labels

# Analyze the clusters
# This part involves statistical analysis of each cluster, for example:
print(credit_df.groupby('Cluster').mean())
```

```python
# Visualizing the clusters (two dimensions only)
plt.scatter(X[cluster_labels == 0, 0], X[cluster_labels == 0, 1], s = 100, c = 'red', label = 'High Income, Low Spenders')
plt.scatter(X[cluster_labels == 1, 0], X[cluster_labels == 1, 1], s = 100, c = 'blue', label = 'Low Income, High Spenders')
plt.scatter(X[cluster_labels == 2, 0], X[cluster_labels == 2, 1], s = 100, c = 'green', label = 'Low Income, Low Spenders')
plt.scatter(X[cluster_labels == 3, 0], X[cluster_labels == 3, 1], s = 100, c = 'cyan', label = 'High Income, High Spenders')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```

### <b> Limitations of expectationmaximization </b>

There are a few issues to be aware of when using the expectationmaximization algorithm:
1. The globally optimal solution may not be achieved.
2. The number of clusters must be selected beforehand.
3. k-means is limited to linear cluster boundaries.
4. k-means is slow for large number of samples.

### **Shortcomings:**

1. **Subjectivity in Choosing the Number of Clusters**:
   - Determining the optimal number of clusters can be subjective and may vary depending on the method used (like the Elbow Method or Silhouette Analysis). Different methods can suggest different numbers of clusters for the same dataset.

2. **Sensitivity to Initial Conditions and Noise**:
   - Some clustering algorithms, like K-means, are sensitive to the initial choice of centroids, which can lead to different results on different runs. They can also be affected by noise and outliers in the data.

3. **Assumption of Cluster Shapes**:
   - Many clustering algorithms assume certain shapes of clusters (like spherical for K-means). This can lead to poor performance if the actual clusters in the data have different shapes.

4. **Difficulty in Clustering High-Dimensional Data**:
   - As the number of dimensions (features) in the data increases, the difficulty of finding meaningful clusters also increases, a phenomenon known as the curse of dimensionality.

5. **Interpretation of Results**:
   - The interpretation of clusters can be challenging and subjective, as it often depends on domain knowledge. Clusters might not always be meaningful or interpretable in the context of the problem.

6. **Lack of Universal Evaluation Metric**:
   - There is no universal standard for evaluating the quality of the clusters formed. Different metrics can provide different perspectives on the quality, but none are conclusive.

7. **Dependence on Domain Knowledge**:
   - Effective clustering often requires domain knowledge to preprocess the data appropriately and interpret the results meaningfully.

8. **Computational Complexity**:
   - For large datasets, some clustering algorithms can be computationally intensive, leading to long processing times or the need for significant computational resources.

9. **Handling of Categorical Data**:
   - Many clustering algorithms are primarily designed for numerical data and may not work well, or require special preprocessing, for categorical data.

10. **No Guarantees of Finding the Optimal Solution**:
    - Some clustering algorithms, particularly those that use iterative improvement (like K-means), may not always converge to the global optimum solution.

### **Hierarchical Clustering**

 Hierarchical clustering is another popular unsupervised learning algorithm used for clustering. Unlike K-means clustering, which requires the number of clusters to be specified beforehand, hierarchical clustering does not require any prior knowledge of the number of clusters. Instead, it creates a hierarchy of clusters by successively merging or dividing existing clusters based on their similarity.

 There are two main types of hierarchical clustering: agglomerative and divisive.

**Agglomerative clustering**, also known as bottom-up clustering,  starts with each data point in its own cluster and successively merges the closest clusters until all data points belong to a single cluster. The distance between clusters can be measured using various metrics such as Euclidean distance or correlation distance.

**Divisive clustering**, also known as top-down clustering,  starts with all data points in a single cluster and successively divides the cluster into smaller sub-clusters until each data point is in its own cluster. This process is typically guided by a hierarchical tree structure that represents the division of the clusters.

Both agglomerative and divisive clustering can be applied to a wide range of data types and are useful for exploring the structure of complex data sets.  The choice between the two depends on the specific problem at hand and the desired level of granularity in the clustering results. 

Clustering1.gif

 The result of hierarchical clustering is typically represented as a dendrogram, which is a tree-like diagram that shows the hierarchy of clusters. The dendrogram can be cut at different heights to obtain different numbers of clusters, depending on the desired granularity of the clustering.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate synthetic data
np.random.seed(42)
X = np.random.rand(10, 2)

# Perform hierarchical clustering
Z = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(Z)
plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram')
plt.show()
```

In this example, we generate a small dataset X with 10 data points and 2 features. Then, we perform hierarchical clustering using the linkage function from scipy.cluster.hierarchy. The method='ward' argument specifies that we want to use the Ward linkage method, which minimizes the variance when merging clusters.

The linkage function computes the distances between data points and iteratively merges the closest pairs of clusters until all data points are in one cluster. The result of hierarchical clustering is a matrix Z containing the information about the merging process.

Finally, we plot the dendrogram using matplotlib to visualize the clustering process. The dendrogram illustrates the hierarchical structure of the clusters and the distances at which clusters are merged. By cutting the dendrogram at different heights, we can obtain different cluster assignments.

Advantages of Hierarchical Clustering:

 Biology: Hierarchical clustering can be used to analyze gene expression data to identify groups of genes that are co-expressed, which can provide insights into biological processes and disease mechanisms.

 Marketing: Hierarchical clustering can be used to segment customers based on their purchasing behavior or demographic information, which can be used to develop targeted marketing campaigns.

 Social Networks: Hierarchical clustering can be used to identify communities within social networks, which can provide insights into the structure of the network and the behavior of its users.

 Computer Vision: Hierarchical clustering can be used for image segmentation, where similar regions in an image are grouped together based on their pixel values or features.

Limitations of Hierarchical Clustering:

 Scalability: Hierarchical clustering can be computationally expensive for large datasets, especially for agglomerative clustering, which requires computing the distance between all pairs of data points.

 Sensitivity to Noise: Hierarchical clustering is sensitive to outliers and noise in the data, which can lead to suboptimal clustering results.

 Subjectivity: The choice of distance metric and linkage method can affect the resulting dendrogram and the number of clusters obtained, which can be subjective.

 Lack of Flexibility: Once the hierarchy of clusters is determined, it cannot be easily modified or updated without recomputing the entire clustering, which can be time-consuming.

###**Density-based Clustering**

Hey hey, have you heard of density-based clustering?  It's a fancy method that groups data points based on their density and proximity. Basically, if points are packed closely together, they're considered a cluster, while sparse areas are considered noise. 

 One of the most popular density-based clustering algorithms is DBSCAN (Density-Based Spatial Clustering of Applications with Noise).  The DBSCAN algorithm defines clusters as regions of the data space where there are many data points close together, separated by regions of lower point density.  DBSCAN works by defining two parameters: epsilon (), which represents the radius of the neighborhood around each point, and minPts, which represents the minimum number of points required to form a dense region or cluster.

 The DBSCAN algorithm starts with a random data point and finds all other data points within its epsilon radius. If the number of points within the radius is greater than or equal to the minPts parameter, a new cluster is formed.  If the number of points within the radius is less than minPts, the data point is labeled as noise or an outlier.

 Here's how it works:

* Choose a random data point.

* Find all other data points that are within a certain distance (epsilon) of the chosen point. These data points are considered "neighbors" of the chosen point.

* If the number of neighbors is greater than or equal to a specified threshold (minPts), then the chosen point is considered a "core point" and a new cluster is formed.

* Expand the cluster by finding all neighbors of the core point, and checking if they also have enough neighbors to be considered core points.

* Repeat steps 3-4 until no more core points can be found.

* Any remaining unclustered points are considered noise or outliers.

The algorithm is efficient because it avoids computing distances between all pairs of data points, and instead focuses on local density. DBSCAN is also robust to noise and outliers, and can handle non-linear and non-spherical clusters.

 **Applications of density-based clustering include:**

* Identifying clusters in spatial data such as  GPS locations or geographic regions. 

* Anomaly detection in network traffic or sensor data. 
Identifying groups of customers with similar purchasing behavior or preferences. 

* Segmentation of biological data such as gene expression data. 

**However, density-based clustering also has some limitations:**

* It can be sensitive to the choice of parameters, especially the epsilon radius and minPts values. 

* It may not work well on data with varying densities or with data points that are highly interconnected. 
* It can be computationally expensive for large datasets.
