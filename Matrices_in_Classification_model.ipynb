## **Confusion Matrix**

A confusion matrix is a table that is often used to describe the performance of a classification model (or "classifier") on a set of test data for which the true values are known. It allows visualization of the performance of an algorithm by showing the count of true positive, true negative, false positive, and false negative predictions.

### Terminology

- **True Positive (TP):** These are the cases where the model correctly predicts the positive class.
- **True Negative (TN):** These are the cases where the model correctly predicts the negative class.
- **False Positive (FP):** These are the cases where the model incorrectly predicts the positive class when the actual class is negative. (Type I error)
- **False Negative (FN):** These are the cases where the model incorrectly predicts the negative class when the actual class is positive. (Type II error)

### Structure
image.png

A confusion matrix is typically organized as follows:

|             | Predicted Positive | Predicted Negative |
|-------------|--------------------|--------------------|
| Actual Positive | True Positive (TP) | False Negative (FN) |
| Actual Negative | False Positive (FP) | True Negative (TN) |

### Interpretation

- **Accuracy:** The overall accuracy of the model can be calculated as (TP + TN) / (TP + TN + FP + FN). It represents the proportion of correctly classified instances among all instances.
- **Precision:** Precision is the ratio of correctly predicted positive observations to the total predicted positives, i.e., TP / (TP + FP). It measures the accuracy of positive predictions.
- **Recall (Sensitivity):** Recall is the ratio of correctly predicted positive observations to the all observations in actual class - yes. It is also known as sensitivity or true positive rate, i.e., TP / (TP + FN).
- **F1-Score:** The F1-Score is the weighted average of Precision and Recall. It is calculated as 2 * (Precision * Recall) / (Precision + Recall). It takes both false positives and false negatives into account.
- **Specificity:** Specificity is the ratio of correctly predicted negative observations to the all observations in actual class - no. It is also known as true negative rate, i.e., TN / (TN + FP).

Confusion matrices are helpful in evaluating the performance of a classification model, especially in scenarios where one class is more prevalent than the other, leading to imbalanced datasets.

### **Precision and Recall**

Precision and recall are two important metrics used to evaluate the performance of classification models, especially in tasks where class imbalance is present. They are commonly used in information retrieval, machine learning, and natural language processing tasks.

**Precision** and **recall** are calculated based on the outcomes of a binary classification task, where one class is typically considered as the positive class (e.g., relevant documents in information retrieval, or the presence of a disease in medical diagnosis) and the other class as the negative class.

#### 3.**Precision:**
Precision measures the accuracy of the positive predictions made by the model. It answers the question: "Out of all the items predicted as positive, how many are actually positive?"

Precision is calculated using the following formula:

$$ \text{Precision} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Positives}}$$

- True Positives (TP) are the cases where the model correctly predicted the positive class.
- False Positives (FP) are the cases where the model incorrectly predicted the positive class.

A high precision indicates that the model is making few false positive predictions, i.e., the majority of the items predicted as positive are indeed positive.

**Use Case:** You are concerned about the reliability of positive predictions made by the classifier.


#### 4. **Recall (Sensitivity):**

Recall, also known as sensitivity or true positive rate, measures the ability of the model to correctly identify all relevant instances in the dataset. It answers the question: "Out of all the actual positive items, how many did the model correctly predict?"

Recall is calculated using the following formula:

$$\text{Recall} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}} $$

- True Positives (TP) are the cases where the model correctly predicted the positive class.
- False Negatives (FN) are the cases where the model failed to predict the positive class when it should have.

A high recall indicates that the model is capturing a large proportion of the actual positive instances, i.e., minimizing false negatives.

**Use Case:** You want to understand how effectively the classifier identifies positive instances

#### Interpretation:
- High precision with low recall: The model makes very few false positive predictions, but misses many positive instances. This could be acceptable in situations where false positives are costly, such as medical diagnosis.
- High recall with low precision: The model captures most of the positive instances but also makes many false positive predictions. This could be acceptable in situations where missing positive instances is more critical than incorrectly labeling negative instances.
- Balanced precision and recall: Ideally, a good model would have both high precision and high recall, indicating that it accurately identifies relevant instances while minimizing false positives and false negatives.

In summary, precision and recall are complementary metrics that provide insights into different aspects of a classification model's performance and are often used together to evaluate its effectiveness.

### 5.**F1 Score:**

The F1 Score is a measure of a model's accuracy on a dataset, considering both the precision and recall of the model. It is especially useful when dealing with imbalanced classes or when false positives and false negatives have different costs.

The F1 Score is the harmonic mean of precision and recall, calculated as follows:

$$
F1 \ Score = 2 \times \frac{Precision \times Recall}{Precision + Recall}
$$

where Precision is the ratio of true positive predictions to the total number of positive predictions, and Recall is the ratio of true positive predictions to the total number of actual positive instances in the dataset.

Precision measures the accuracy of positive predictions, while recall measures the coverage of positive instances.

A higher F1 Score indicates better performance, as it balances precision and recall. It provides a single value that summarizes both precision and recall, making it a useful metric for evaluating model performance, especially in situations where classes are imbalanced.

In summary, while accuracy measures the overall correctness of predictions, the F1 Score provides a more nuanced evaluation by considering both precision and recall, making it particularly valuable in scenarios where class distributions are imbalanced or where false positives and false negatives have different consequences.


### **Precision, Recall, and F1-Score for Multi-Classification**

Precision, recall, and F1-score are commonly used metrics to evaluate the performance of classification models, especially in multi-class classification scenarios. These metrics help assess the quality of the model's predictions by considering different aspects of its performance.

#### **Precision**

Precision measures the accuracy of positive predictions made by the model. It calculates the ratio of true positive predictions to all positive predictions made by the model.

**Precision** = (True Positives) / (True Positives + False Positives)

#### **Recall**

Recall, also known as sensitivity, measures the ability of the model to correctly identify positive instances from all actual positive instances. It calculates the ratio of true positive predictions to all actual positive instances.

**Recall** = (True Positives) / (True Positives + False Negatives)

#### **F1-Score**

F1-score is the harmonic mean of precision and recall. It provides a balance between precision and recall, making it a useful metric when you need to consider both false positives and false negatives. F1-score reaches its best value at 1 and worst at 0.

**F1-Score** = 2 * (Precision * Recall) / (Precision + Recall)

#### Steps to Calculate Precision, Recall, and F1-Score for Multi-Classification:

1. **Count True Positives (TP), False Positives (FP), and False Negatives (FN) for each class**: Calculate the number of true positive, false positive, and false negative predictions for each class.

2. **Calculate Precision, Recall, and F1-Score for each class**: Use the formulas mentioned above to calculate precision, recall, and F1-score for each class individually.

3. **Weighted Average**: Optionally, you can calculate a weighted average of precision, recall, and F1-score across all classes to get an overall performance measure. This is particularly useful when classes are imbalanced.

#### Example:

Let's consider a multi-class classification problem with three classes: A, B, and C.

- For class A:
  - True Positives (TP_A) = 50
  - False Positives (FP_A) = 10
  - False Negatives (FN_A) = 5

- For class B:
  - True Positives (TP_B) = 40
  - False Positives (FP_B) = 8
  - False Negatives (FN_B) = 12

- For class C:
  - True Positives (TP_C) = 30
  - False Positives (FP_C) = 5
  - False Negatives (FN_C) = 8

**Precision**:

Precision for class A = 50 / (50 + 10) = 0.833

Precision for class B = 40 / (40 + 8) = 0.833

Precision for class C = 30 / (30 + 5) = 0.857

**Recall**:

Recall for class A = 50 / (50 + 5) = 0.909

Recall for class B = 40 / (40 + 12) = 0.769

Recall for class C = 30 / (30 + 8) = 0.789

**F1-Score**:

F1-Score for class A = 2 * (0.833 * 0.909) / (0.833 + 0.909) = 0.870

F1-Score for class B = 2 * (0.833 * 0.769) / (0.833 + 0.769) = 0.800

F1-Score for class C = 2 * (0.857 * 0.789) / (0.857 + 0.789) = 0.822

#### Notable Features/Functionality:

- **Balanced Evaluation**: Precision, recall, and F1-score provide a balanced evaluation of a classifier's performance, especially in situations where classes are imbalanced.
- **Interpretability**: These metrics are easy to interpret, making them useful for conveying the performance of a classifier to stakeholders.
- **Threshold Independence**: These metrics are threshold independent, meaning they do not depend on the decision threshold set for classification, unlike metrics like accuracy.

By calculating precision, recall, and F1-score for each class, and potentially computing a weighted average across all classes, you can effectively evaluate the performance of a multi-class classification model.

### **True Positive Rate (TPR):**
TPR, also known as sensitivity or recall, measures the proportion of actual positive instances that are correctly identified by the model.

Mathematically, TPR is calculated as the ratio of true positives to the sum of true positives and false negatives:
$$
\text{TPR} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}}
$$
TPR represents the model's ability to correctly classify positive instances among all actual positive instances.

### **False Positive Rate (FPR):**
FPR measures the proportion of actual negative instances that are incorrectly classified as positive by the model.

Mathematically, FPR is calculated as the ratio of false positives to the sum of false positives and true negatives:
$$
\text{FPR} = \frac{\text{False Positives}}{\text{False Positives} + \text{True Negatives}}
$$
FPR represents the model's tendency to generate false alarms or false positives.

### **Comparison with Recall and Precision:**
- **Recall (Sensitivity/TPR)**: Recall is another name for TPR. It measures the ability of the classifier to find all the positive samples. A high recall indicates that the classifier is good at capturing positive instances.
- **Precision**: Precision measures the accuracy of positive predictions made by the model. It is the ratio of true positives to the sum of true positives and false positives. A high precision indicates that the classifier is good at avoiding false positives.

In summary, while TPR and FPR focus on the performance of the model with respect to positive and negative instances, recall and precision provide insights into the accuracy of positive predictions and the ability to avoid false positives, respectively.


##**ROC Curve**
------------
image.png

**ROC Curve (Receiver Operating Characteristic Curve):**

- **Definition:** The ROC curve is a graphical representation of the performance of a binary classification model across different threshold settings. It plots the True Positive Rate (TPR) against the False Positive Rate (FPR) at various threshold values.

- **Use:** The ROC curve is widely used in evaluating the performance of binary classification models, especially in scenarios where the class distribution is imbalanced. It helps in comparing the trade-offs between the true positive rate and the false positive rate, allowing users to select the most suitable threshold for their specific application.

- **Advantages:**
  - Provides a comprehensive view of a classifier's performance across all possible threshold values.
  - Helps in understanding the trade-offs between sensitivity (TPR) and specificity (1 - FPR).
  - Effective in comparing the performance of different models.
  - Robust to class imbalance.

- **Disadvantages:**
  - It may not be suitable for multi-class classification problems.
  - The interpretation of the ROC curve may not be straightforward, especially when the curve is close to the diagonal line (random classifier).

- **Mathematical Formula:**
  - **True Positive Rate (Sensitivity/Recall/TPR):**
$$
\text{TPR} = \frac{\text{True Positives}}{\text{True Positives} + \text{False Negatives}} $$
    
  - **False Positive Rate (FPR):**
    $$\text{FPR} = \frac{\text{False Positives}}{\text{False Positives} + \text{True Negatives}}
    $$

- **Construction:**
  1. Generate predicted probabilities for each instance in the dataset.
  2. Sort instances by their predicted probabilities.
  3. Start with a threshold where all instances are classified as positive (high threshold).
  4. Decrease the threshold gradually, calculating TPR and FPR at each step.
  5. Plot TPR against FPR for each threshold setting, resulting in the ROC curve.

The ROC curve is typically visualized using a line plot, where each point on the curve represents the performance of the classifier at a particular threshold setting. A perfect classifier would have an ROC curve that passes through the top-left corner of the plot (TPR = 1, FPR = 0), while a random classifier would produce a diagonal line from bottom-left to top-right.

**AUC-ROC (Area Under the Receiver Operating Characteristic Curve):**

- **Definition:** AUC-ROC is a metric that quantifies the performance of a binary classification model by measuring the area under the Receiver Operating Characteristic (ROC) curve. It represents the probability that the classifier will rank a randomly chosen positive instance higher than a randomly chosen negative instance.

- **Use:** AUC-ROC is widely used in evaluating the performance of binary classification models, providing a single scalar value that summarizes the classifier's ability to discriminate between positive and negative instances across all possible threshold settings.

- **Advantages:**
  - Provides a single value metric for comparing the performance of different classifiers.
  - Robust to class imbalance.
  - Interpretable and easy to understand.
  - Effective in scenarios where the choice of threshold is not critical.

- **Disadvantages:**
  - May not be suitable for multi-class classification problems.
  - Less informative about the specific threshold selection compared to other metrics like precision-recall curve.

- **Mathematical Formula:**
  - The AUC-ROC represents the area under the ROC curve, which is calculated using integration or approximation methods.
  
- **Construction:**
  1. Generate predicted probabilities for each instance in the dataset.
  2. Sort instances by their predicted probabilities.
  3. Start with a threshold where all instances are classified as positive (high threshold).
  4. Decrease the threshold gradually, calculating TPR and FPR at each step.
  5. Plot the ROC curve and calculate the area under it using integration or approximation methods.

The AUC-ROC value ranges from 0 to 1, where:
- AUC = 1 indicates a perfect classifier.
- AUC = 0.5 indicates a random classifier (no discriminative power).
- AUC < 0.5 indicates a classifier performing worse than random.

The AUC-ROC metric provides a comprehensive assessment of a classifier's performance across all possible threshold settings, making it a popular choice for model evaluation in binary classification tasks.


##**Definition**
**Sensitivity / True Positive Rate / Recall:**

- **Definition:** Sensitivity, also known as True Positive Rate (TPR) or Recall, measures the proportion of positive instances that are correctly identified by the classifier. It indicates how well the classifier performs in detecting positive instances from the total actual positives.

- **Formula :**
  $$Sensitivity = \frac{True Positives}{True Positives + False Negatives}$$

**False Negative Rate:**

- **Definition:** False Negative Rate (FNR) measures the proportion of positive instances that are incorrectly classified as negative by the classifier. It represents the rate of missed positive predictions.

- **Formula :**
  $$FNR = \frac{False Negatives}{False Negatives + True Positives}$$

**Specificity / True Negative Rate:**

- **Definition:** Specificity, also known as True Negative Rate (TNR), measures the proportion of negative instances that are correctly identified by the classifier. It indicates how well the classifier performs in detecting negative instances from the total actual negatives.

- **Formula :**
  $$Specificity = \frac{True Negatives}{True Negatives + False Positives}$$

**False Positive Rate:**

- **Definition:** False Positive Rate (FPR) measures the proportion of negative instances that are incorrectly classified as positive by the classifier. It represents the rate of false alarms or false positive predictions.

- **Formula :**

  $$FPR = \frac{False Positives}{False Positives + True Negatives}= 1-Specificity$$
- **Interpretation:**
  - A higher TPR (Sensitivity) and a lower FNR are desirable as they indicate a higher rate of correctly identifying positive instances.
  - A higher TNR (Specificity) and a lower FPR are desirable as they indicate a higher rate of correctly identifying negative instances.

These metrics are crucial in evaluating the performance of binary classification models and understanding their ability to correctly classify positive and negative instances.


##**Kappa Statistic (Cohen's Kappa)**
**Cohen's Kappa**, also known as Kappa statistic, is a measure of inter-rater agreement or reliability for categorical items, particularly in classification tasks. It assesses the level of agreement between two raters (or two sets of classifications) while taking into account the possibility of agreement occurring by chance. Cohen's Kappa is widely used in various fields, including machine learning, psychology, and social sciences.

**Advantages:**
1. **Accounts for Chance Agreement**: Unlike simple agreement measures, Cohen's Kappa adjusts for the possibility of agreement occurring by chance alone.
2. **Applicable to Categorical Data**: It can be used to assess agreement between raters or classifiers when dealing with categorical variables.
3. **Robust to Imbalanced Classes**: It is less affected by class imbalance compared to simple accuracy.

**Disadvantages:**
1. **Sensitive to Marginal Distributions**: Cohen's Kappa can be affected by large differences in marginal distributions of categories.
2. **Depends on Number of Categories**: The interpretation of Kappa can vary depending on the number of categories involved.
3. **Limited to Pairwise Agreement**: It measures agreement between only two raters or classifiers at a time, limiting its applicability in scenarios with more than two raters.

**Use Cases:**
1. **Annotation Agreement**: Assessing agreement between human annotators in labeling tasks, such as sentiment analysis or image classification.
2. **Evaluation of Classification Models**: Measuring agreement between different models or algorithms in assigning categorical labels to instances.
3. **Assessment of Diagnostic Tests**: Evaluating the agreement between different medical tests or diagnostic tools in categorizing patients into disease categories.

**Mathematical Formula:**
Cohen's Kappa is calculated using the observed agreement and expected agreement among raters or classifiers. The formula is given by:

$$
\kappa = \frac{{P_o - P_e}}{{1 - P_e}}
$$

Where:
- $P_o$ is the observed proportion of agreement.
- $P_e$ is the expected proportion of agreement by chance alone.

The expected agreement $P_e$ is calculated as the product of the marginal probabilities of agreement for each category, summed over all categories.

Cohen's Kappa ranges from -1 to 1, where:
- $\kappa = 1$ indicates perfect agreement beyond chance.
- $\kappa = 0$ indicates agreement equal to that expected by chance alone.
- $\kappa < 0$ indicates agreement worse than chance.

Overall, Cohen's Kappa provides a valuable measure of agreement that accounts for chance agreement and is widely used in various fields for evaluating the reliability and consistency of categorical classifications.
