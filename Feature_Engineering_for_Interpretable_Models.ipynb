## **Handling Imbalanced Dataset**

The class imbalance problem in machine learning refers to situations where the classes (categories) in the dataset are not represented equally. This imbalance can significantly skew the performance of machine learning models, especially those that assume an equal or similar distribution of classes.

Here are some key points about the class imbalance problem:

1. **Definition**: Class imbalance occurs when there are significantly more instances of some classes than others in the dataset. For example, in a binary classification problem, you might have a lot of examples for one class (the majority class) and very few for the other (the minority class).

2. **Impact on Model Performance**: Most machine learning algorithms are designed to maximize overall accuracy, which can be misleading when the classes are imbalanced. They tend to be biased towards the majority class, as correctly predicting the majority class most of the time can still achieve high accuracy.

3. **Real-world Examples**: Class imbalance is common in real-world scenarios. For example, in fraud detection, the number of fraudulent transactions is typically much smaller compared to legitimate transactions. In medical diagnosis, the number of patients with a rare disease is usually much lower than those without it.

4. **Strategies to Address the Issue**:
   - **Resampling Techniques**: Balancing the dataset by either oversampling the minority class or undersampling the majority class.
   - **Algorithmic Level Solutions**: Using algorithms that are inherently more robust to class imbalance, such as tree-based methods.
   - **Cost-Sensitive Learning**: Assigning a higher cost to misclassifying the minority class.
   - **Evaluation Metrics**: Using metrics that are more informative than accuracy in imbalanced settings, such as precision, recall, F1-score, or the area under the ROC curve.

5. **Consideration of Data Quality**: While addressing class imbalance, it's also important to maintain the quality and representativeness of the data. Simply adding more samples of the minority class (through techniques like synthetic data generation) or removing samples from the majority class can sometimes lead to overfitting or loss of important information.

Understanding and effectively handling class imbalance is crucial for developing models that perform well in real-world conditions, where balanced datasets are often not the norm.

### Resampling Techniques

A widely adopted technique for dealing with highly unbalanced datasets is called resampling. It consists of removing samples from the majority class (under-sampling) and/or adding more examples from the minority class (over-sampling).

Screenshot 2020-12-26 at 3.30.55 AM.png

Despite the advantage of balancing classes, these techniques also have their own weaknesses.

The simplest implementation of over-sampling is to duplicate random records from the minority class, which can cause overfitting.

In under-sampling, the simplest technique involves removing random records from the majority class, which can cause loss of information.



Let‚Äôs implement this with the fraud detection example.

We will start by separating the class that will be 0 and class 1.

````python
# Class count
class_count_0, class_count_1 = data['Fraud'].value_counts()

# Separate class
class_0 = data[data['Fraud'] == 0]
class_1 = data[data['Fraud'] == 1]
print('class 0:', class_0.shape)
print('class 1:', class_1.shape)
````

#### 1. **Random Under-Sampling**

Undersampling can be defined as removing some observations of the majority class. This is done until the majority and minority class is balanced out.

Undersampling can be a good choice when you have a ton of data -think millions of rows. But a drawback to undersampling is that we are removing information that may be valuable.

````python
from imblearn.under_sampling import RandomUnderSampler

rus = RandomUnderSampler(random_state=42, replacement=True)
x_rus, y_rus = rus.fit_resample(data.iloc[:,0:-1], data['Fraud'])

print('Original dataset shape:', len(data))
print('Resampled dataset shape', len(y_rus))
````

#### 2. **Random Over-Sampling**

Oversampling can be defined as adding more copies to the minority class. Oversampling can be a good choice when you don‚Äôt have a ton of data to work with.

A con to consider when oversampling is that it can cause overfitting and poor generalization to your test set.

````python
from imblearn.over_sampling import RandomOverSampler

ros = RandomOverSampler(random_state=42)
x_ros, y_ros = ros.fit_resample(data.iloc[:,0:-1], data['Fraud'])

print('Original dataset shape:', len(data))
print('Resampled dataset shape', len(y_ros))
````

#### 3. **Under-Sampling: Tomek links**

Tomek links are pairs of very close instances but of opposite classes. Removing the instances of the majority class of each pair increases the space between the two classes, facilitating the classification process.

Tomek‚Äôs link exists if the two samples are the nearest neighbors of each other.

````python
from imblearn.under_sampling import TomekLinks

tl = TomekLinks(sampling_strategy='majority')

X_tl, y_tl = tl.fit_resample(data.iloc[:,0:-1], data['Fraud'])

print('Original dataset shape:', len(data))
print('Resampled dataset shape', len(y_tl))
````

#### 4. **Synthetic Minority Oversampling Technique (SMOTE)**

This technique generates synthetic data for the minority class.

SMOTE (Synthetic Minority Oversampling Technique) works by randomly picking a point from the minority class and computing the k-nearest neighbors for this point. The synthetic points are added between the chosen point and its neighbors.

Screenshot 2020-12-26 at 3.53.10 AM.png

SMOTE algorithm works in 4 simple steps:

* Choose a minority class as the input vector
* Find its k nearest neighbors (k_neighbors is specified as an argument in the SMOTE() function)
* Choose one of these neighbors and place a synthetic point anywhere on the line joining the point under consideration and its chosen neighbor
* Repeat the steps until data is balanced

````python
from imblearn.over_sampling import SMOTE

smote = SMOTE()

# fit predictor and target variable
x_smote, y_smote = smote.fit_resample(data.iloc[:,0:-1], data['Fraud'])

print('Original dataset shape', len(data))
print('Resampled dataset shape', len(y_smote))
````

### **Choose the evaluation metric wisely**

Accuracy is not the best metric to use when evaluating imbalanced datasets as it can be misleading.

Metrics that can provide better insight are:

* **Confusion Matrix**: a table showing correct predictions and types of incorrect predictions.
* **Precision**: the number of true positives divided by all positive predictions. Precision is also called Positive Predictive Value. It is a measure of a classifier‚Äôs exactness. Low precision indicates a high number of false positives.
* **Recall**: the number of true positives divided by the number of positive values in the test data. The recall is also called Sensitivity or the True Positive Rate. It is a measure of a classifier‚Äôs completeness. Low recall indicates a high number of false negatives.
* **F1 Score**: the weighted average of precision and recall.
* **Area Under ROC Curve (AUC-ROC)**: AUC-ROC represents the likelihood of your model distinguishing observations from two classes. In other words, if you randomly select one observation from each class, what‚Äôs the probability that your model will be able to ‚Äúrank‚Äù them correctly?

##**Feature selection: Techniques and their applicability**

Feature selection is a crucial step in the data analysis and machine learning process. Think of it as deciding what ingredients to put in a recipe. Just like in cooking, where using too many or the wrong ingredients can spoil a dish, in data analysis, using too many or irrelevant features (which are like individual ingredients for your data recipe) can make your analysis or model less effective. Here's why feature selection is important:

1. **Improves Model Performance:** Using only relevant features can help your model perform better. It's like focusing on the ingredients that really matter for your dish to taste great.

2. **Reduces Complexity:** Fewer features mean simpler models. Simpler models are generally easier to understand and faster to run, just like a simpler recipe is easier to follow and quicker to cook.

3. **Avoids Overfitting:** Overfitting is when a model is too closely tailored to the training data and doesn‚Äôt perform well on new, unseen data. It's like a recipe that‚Äôs perfect for one specific occasion but doesn‚Äôt work well for others. Feature selection helps avoid this by removing irrelevant or redundant data.

4. **Easier Interpretation:** With fewer features, it‚Äôs easier to understand the results and explain them to others. A simple recipe with fewer ingredients is easier to explain than one with dozens of components.

5. **Faster Training Time:** Fewer features mean less data for the model to process, which can speed up the training process. It‚Äôs like cooking a meal faster because you have fewer ingredients to prepare.

6. **Reduces Risk of Errors:** Each additional feature introduces a chance of errors, either in the data itself or in how the model processes it. Fewer features can mean fewer opportunities for errors.

7. **Improves Data Quality:** Feature selection can help you identify and remove data that doesn't contribute to your analysis or might even be misleading.

In summary, feature selection helps you choose the right data for your analysis, much like choosing the right ingredients for a recipe. This can make your models simpler, faster, more accurate, and easier to understand.

##### <u>**Example:**</u>

**Step 1: Create Sample Dataset with Random Features**
We'll create a dataset with a few random features and a target variable.

**Step 2: Add a Constant Feature to the Dataset**
Now, we'll add a constant feature to our dataset. This feature will have the same value for all observations. Let's add this feature and then proceed with fitting a linear regression model.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Creating a sample dataset with random features
np.random.seed(0)  # Seed for reproducibility
sample_data = {
    'feature1': np.random.rand(100),
    'feature2': np.random.rand(100),
    'feature3': np.random.rand(100),
    'target': np.random.randint(0, 2, 100)  # Binary target
}

sample_df = pd.DataFrame(sample_data)
# Adding a constant feature to the dataset
sample_df['constant_feature'] = 1

# Display the first few rows of the dataframe
sample_df.head()
````

Let's visualize how the constant feature is related to the target variable.

````python
# Scatter plot for Constant Feature vs Target
plt.figure(figsize=(6, 4))
sns.lineplot(data=sample_df, x='constant_feature', y='target')
plt.title("Constant Feature vs Target")
plt.show()
````

**Step 3: Fit a Linear Regression Model and Evaluate Performance**

Next, we'll fit a linear regression model to this data, including the constant feature. We'll then evaluate the model's performance using the mean squared error (MSE). This will help us understand the impact of the constant feature on the model.

````python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score

# Preparing the data for modeling
X_sample = sample_df.drop('target', axis=1)
y_sample = sample_df['target']


# Fit the linear regression model
model_sample = LinearRegression()
model_sample.fit(X_sample, y_sample)

# Make predictions and calculate MSE
predictions = model_sample.predict(X_sample)

mse = mean_squared_error(y_sample, predictions)

mse
````

The Mean Squared Error (MSE) for our model is approximately 0.24326. This gives us an initial indication of the model's performance, including the constant feature.

Variance Inflation Factor (VIF) is a measure used primarily to detect multicollinearity among predictors within a multiple regression model. However, it can also be indirectly used as a feature selection method, particularly in the context of linear models. Here's how it works and how it can be applied in feature selection:

1. **Purpose**: VIF assesses how much the variance of an estimated regression coefficient increases if your predictors are correlated. A high VIF indicates that the predictor is highly collinear with the other predictors, and therefore, its coefficient is not well estimated.

2. **Calculation**: For each predictor variable, a VIF value is calculated. It's the ratio of variance in a model with multiple terms, divided by the variance of a model with one term alone. Mathematically, $( \text{VIF}_i = \frac{1}{1 - R_i^2})$, where $( R_i^2)$ is the coefficient of determination of a regression of predictor $( i)$ on all the other predictors.

**Step 4: Calculate VIF**

Now, we'll calculate the Variance Inflation Factor (VIF) for each feature. This will help us assess the impact of multicollinearity, especially the influence of the constant feature. High VIF values can indicate problematic multicollinearity. Let's see how the constant feature affects the VIF values.

````python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calculate_vif(dataframe):
    vif_data = pd.DataFrame()
    vif_data["feature"] = dataframe.columns

    # Calculating VIF for each feature
    vif_data["VIF"] = [variance_inflation_factor(dataframe.values, i) for i in range(len(dataframe.columns))]
    return vif_data

vif_sample_df = calculate_vif(X_sample)

vif_sample_df
````

**Interpretation**:
   - A VIF of 1 means that the predictor is not correlated with other variables.
   - A VIF between 1 and 5 typically indicates moderate correlation, but it's often not severe enough to require attention.
   - A VIF above 5 or 10 (thresholds can vary based on the field of study) suggests high correlation that may be problematic.

### **What is curse of dimensionality?**

The "curse of dimensionality" is a term used in data science and machine learning that describes the problems that arise when analyzing and organizing data in high-dimensional spaces (that is, spaces with many features or variables). It's a bit like trying to find your way in a very, very large and complex maze with many, many paths.

Here's a simple explanation:

1. **Harder to Visualize:** Just like it's easy to picture a line (1D) or a square (2D), but hard to imagine a 4D or 5D shape, high-dimensional data is very hard for us to visualize or understand.

2. **More Data Needed:** In high dimensions, the amount of data you need to make good predictions grows exponentially. It's like if you're in a big, multi-level maze, you need many more signs to guide you through it than in a simple one-level maze.

3. **Distance Measures Become Less Useful:** In high-dimensional spaces, traditional measures of distance (like Euclidean distance) start to become less meaningful. This can make some machine learning algorithms less effective, as they rely on these distance measures.

4. **Overfitting Risk:** With lots of features, there's a higher chance that your model will start to "memorize" the training data (overfit) rather than learning to generalize from it. It's like focusing too much on the tiny details of the maze and forgetting the overall path to the exit.

5. **Increased Computational Complexity:** More dimensions mean more calculations for the computer, which can slow down data analysis significantly. Imagine how much longer it would take to solve a maze with hundreds of floors compared to a simple flat one.

In summary, the curse of dimensionality refers to the various issues that arise when dealing with very high-dimensional data. It can make data analysis more complex, less accurate, and more computationally intensive. That's why techniques like feature selection, dimensionality reduction, or choosing appropriate models become critical in high-dimensional settings.

### **Why dimensionality reduction is important?**

Dimensionality reduction techniques are important in data science and machine learning because they help simplify complex datasets. Imagine you have a backpack full of tools, but you're only going on a short hike. Carrying all those tools makes your backpack heavy and your hike more difficult. If you take only what you need, your hike becomes easier and more enjoyable. Dimensionality reduction is like choosing the essential tools for your hike.

Here's why these techniques are necessary:

1. **Reduces Complexity:** High-dimensional data (data with many features) can be very complex. Reducing dimensions makes the data simpler to work with, just like a lighter backpack is easier to carry.


2. **Improves Model Performance:** Too many features can lead to overfitting, where the model learns the noise in the data rather than the actual trend.
Reducing dimensions can help the model focus on the most important aspects, just like taking only necessary tools can make your work more efficient.

3. **Faster Computation:** Fewer dimensions mean fewer calculations, which speeds up data processing and analysis. It's like how a smaller backpack is quicker and easier to pack and carry.

4. **Easier Visualization and Understanding:** It's hard to visualize and understand data in many dimensions. Reducing dimensions can help you see patterns and relationships more clearly, similar to how a simpler map is easier to read than a complex one.

5. **Reduces Storage and Memory Needs:** Less dimensional data takes up less space, which can be important for storing and processing large datasets.

6. **Removes Redundant Features:** Some features may be redundant or irrelevant. Dimensionality reduction helps identify and remove these, making the dataset more effective.

In short, dimensionality reduction helps make large, complex datasets more manageable, understandable, and useful, much like simplifying your gear makes a hike more enjoyable and less burdensome.


Dimensionality reduction techniques, such as Principal Component Analysis (PCA) or Singular Value Decomposition (SVD), transform the features into a lower-dimensional space while retaining the most important information. It's like condensing our data into a more manageable form without losing too much!

**Step 2: Calculate the Covariance Matrix**

For a 2x2 matrix, the covariance matrix is calculated as follows:

1. Calculate the mean of each feature.

2. Calculate the covariance for each pair of features. The formula for covariance between two variables X and Y is:
\begin{align*}
  Cov(X,Y) &= \frac{1}{n-1}\sum_{i=1}^{n} (X_{i} - \bar X)(Y_{i} - \bar Y) \\
\end{align*}

````python
# Calculating the covariance matrix for the standardized 2x2 matrix
covariance_matrix_2x2 = np.cov(data_2x2_standardized)

covariance_matrix_2x2
````

Now, we'll apply PCA to the standardized data. PCA will transform the original data into a new set of variables, the principal components, which are orthogonal (uncorrelated) and maximize the variance.

We'll start by computing PCA and then analyze the amount of variance captured by each principal component. This will give us insight into how many components we might want to keep. Let's perform PCA using scikit-learn's PCA implementation.

````python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Applying PCA
pca = PCA()
student_pca = pca.fit_transform(student_df_standardized)

# Convert PCA components to a DataFrame for better readability
student_pca_df = pd.DataFrame(data=student_pca, columns=[f'PC{i+1}' for i in range(len(student_df.columns))])

# Display the first few rows of the PCA components
student_pca_df.head()
````

## **Best Practices:**

Feature engineering is a crucial step in the data preprocessing phase of building a machine learning model. It involves creating new features or modifying existing ones to improve the performance of the model. Here are some best practices for feature engineering:

1. **Understand Your Data and Domain**:
   - Know the context and specifics of your dataset.
   - Work with domain experts to gain insights that could guide your feature engineering efforts.

2. **Start with a Clear Objective**:
   - Identify what you're trying to predict or classify and tailor your features to support this goal.
   - Ensure that features are relevant to the problem you are solving.

3. **Keep It Simple**:
   - Start with simple features before trying complex ones.
   - Avoid over-engineering, which can lead to overly complex models and overfitting.

4. **Handle Missing Data Carefully**:
   - Use strategies like imputation, removal, or indicator variables for missing data.
   - Understand the implications of each method on your model and data.

5. **Encode Categorical Variables**:
   - Convert categorical data into a format that can be easily used by machine learning algorithms (like one-hot encoding or label encoding).
   - Consider the impact of encoding on dimensionality and model complexity.

6. **Normalize or Scale Features**:
   - Apply normalization or scaling when features have different scales or units.
   - This is especially important for models sensitive to the scale of data like KNN, SVM, and neural networks.

7. **Feature Selection**:
   - Use techniques like correlation analysis, backward elimination, or machine learning-based methods to select relevant features.
   - Avoid including irrelevant or redundant features that can reduce model performance.

8. **Feature Extraction and Dimensionality Reduction**:
   - Use techniques like PCA (Principal Component Analysis) to reduce the number of features while retaining important information.
   - Useful in high-dimensional datasets to combat the curse of dimensionality.

9. **Temporal and Spatial Feature Engineering**:
   - For time series or spatial data, create features that capture temporal dynamics (like seasonality, trends) or spatial patterns.
   - Be mindful of leakage, ensuring that future data is not inadvertently used.

10. **Iterative Process**:
    - Treat feature engineering as an iterative process.
    - Regularly revisit and refine your features based on model performance and new insights.

11. **Evaluate Feature Importance**:
    - Use model feedback to understand which features are most important.
    - This can guide further feature engineering efforts.

12. **Be Aware of Data Leakage**:
    - Ensure that your features do not inadvertently include information from the future or from the target variable.

Remember, the most effective feature engineering strategies often depend on the specific characteristics of the dataset and the problem at hand. Regular experimentation and validation are key to identifying the most impactful features.

## **Shortcomings:**

Feature engineering, while essential in machine learning, comes with its own set of challenges and limitations. Understanding these shortcomings is important for effective model development. Here are some key ones:

1. **Time-Consuming Process**:
   - Feature engineering often requires a significant amount of time and effort. It involves understanding the data deeply, creating hypotheses, and iteratively testing them.

2. **Requires Domain Expertise**:
   - Effective feature engineering often relies on domain knowledge to create meaningful features. Lack of such knowledge can lead to the omission of important features or the creation of irrelevant ones.

3. **Risk of Overfitting**:
   - Over-engineering features can lead to models that are too complex and fit too closely to the training data, performing poorly on unseen data (overfitting).

4. **Model Complexity**:
   - Adding too many features, or overly complex features, can increase the complexity of the model. This might make the model harder to understand and interpret.

5. **Data Leakage**:
   - There's a risk of accidentally introducing information about the target variable into the features, a problem known as data leakage, which can lead to overly optimistic performance estimates.

6. **Scalability Issues**:
   - Some feature engineering techniques may not scale well with large datasets, either due to computational constraints or because they rely on fine-tuned manual processes.

7. **Dynamic Data Changes**:
   - Features engineered based on historical data may become irrelevant or misleading if the underlying data distribution changes over time (concept drift).

8. **Bias Amplification**:
   - If the original data contains biases, feature engineering can inadvertently amplify these biases, leading to unfair or biased model predictions.

9. **Dependency on Quality of Data**:
   - The effectiveness of feature engineering is heavily dependent on the quality of the underlying data. Poor quality data (noisy, erroneous, or incomplete) can severely limit the potential gains from feature engineering.

10. **Difficulty in Automation**:
    - While some aspects of feature engineering can be automated, many require creative and iterative human input, making full automation challenging.

Despite these challenges, feature engineering remains a critical step in building effective machine learning models. The key is to be aware of these pitfalls and approach feature engineering with a careful, informed strategy.

#### **2.	Imputing Random Values**

Random Value imputation consists of taking random observation from the dataset and we can use this observation to replace the missing value.

````python
import pandas as pd
import numpy as np

# Define a function to impute random values for missing entries
def impute_random_values(df, column):
    # Identify missing indices
    missing_indices = df[df[column].isnull()].index

    # Calculate the mean and standard deviation of the column
    column_mean = df[column].mean()
    column_std = df[column].std()

    # Generate random values with mean and std deviation
    random_values = np.random.normal(column_mean, column_std, len(missing_indices))

    # Update missing values with random values
    df.loc[missing_indices, column] = random_values

    return df
````

````python
# Create a sample DataFrame
data = {
    'Age': [25, 30, np.nan, 22, 28, np.nan, 35, 40, 29, np.nan],
    'Salary': [50000, 60000, 55000, np.nan, 75000, np.nan, 80000, np.nan, 70000, 72000]
}

df = pd.DataFrame(data)
# Display the original DataFrame
print("Original DataFrame:")
print(df)
````

````python
# Impute random values for missing entries in 'Age' and 'Salary'
df = impute_random_values(df, 'Age')
df = impute_random_values(df, 'Salary')

# Display the DataFrame after imputation
print("\nDataFrame after random imputation:")
print(df)
````

#### **5.	End of Distribution Imputation**

If there is suspicion that the missing value is not at random then capturing that information is important. In this scenario, one would want to replace missing data with values that are at the tails of the distribution of the variable. The advantage is that it is quick and captures the importance of missing values (if one suspects the missing data is valuable).

````python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Create a sample DataFrame with missing values
data = pd.DataFrame({
    'Age': [25, np.nan, 30, 22, np.nan, 40, 28, np.nan, 35],
    'Income': [50000, 60000, 75000, np.nan, 45000, np.nan, 70000, 55000, 80000]
})

print("Original data:")
print(data)

# Define a function for end-of-distribution imputation
def end_of_distribution_imputation(data, column_name):
    std_dev = data[column_name].std()
    mean = data[column_name].mean()
    extreme_value = mean + (3 * std_dev)  # 3 standard deviations above the mean

    data[column_name].fillna(extreme_value, inplace=True)
    return data

# Impute missing values using end-of-distribution approach
imputed_data = end_of_distribution_imputation(data, 'Age')
imputed_data = end_of_distribution_imputation(imputed_data, 'Income')

print("\nData after end-of-distribution imputation:")
print(imputed_data)
````

#### **6.	Frequent Categories Imputation**

#### **1. Using Standard Deviation in Symmetric Curve**

The standard deviation is a measure of how spread out the values in a dataset are around the mean. The idea behind using standard deviation for outlier detection is that data points that are significantly far from the mean are considered outliers.

**Here's how it works:**

- Calculate Mean and Standard Deviation: Calculate the mean (average) and standard deviation of the dataset.

- Define Threshold: Determine a threshold value (often denoted as k) that represents how many standard deviations away from the mean a value must be in order to be considered an outlier. Generally if any data point lies 3 standard deviations away from the mean we can say that data point is an outlier.

- Detect Outliers: Any data point whose value is more than k times the standard deviation away from the mean can be classified as an outlier. This can be determined using the formula: value > mean + k * std_dev or value < mean - k * std_dev.

- Outlier Treatment using Standard Deviation: Once outliers are detected, they can be treated using various methods. One common approach is to cap or clip the outlier values to a certain range, which can be determined based on the mean and standard deviation.

````python
import pandas as pd
import numpy as np

# Create a sample DataFrame with a symmetric distribution and outliers
data = pd.DataFrame({
    'Value': [20, 21, 19, 25, 24, 22, 23, 18, 30, 27, 26, 28, 100]
})

print("Original data:")
print(data)
````

````python
# Calculate mean and standard deviation
mean = data['Value'].mean()
std_dev = data['Value'].std()

print("Mean of the data: ", mean)
print("Standard deviation of the data: ", std_dev)
````

````python
# Detect outliers using standard deviation
std_dev_threshold = 3
lower_bound = mean - std_dev_threshold * std_dev
upper_bound = mean + std_dev_threshold * std_dev
outliers = data[(data['Value'] < lower_bound) | (data['Value'] > upper_bound)]

print("\nOutliers detected using standard deviation:")
print(outliers)
````

````python
# Define a function to treat outliers using standard deviation
def treat_outliers_std(data, column_name, std_dev_threshold):
    lower_bound = mean - std_dev_threshold * std_dev
    upper_bound = mean + std_dev_threshold * std_dev
    data[column_name] = np.clip(data[column_name], lower_bound, upper_bound)
    return data

# Treat outliers using standard deviation
data_treated = treat_outliers_std(data.copy(), 'Value', std_dev_threshold)

print("\nData after outlier treatment using standard deviation:")
print(data_treated)
````

Using the standard deviation for outlier detection and treatment assumes that the data is approximately normally distributed (symmetric). It's a simple and intuitive method, but it might not be suitable for all types of data distributions. Additionally, the choice of the threshold value (k) requires careful consideration to balance between sensitivity to outliers and preserving data integrity.

Remember that outlier treatment should be done thoughtfully and in line with the characteristics of your data and the goals of your analysis.

#### **2. Using IQR in Skew-symmetric Curve**

The box plot is a useful graphical display for describing the behavior of the data in the middle as well as at the ends of the distributions. The box plot uses the median and the lower and upper quartiles (defined as the 25th and 75th percentiles). If the lower quartile is Q1 and the upper quartile is Q3, then the difference (Q3 ‚Äî Q1) is called the interquartile range or IQR.

A box plot is constructed by drawing a box between the upper and lower quartiles with a solid line drawn across the box to locate the median. The following quantities (called fences) are needed for identifying extreme values in the tails of the distribution:

1.	lower inner fence: Q1‚Äì1.5*IQ
2.	upper inner fence: Q3 + 1.5*IQ
3.	lower outer fence: Q1‚Äì3*IQ
4.	upper outer fence: Q3 + 3*IQ


````python
import pandas as pd
import numpy as np

# Create a sample DataFrame with a skew-symmetric distribution and outliers
data = pd.DataFrame({
    'Value': [10, 12, 13, 15, 16, 17, 18, 19, 20, 22, 23, 24, 100]
})

print("Original data:")
print(data)
````

````python
# Calculate the first quartile (Q1) and third quartile (Q3)
q1 = data['Value'].quantile(0.25)
q3 = data['Value'].quantile(0.75)

# Calculate the Interquartile Range (IQR)
iqr = q3 - q1

# Detect outliers using IQR
iqr_multiplier = 1.5
lower_bound = q1 - iqr_multiplier * iqr
upper_bound = q3 + iqr_multiplier * iqr
outliers = data[(data['Value'] < lower_bound) | (data['Value'] > upper_bound)]

print("\nOutliers detected using IQR:")
print(outliers)
````

````python
# Define a function to treat outliers using IQR
def treat_outliers_iqr(data, column_name, iqr_multiplier):
    lower_bound = q1 - iqr_multiplier * iqr
    upper_bound = q3 + iqr_multiplier * iqr
    data[column_name] = np.clip(data[column_name], lower_bound, upper_bound)
    return data

# Treat outliers using IQR
data_treated = treat_outliers_iqr(data.copy(), 'Value', iqr_multiplier)

print("\nData after outlier treatment using IQR:")
print(data_treated)
````

### **Categorical Encoding**

This means that categorical data must be encoded into numbers before we can use it to fit and evaluate a model. There are many ways to encode categorical variables for modeling as follows.

#### **1.Nominal Encoding**

Variable comprises a finite set of discrete values with no relationship between values. Some examples include:

‚Ä¢	A ‚Äúpet‚Äù variable with the values: ‚Äúdog‚Äù and ‚Äúcat‚Äú.

‚Ä¢	A ‚Äúcolor‚Äù variable with the values: ‚Äúred‚Äú, ‚Äúgreen‚Äú, and ‚Äúblue‚Äú.

‚Ä¢	A ‚Äúplace‚Äù variable with the values: ‚Äúfirst‚Äú, ‚Äúsecond‚Äú, and ‚Äúthird‚Äú


**1.	One Hot Encoding**

One-hot encoding, also known as dummy coding, is a technique where each category is transformed into a binary column. Each category becomes a new binary column, and the presence of the category is indicated by a 1, while other columns remain 0. This approach is suitable when the categorical variable has no ordinal relationship.

**Example:**

Suppose you have a categorical variable "Color" with categories "Red," "Blue," and "Green." One-hot encoding transforms these categories into binary columns as follows:

| Color | Red | Blue | Green |
|-------|-----|------|-------|
| Red   | 1   | 0    | 0     |
| Blue  | 0   | 1    | 0     |
| Green | 0   | 0    | 1     |

In this example, each category "Red," "Blue," and "Green" has been transformed into its own binary column, indicating its presence or absence in each row of the dataset.


**2.	One Hot Encoding with many  Categorical**

When dealing with datasets that have multiple categorical variables, you might encounter scenarios where you need to apply one-hot encoding to all of these variables. One-hot encoding individually for each variable creates binary columns for each category, indicating its presence or absence. It is the same as one-hot encoding but the difference is Let you have more than 20 unique categories. Then if you will perform one-hot encoding to all those you will have additional 20 columns. This will lead to the Curse of Dimensionality. So, It has been observed from the KDD Orange Cup challenge that some top n no. of maximum categories are taken and encoded.

**Example:**

Suppose you have a dataset with two categorical variables, "Color" and "Size." The categories for each variable are as follows:

| Color  | Size   |
|--------|--------|
| Red    | Small  |
| Blue   | Medium |
| Green  | Large  |

After applying one-hot encoding, the categorical variables are transformed into binary columns as shown below:

**Encoded Values:**

| Color | Red | Blue | Green | Size   | Small | Medium | Large |
|-------|-----|------|-------|--------|-------|--------|-------|
| Red   | 1   | 0    | 0     | Small  | 1     | 0      | 0     |
| Blue  | 0   | 1    | 0     | Medium | 0     | 1      | 0     |
| Green | 0   | 0    | 1     | Large  | 0     | 0      | 1     |

Each categorical variable has been transformed into separate binary columns, capturing the presence or absence of each category in the dataset. This expanded representation is helpful for machine learning algorithms to understand the categorical information.


**3. Mean Encoding (Target Encoding)**

Mean Encoding, also known as Target Encoding, is a technique used to transform categorical variables into a numeric format based on the mean (or other statistical metrics) of the target variable for each category. This technique is particularly useful when dealing with categorical variables in regression or classification tasks.

**How it Works:**

1. **Calculate the Mean for Each Category:**
   - For each category in the categorical variable, calculate the mean of the target variable for the samples that belong to that category.

2. **Replace Categories with Mean Values:**
   - Replace the original category labels with the corresponding mean values calculated in the previous step.

**Example:**

Suppose you have a dataset with a categorical variable "Color" and a target variable "Price." The "Color" variable has categories "Red," "Blue," and "Green," and the target variable "Price" contains the prices of products. The mean encoding process would look like this:

| Color  | Price in Dollars |
|--------|-------|
| Red    | 100  |
| Blue   | 150  |
| Green  | 120  |
| Blue   | 160  |
| Red    | 110  |

**Encoded Values:**

| Color  | Encoded Price Mean |
|--------|---------------------|
| Red    | 105                |
| Blue   | 155                |
| Green  | 120                |
| Blue   | 155                |
| Red    | 105                |

In this example, the original "Color" categories are replaced with the mean prices of products that belong to each category. The mean encoding provides a way to incorporate categorical information into your models while leveraging the relationship between categories and the target variable.


#### **2. Ordinal Encoding**

Variable comprises a finite set of discrete values with a ranked ordering between values. In ordinal encoding, each unique category value is assigned an integer value.

For example, ‚Äúred‚Äù is 1, ‚Äúgreen‚Äù is 2, and ‚Äúblue‚Äù is 3.


**1. Label Encoding**


Label Encoding is a technique used to convert categorical variables into numerical format. In this method, each unique category is assigned a unique integer value. This technique is suitable when the categorical variable has an ordinal relationship or when working with algorithms that require numeric input.

**How it Works:**

1. **Assign Integer Labels:**
   - For each unique category in the categorical variable, assign a unique integer label starting from 0 or 1.

2. **Replace Categories with Labels:**
   - Replace the original category labels with the corresponding integer labels assigned in the previous step.

**Example:**

Suppose you have a dataset with a categorical variable "Size," which has categories "Small," "Medium," and "Large." The label encoding process would look like this:

| Size    |
|---------|
| Medium  |
| Small   |
| Large   |
| Medium  |
| Small   |

**Encoded Values:**

| Size    | Encoded Label |
|---------|---------------|
| Medium  | 1             |
| Small   | 0             |
| Large   | 2             |
| Medium  | 1             |
| Small   | 0             |

In this example, the original "Size" categories are replaced with integer labels. Label Encoding is simple to implement and can work well with algorithms that understand numeric values. However, it's important to note that some algorithms might misinterpret the numeric labels as having ordinal relationships, even when they don't exist.


````python
#Example Code
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Create a sample DataFrame with a categorical variable
data = pd.DataFrame({
    'Size': ['Medium', 'Small', 'Large', 'Medium', 'Small']
})

print("Original data:")
print(data)

# Initialize LabelEncoder
encoder = LabelEncoder()

# Fit and transform the encoder on the categorical variable
encoded_data = encoder.fit_transform(data['Size'])

# Create a new column with the encoded values
data['Encoded_Size'] = encoded_data

print("\nEncoded data:")
print(data)
````

**2. Target Guided Ordinal Encoding**



Target Guided Ordinal Encoding is a technique used to convert categorical variables into numerical format while considering the relationship between the categories and the target variable. It assigns labels to categories based on the mean of the target variable for each category. This method is especially useful when the categorical variable exhibits a clear ordinal relationship with the target variable.so, in other words, ‚Äútarget guided ordinal encoding = mean encoding +label encoding.‚Äù

**How it Works:**

1. **Calculate Mean Target for Each Category:**
   - For each category in the categorical variable, calculate the mean of the target variable for the samples that belong to that category.

2. **Assign Labels Based on Mean Target:**
   - Assign labels to categories in ascending order of their mean target values. Categories with higher mean target values receive higher labels.

**Example:**

Suppose you have a dataset with a categorical variable "Education" and a target variable "Income." The "Education" variable has categories "Bachelor," "Master," and "PhD," and the target variable "Income" represents income levels. The target guided ordinal encoding process would look like this:

| Education | Income |
|-----------|--------|
| Bachelor  | 50000 |
| Master    | 75000 |
| PhD       | 60000 |
| Bachelor  | 55000 |
| PhD       | 70000 |

**Encoded Values:**

| Education | Encoded Education |
|-----------|-------------------|
| Bachelor  | 1                 |
| Master    | 2                 |
| PhD       | 3                 |
| Bachelor  | 1                 |
| PhD       | 3                 |

In this example, the "Education" categories are assigned labels based on their mean income levels. "Bachelor" is assigned the lowest label (1) since it has the lowest mean income, while "PhD" is assigned the highest label (3) due to its higher mean income. Target guided ordinal encoding leverages the relationship between categories and the target variable, making it suitable for ordinal categorical variables.


````python
import pandas as pd

# Create a sample DataFrame with categorical variable and target variable
data = pd.DataFrame({
    'Education': ['Bachelor', 'Master', 'PhD', 'Bachelor', 'PhD'],
    'Income': [50000, 75000, 60000, 55000, 70000]
})

print("Original data:")
print(data)

# Calculate mean income for each education level
mean_income_by_education = data.groupby('Education')['Income'].mean().sort_values()

# Create a dictionary to map education levels to labels based on mean income
education_labels = {education: label + 1 for label, education in enumerate(mean_income_by_education.index)}

# Apply target guided ordinal encoding
data['Encoded_Education'] = data['Education'].map(education_labels)

print("\nEncoded data:")
print(data)
````

**3.	Count or Frequency Encoding**



Count or Frequency Encoding is a technique used to convert categorical variables into numerical format based on the frequency or count of each category in the dataset. It assigns labels to categories based on how often each category appears. This method is suitable when the categorical variable doesn't have an ordinal relationship and when you want to capture the prevalence of each category.

**How it Works:**

1. **Calculate Frequency of Each Category:**
   - For each category in the categorical variable, calculate the frequency or count of how many times the category appears in the dataset.

2. **Assign Labels Based on Frequency:**
   - Assign labels to categories based on their frequency in ascending or descending order. Categories with higher frequencies receive higher labels.

**Example:**

Suppose you have a dataset with a categorical variable "Color." The "Color" variable has categories "Red," "Blue," and "Green." The count or frequency encoding process would look like this:

| Color  |
|--------|
| Red    |
| Blue   |
| Green  |
| Blue   |
| Red    |

**Encoded Values:**

| Color  | Encoded Color |
|--------|---------------|
| Red    | 2             |
| Blue   | 2             |
| Green  | 1             |
| Blue   | 2             |
| Red    | 2             |

In this example, the "Color" categories are assigned labels based on their frequency. "Blue" appears the most frequently (count = 3), so it receives the highest label (2), while "Green" appears the least frequently (count = 1) and gets the lowest label (1). Count or frequency encoding helps capture the distribution of categories in the dataset.


````python
import pandas as pd

# Create a sample DataFrame with a categorical variable
data = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green', 'Blue', 'Red']
})

# Print the original data
print("Original data:")
print(data)

# Calculate frequency of each color using value_counts()
color_counts = data['Color'].value_counts()

# Create a dictionary to map colors to labels based on frequency with equal ranking for ties
color_labels = {color: count for color, count in color_counts.iteritems()}

# Apply count or frequency encoding using the map() function
data['Encoded_Color'] = data['Color'].map(color_labels)

# Print the encoded data
print("\nEncoded data:")
print(data)
````

**4. Probability Ratio Encoding**




This encoding is purely based on probability. If it‚Äôs possible to calculate the probability or it‚Äôs a probability type problem, You can use this type of encoding by calculating the probabilities. For example, If you are working with some dataset like survive and not- survive in ‚ÄúTitanic Dataset‚Äù. There you can use probability encoding.



Suppose we have a dataset with a categorical variable "Category" and a binary target variable "Outcome" (1 for positive outcome, 0 for negative outcome). We want to perform Probability Ratio Encoding on the "Category" variable to encode it based on the likelihood of a positive outcome for each category. For example in analogy to the titanic datast the category could be the "cabin" feature and the outcome could be "survived" feature

**Dataset:**

| Category | Outcome |
|----------|---------|
| A        | 1       |
| B        | 0       |
| A        | 1       |
| C        | 0       |
| B        | 1       |
| C        | 1       |
| A        | 0       |
| B        | 0       |

**Step-by-Step Process:**

 **1. Calculate Probability for Each Category and Target:**

   Calculate the probability of a positive outcome (target = 1) and a negative outcome (target = 0) for each category:
   
   | Category | P(Outcome=1) | P(Outcome=0) |
   |----------|--------------|--------------|
   | A        | 2/3 (0.67)   | 1/3 (0.33)   |
   | B        | 1/3 (0.33)   | 2/3 (0.67)   |
   | C        | 1/2 (0.50)   | 1/2 (0.50)   |

**2. Calculate Probability Ratio:**

   Calculate the probability ratio by dividing the probability of a positive outcome by the probability of a negative outcome for each category:
   
   | Category | Probability Ratio |
   |----------|-------------------|
   | A        | 2.0               |
   | B        | 0.5               |
   | C        | 1.0               |

**3. Map Categories to Probability Ratios:**

   Replace the original "Category" categories with the calculated probability ratios:
   
   | Category | Encoded Outcome Probability Ratio |
   |----------|----------------------------------|
   | A        | 2.0                              |
   | B        | 0.5                              |
   | A        | 2.0                              |
   | C        | 1.0                              |
   | B        | 0.5                              |
   | C        | 1.0                              |
   | A        | 2.0                              |
   | B        | 0.5                              |


**4. Apply Encoding to the Dataset:**

   Map the probability ratio-encoded values to the original "Category" column:
   
   | Category | Outcome | Encoded Outcome Probability Ratio |
   |----------|---------|----------------------------------|
   | A        | 1       | 2.0                              |
   | B        | 0       | 0.5                              |
   | A        | 1       | 2.0                              |
   | C        | 0       | 1.0                              |
   | B        | 1       | 0.5                              |
   | C        | 1       | 1.0                              |
   | A        | 0       | 2.0                              |
   | B        | 0       | 0.5                              |

In this example, Probability Ratio Encoding is applied to the "Category" variable. The encoding captures the likelihood of a positive outcome for each category. Categories with a higher probability of a positive outcome are assigned higher probability ratio-encoded values.


````python
import pandas as pd

# Creating the dataset
data = {
    'Category': ['A', 'B', 'A', 'C', 'B', 'C', 'A', 'B'],
    'Outcome': [1, 0, 1, 0, 1, 1, 0, 0]
}

df = pd.DataFrame(data)

# Step 1: Calculate probability for each category and target
prob_df = df.groupby('Category')['Outcome'].mean()
prob_df = prob_df.to_frame(name='P(Outcome=1)')

# Calculate probability of no outcome (Outcome=0)
prob_df['P(Outcome=0)'] = 1 - prob_df['P(Outcome=1)']

# Step 2: Calculate probability ratio
prob_df['Probability Ratio'] = prob_df['P(Outcome=1)'] / prob_df['P(Outcome=0)']

# Step 3: Map categories to probability ratios
prob_encode = prob_df['Probability Ratio'].to_dict()
df['Encoded Outcome Probability Ratio'] = df['Category'].map(prob_encode)

# Display the encoded dataset
print(df)
````

‚Ä¢	A value closer to 1 implies stronger positive correlation

‚Ä¢	A value closer to -1 implies stronger negative correlation


````python
#Example Code
def correlation(df, threshold):
    col_corr = set()  # Set to store correlated column names
    corr_matrix = df.corr()  # Calculate the correlation matrix

    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            corr_value = corr_matrix.iloc[i, j]  # Get the correlation value
            if abs(corr_value) > threshold:
                colname = corr_matrix.columns[i]  # Get the name of the correlated column
                col_corr.add((colname, corr_value))  # Add the column name and correlation value to the set

    return col_corr  # Return the set of correlated column names and their correlation values
````

Here's how you can use the funtion with the example DataFrame:

````python
import pandas as pd

# Create a sample DataFrame
data = {
    'Feature1': [1, 2, 3, 4, 5],
    'Feature2': [5, 4, 3, 2, 1],
    'Feature3': [2, 2, 2, 2, 2],
    'Target': [0, 1, 0, 1, 0]
}

df = pd.DataFrame(data)

# Define the correlation threshold
threshold = 0.5

# Call the correlation function
highly_correlated_info = correlation(df, threshold)

print("Columns with high correlation:")
for colname, corr_value in highly_correlated_info:
    print(f"{colname} with correlation value: {corr_value}")
````

**2. Variable inflation Factor**

Collinearity is the state where two variables are highly correlated and contain similar information about the variance within a given dataset. To detect collinearity among variables, simply create a correlation matrix and find variables with large absolute values.

Steps for Implementing VIF

‚Ä¢	Calculate the VIF .

‚Ä¢	Inspect the factors for each predictor variable, if the VIF is between 5‚Äì10, multicollinearity is likely present and you should consider dropping the variable.

In VIF method, we pick each feature and regress it against all of the other features. For each regression, the factor is calculated as :

VIF=$\frac{1}{1-R^2}$

Where, R-squared is the coefficient of determination in linear regression. Its value lies between 0 and 1.


````python
import pandas as pd
from statsmodels.stats.outliers_influence import variance_inflation_factor

# Create a sample dataset
data = {
    'Income': [50000, 60000, 70000, 55000, 75000],
    'Age': [28, 32, 45, 29, 38],
    'Experience': [5, 8, 15, 6, 10],
    'Target': [1, 0, 1, 0, 1]
}

df = pd.DataFrame(data)

# Separate features and target variable
X = df[['Income', 'Age', 'Experience']]
y = df['Target']

# Calculate VIF for each variable
vif_data = pd.DataFrame()
vif_data["Variable"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print("VIF values:")
print(vif_data)
````

**6. Embedded Method**

Embedded methods are iterative in a sense that takes care of each iteration of the model training process and carefully extract those features which contribute the most to the training for a particular iteration.



In this example, we load the Iris dataset from sklearn.datasets. The dataset contains features related to iris flowers and a target variable "Species" representing the species of the iris flowers. The embedded_feature_selection function is used with Lasso Regression to perform feature selection. The alpha parameter is set to 0.01 for regularization.

Please note that the Iris dataset is used for illustrative purposes, and the concept of embedded feature selection can be applied to more complex datasets as well.

````python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.linear_model import Lasso

def embedded_feature_selection(X, y, alpha=1.0):
    model = Lasso(alpha=alpha)
    model.fit(X, y)
    selected_features = X.columns[model.coef_ != 0]
    return selected_features

# Load the Iris dataset
iris_data = load_iris()
X = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
y = pd.Series(iris_data.target, name='Species')

# Call the embedded_feature_selection function
selected_features = embedded_feature_selection(X, y, alpha=0.01)

print("Selected features:", selected_features)
````

Fisher Score is a statistical method used for feature selection to identify the importance of each feature in distinguishing between different classes or groups within the target variable. It is particularly useful for classification problems with categorical or discrete features. Fisher Score is based on the Fisher's Discriminant Ratio, which quantifies the ratio of the variance between classes to the variance within classes.

**Here's how Fisher Score works:**

- **Calculate Mean and Variance:** Compute the mean and variance of each feature for each class separately.
Calculate the mean of each feature across all classes and the total variance within each class.

- **Calculate Fisher's Discriminant Ratio (FDR):** The Fisher's Discriminant Ratio for a feature is calculated as the ratio of the between-class variance to the within-class variance.
Select Features:

Features with higher Fisher's Discriminant Ratio are considered more important, as they exhibit higher variance between classes compared to the variance within classes.Select the top features with the highest FDR values as they contribute the most to the separability between classes.

Here's an example of implementing Fisher Score for feature selection in Python using the sklearn library:

````python
import pandas as pd
from sklearn.datasets import load_iris
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2

# Load the Iris dataset
iris_data = load_iris()
X = pd.DataFrame(iris_data.data, columns=iris_data.feature_names)
y = pd.Series(iris_data.target, name='Species')

# Select the top 2 features using Fisher Score
num_features = 2
fisher_selector = SelectKBest(chi2, k=num_features)
X_selected = fisher_selector.fit_transform(X, y)

# Get the indices of selected features
selected_indices = fisher_selector.get_support(indices=True)

# Get the names of selected features
selected_features = X.columns[selected_indices]

print("Selected features using Fisher Score:")
print(selected_features)
````

### <b> Learning the math behind PCA </b>

<b> Step 1 : Take the whole dataset with dimension $d$ (which corresponds to the $d$ features) </b>


<b> Step 2 : Compute the mean of every dimension of the whole dataset. </b>

<b> Step 3 : Compute the covariance matrix of the whole dataset </b>
* ### \begin{align*}
  Cov(X,Y) &= \frac{1}{n-1}\sum_{i=1}^{n} (X_{i} - \bar X)(Y_{i} - \bar Y) \\
\end{align*}

<b> Step 4 : Compute Eigenvectors and corresponding Eigenvalues for the covariance matrix A </b>
* Let A be a square matrix, ŒΩ a vector and Œª a scalar that satisfies $AŒΩ = ŒªŒΩ$, then Œª is called eigenvalue associated with eigenvector ŒΩ of A.

<b> Step 5 : Sort the eigenvectors by decreasing eigenvalues and choose $k$ eigenvectors with the largest eigenvalues to form a $d √ó k$ dimensional matrix $W$. </b>

<b> Step 6 : Transform the samples onto the new subspace</b>
* In the last step, we use the $d √ó k$ dimensional matrix $W$ that we just computed to transform our samples onto the new subspace via the equation $y = W‚Ä≤ √ó x$ where $W‚Ä≤$ is the transpose of the matrix $W$.
