##**Important formula in Machine learning:**

1) **Gradient Descent:** $\theta_{j+1} = \theta_j - \alpha\nabla J(\theta_j)$
Definition: An optimization algorithm used to find the values of parameters $(\theta)$ that minimize a cost function $(J)$ over iterations.
Use: Training machine learning models by updating weights/coefficients to minimize the cost/error.

2) **Normal Distribution:** $f(x|\mu,\sigma^2) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{(x-\mu)^2}{2\sigma^2}\right)$
Definition: A continuous probability distribution describing many natural phenomena.
Use: Modeling data that clusters around a mean, used in statistics and probability theory.

3)**Z-score:** $z = \frac{x - \mu}{\sigma}$
Definition: The number of standard deviations a data point is away from the mean.
Use: Standardizing data to have a mean of 0 and standard deviation of 1.

4) **Sigmoid:** $\sigma(x) = \frac{1}{1 + e^{-x}}$
Definition: A mathematical function that maps any real value to between 0 and 1.
Use: Activation function in logistic regression and neural networks.

5) **Correlation:** $\text{Correlation} = \frac{\text{Cov}(X,Y)}{\text{Std}(X) \cdot \text{Std}(Y)}$
Definition: A measure of the linear relationship between two variables.
Use: Analyzing the strength and direction of relationships between variables.

6) **Cosine Similarity:** $\text{similarity}= \frac{A \cdot B}{\|A\| \|B\|}$
Definition: A measure of similarity between two non-zero vectors.
Use: Text mining, information retrieval, data clustering.

7) **Naive Bayes:** $P(y|x_1,...,x_n) = \frac{P(y)\prod_{i=1}^{n} P(x_i|y)}{P(x_1,...,x_n)}$
Definition: A probabilistic classifier based on Bayes' theorem with naive independence assumptions.
Use: Text classification, spam filtering, recommendation systems.

8) **MLE:** $\arg\max_\theta \prod_{i=1}^{n} P(x_i|\theta)$
Definition: The parameter value that maximizes the likelihood function.
Use: Estimating parameters in probabilistic models like logistic regression.

9) **OLS:** $\boldsymbol{\beta} = (X^T X)^{-1} X^T y$
Definition: Ordinary Least Squares estimator for linear regression coefficients.
Use: Finding the best-fitting linear model for data.

10) **F1 Score:** $\frac{2 \cdot P \cdot R}{P + R}$
Definition: The harmonic mean of precision and recall.
Use: Evaluating binary classification models with a single score.

11) **ReLU:** $\max(0, x)$
Definition: A non-linear activation function that outputs $x$ if positive, 0 otherwise.
Use: Introducing non-linearity in neural networks.

12) **Softmax:** $P(y_i = j|x) = \frac{e^{x_j}}{\sum_{k=1}^{K} e^{x_k}}$
Definition: A function that converts logits to a probability distribution over classes.
Use: Output layer for multi-class classification neural networks.

13) **R2 score:** $R^2 = 1 - \frac{\sum_{i=1}^{n} (y_i - \hat{y}_i)^2}{\sum_{i=1}^{n} (y_i - \bar{y})^2}$
Definition: A measure of how well a regression model fits the data.
Use: Evaluating the predictive performance of regression models.

14) **MSE:** $\text{MSE} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2$
Definition: The average squared difference between predicted and actual values.
Use: Evaluating regression and forecasting model performance.

15) **MSE + L2 Reg:** $\text{MSE}_{\text{regularized}} = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2 + \lambda \sum_{j=1}^{p} \beta_j^2$
Definition: MSE with an L2 regularization penalty term.
Use: Preventing overfitting in linear models and neural networks.

16) **Eigen vectors:** $Av = \lambda v$
Definition: Non-zero vectors that are scaled by a scalar value (eigenvalue) under a linear transformation.
Use: Principal Component Analysis (PCA), Eigen decompositions.

17) **Entropy:** $\text{Entropy} = -\sum_{i=1}^{n} p_i \log_2(p_i)$
Definition: A measure of the average information content or uncertainty in a random variable.
Use: Information theory, data compression, decision trees.

18) **KMeans:** $\arg\min_S \sum_{i=1}^{K} \sum_{x \in S_i} \|x - \mu_i\|^2$
Definition: An iterative clustering algorithm that partitions data into K clusters.
Use: Unsupervised clustering of unlabeled data.

19) **KL Divergence:** $D_{KL}(P|Q) = \sum_{x \in \mathcal{X}} P(x) \log\left(\frac{P(x)}{Q(x)}\right)$
Definition: A measure of how different two probability distributions are.
Use: Comparing learned and true distributions in probabilistic models.

20) **Log-loss:** $\frac{1}{N} \sum_{i=1}^{N} \left( y_i \log(\hat{y}_i) + (1 - y_i)\log(1 - \hat{y}_i) \right)$
Definition: The negative log-likelihood for binary classification.
Use: Evaluating and optimizing logistic regression and neural network classifiers.

21) **SVM:** $\min \frac{1}{2} \|w\|^2 + C \sum_{i=1}^{n} \max\left(0, 1 - y_i(w \cdot x_i - b)\right)$
Definition: A discriminative classifier that finds the optimal hyperplane separating classes.
Use: Classification, regression, and novelty detection.

22) **Linear Regression:** $y = \beta_0 + \beta_1x_1 + \beta_2x_2 + \cdots + \beta_nx_n + \epsilon$
Definition: A linear model with a sum of weighted inputs and a constant term.
Use: Predicting a continuous target variable from input features.

23) **SVD:** $A = U\Sigma V^T$
Definition: Singular Value Decomposition of a matrix into three component matrices.
Use: Data compression, recommendation systems, signal processing.

24) **Lagrange Multiplier:** $\max f(x) \; \text{s.t.} \; g(x) = 0, \; L(x, \lambda) = f(x) - \lambda \cdot g(x)$
Definition: A strategy to find the local maxima/minima of an objective function subject to constraints.
Use: Optimization under equality constraints.
