# **K-Nearest Neighbors (KNN)**

## Definition
K-Nearest Neighbors (KNN) is a non-parametric and lazy learning algorithm used for classification and regression tasks. It is considered a simple yet effective algorithm in the field of supervised learning.

## How It Works
1. **Initialization**: During the training phase, KNN stores all available data points and their corresponding labels.
2. **Prediction**: When making predictions for a new data point, KNN calculates the distance between the new point and all other points in the training set.
3. **Nearest Neighbors**: It then selects the K nearest neighbors based on the calculated distances.
4. **Voting**: For classification tasks, KNN assigns the class label that is most frequent among the K nearest neighbors. For regression tasks, it takes the average of the K nearest neighbors' target values.

## Mathematical Formula
The KNN algorithm uses the Euclidean distance formula to measure the distance between two points $X$ and $Y$ in $n$-dimensional space:

$$
\text{Euclidean Distance}(X, Y) = \sqrt{\sum_{i=1}^{n} (X_i - Y_i)^2}
$$

## Types
- **Classification**: KNN assigns the class label based on the majority vote of its nearest neighbors.
- **Regression**: KNN predicts the value of a continuous target variable by averaging the values of its nearest neighbors.

## Assumptions
- KNN assumes that similar instances tend to have similar class labels or target values.
- It also assumes that the data space can be divided into regions with different classes or target values.

## Limitations
- **Computational Complexity**: KNN can be computationally expensive, especially for large datasets, as it requires calculating distances for each query instance.
- **Memory Usage**: KNN stores the entire training dataset in memory, which can be prohibitive for large datasets.
- **Sensitivity to Noise and Irrelevant Features**: KNN can perform poorly if the dataset contains noisy or irrelevant features.

## Advantages
- **Simple to Implement**: KNN is easy to understand and implement, making it suitable for beginners in machine learning.
- **No Training Phase**: KNN is a lazy learner, meaning it does not require a training phase. It directly uses the training data for prediction.
- **Non-Parametric**: KNN makes no assumptions about the underlying data distribution, making it versatile for different types of data.

## Disadvantages
- **Slow Prediction Time**: KNN's prediction time can be slow, especially for large datasets, as it needs to compute distances for each query instance.
- **High Memory Usage**: Storing the entire dataset in memory can be memory-intensive, especially for large datasets.
- **Sensitive to Outliers**: Outliers in the dataset can significantly impact the performance of KNN, as it relies on the nearest neighbors for prediction.

## Structure
- **Training Data**: KNN stores the entire training dataset in memory.
- **Distance Metric**: It uses distance metrics such as Euclidean distance, Manhattan distance, or cosine similarity to measure the similarity between data points.
- **K Value**: The number of nearest neighbors to consider, which is a hyperparameter that needs to be tuned.

## Examples and Use Cases
- **Classification**: Classifying emails as spam or non-spam based on their features.
- **Regression**: Predicting the price of a house based on its features like area, number of bedrooms, etc.
- **Anomaly Detection**: Identifying outliers or anomalies in data.
- **Recommendation Systems**: Recommending products or services based on similar users' preferences.

## Notable Features
- **Instance-Based Learning**: KNN is an instance-based learning algorithm, meaning it memorizes the training instances and uses them directly during the prediction phase.
- **Adaptability**: KNN can adapt to changes in the dataset without requiring retraining, making it suitable for dynamic environments.

In summary, KNN is a versatile and simple algorithm for classification and regression tasks. While it has its limitations, such as computational complexity and sensitivity to noise, it remains a popular choice due to its ease of implementation and adaptability to different types of data.

## **K-Nearest Neighbors (KNN) Classifier**

## Description

K-Nearest Neighbors (KNN) is a simple and effective supervised learning algorithm used for classification and regression tasks. In classification, it predicts the class of a given data point by analyzing the 'k' nearest data points in the feature space. The class label of the majority of these neighbors determines the class label of the given data point.

## Definition

Given a dataset with labeled instances, the KNN algorithm classifies new instances based on their similarity to existing instances. It assumes that similar instances exist in close proximity in the feature space.

## Assumptions

- Instances in the same class are close to each other in the feature space.
- The data has a meaningful distance metric to measure the similarity between instances.

## Mathematical Formula

Given a query instance $x_q$, the predicted class $\hat{y}$ is determined by a majority vote among its 'k' nearest neighbors:

$\hat{y} = \text{mode}(y_i)$ for $i = 1, 2, ..., k$

where $y_i$ represents the class labels of the 'k' nearest neighbors.

## Advantage

- Simple to understand and implement.
- Non-parametric: Does not make any assumptions about the underlying data distribution.
- Can be used for both classification and regression tasks.
- Robust to noisy data.

## Disadvantage

- Computationally expensive: Requires storing and computing distances for all instances in the dataset.
- Sensitive to the choice of 'k' and the distance metric.
- Performance may degrade with high-dimensional data.
- Not suitable for large datasets due to its high computational cost during inference.

## Structure

1. **Training Phase**: Store all training instances with their respective class labels.
2. **Prediction Phase**: For each new instance, find its 'k' nearest neighbors based on a distance metric (e.g., Euclidean distance), and predict its class label based on the majority class among these neighbors.

## How It Works

1. Calculate the distance between the query instance and all training instances using a chosen distance metric.
2. Sort the distances and select the 'k' nearest neighbors.
3. Perform a majority vote among these neighbors to determine the predicted class label for the query instance.

## Notable Features and Functionality

- Flexibility in choosing distance metrics (e.g., Euclidean, Manhattan, Minkowski).
- Can handle multi-class classification problems.
- Can be combined with dimensionality reduction techniques like PCA to improve performance in high-dimensional spaces.

## Examples and Use Cases

- **Medical Diagnosis**: Classify patients into different disease categories based on their symptoms and medical history.
- **Recommendation Systems**: Predict user preferences based on the preferences of similar users.
- **Anomaly Detection**: Identify outliers by classifying instances that deviate significantly from the majority of data points.

KNN provides a versatile and intuitive approach to classification tasks, especially when the data has a clear underlying structure and the computational cost is not prohibitive. However, its performance heavily depends on the choice of parameters and the quality of the data.

##**K-Nearest Neighbors (KNN) Regressor**

### Definition:
K-Nearest Neighbors (KNN) Regressor is a type of supervised learning algorithm used for regression tasks. Unlike classification algorithms that predict a class label for a given input, the KNN regressor predicts a continuous value for the target variable based on the average or weighted average of the $k$ nearest data points.

### Types:
KNN Regressor is a type of instance-based learning algorithm. It doesn't involve explicit training, rather it memorizes the entire training dataset and makes predictions based on similarity measures between new instances and existing instances in the training data.

### Assumptions:
- KNN assumes that similar data points have similar target values.
- It assumes that the feature space is continuous.

### Limitations:
- Computationally expensive, especially on large datasets as it requires calculating distances between the new instance and all training instances.
- Sensitive to irrelevant features and noisy data.
- Choice of the parameter $k$ can significantly affect the performance of the model.

### Mathematical Formula:
The predicted value $\hat{y}$ for a new instance $x$ is calculated as the average or weighted average of the target values of its $k$ nearest neighbors:

$$\hat{y} = \frac{1}{k} \sum_{i=1}^{k} y_i$$

or

$$\hat{y} = \frac{\sum_{i=1}^{k} w_i \cdot y_i}{\sum_{i=1}^{k} w_i}$$

Where:
- $\hat{y}$ is the predicted value.
- $y_i$ is the target value of the $i^{th}$ nearest neighbor.
- $w_i$ is the weight assigned to the $i^{th}$ nearest neighbor (if weighted KNN is used).

### Advantage:
- Simple and easy to implement.
- Non-parametric method, thus it doesn't make any assumptions about the underlying data distribution.
- Can be effective for small to medium-sized datasets with fewer dimensions.

### Disadvantage:
- Computationally expensive during prediction phase, especially with large datasets.
- Performs poorly with high-dimensional data due to the curse of dimensionality.
- Choice of $k$ can be critical and may require tuning.

### Structure:
- KNN Regressor does not have a specific structure as it doesn't learn explicit parameters from the data. Instead, it stores the entire training dataset for prediction.

### How it Works:
1. Calculate the distance between the new instance and all instances in the training dataset using a distance metric like Euclidean distance.
2. Select the $k$ instances with the smallest distances.
3. Calculate the predicted value for regression by averaging the target values of the $k$ nearest neighbors.

### Notable Features/Functionality:
- KNN can handle both regression and classification tasks.
- It's a lazy learner as it doesn't build a model during training, rather it memorizes the training data for later use in prediction.
- Can handle multi-output regression problems.

### Examples and Use Cases:
- Predicting house prices based on features like area, number of bedrooms, etc.
- Estimating stock prices based on historical data and relevant market features.
- Forecasting demand for products based on historical sales data and other factors.

KNN Regressor can be a useful algorithm in situations where the underlying data distribution is not known, and a quick and simple model is needed for prediction. However, its performance heavily depends on the choice of the parameter $k$ and the quality of the training data.

### **Mathematical Formulation**

The goal of Neighborhood Component Analysis (NCA) is to learn an optimal linear transformation matrix of size $(n_{\text{components}}, n_{\text{features}})$, which maximizes the sum over all samples $i$ of the probability $P_i$ that sample $i$ is correctly classified. Mathematically, this can be represented as:

$$
\max_{W} \sum_{i=1}^{N} P_i
$$

Where $N$ is the total number of samples, and $P_i$ is the probability of sample $i$ being correctly classified according to a stochastic nearest neighbors rule in the learned embedded space. This probability $P_i$ is defined as:

$$
P_i = \frac{\sum_{j \in S_i} e^{-\|Wx_i - Wx_j\|^2}}{\sum_{k=1}^{N} e^{-\|Wx_i - Wx_k\|^2}}
$$

Here, $S_i$ is the set of points in the same class as sample $i$, and $x_i$ represents the feature vector of sample $i$. The term $\|Wx_i - Wx_j\|^2$ denotes the squared Euclidean distance between the transformed feature vectors $Wx_i$ and $Wx_j$. The softmax function is applied over these distances to calculate the probability of correct classification.

#### Mahalanobis Distance

Neighborhood Component Analysis (NCA) can also be interpreted as learning a squared Mahalanobis distance metric. The Mahalanobis distance $d_M^2$ between two samples $i$ and $j$ is given by:

$$
d_M^2(i, j) = (Wx_i - Wx_j)^T M (Wx_i - Wx_j)
$$

Here, $M$ is a symmetric positive semi-definite matrix of size $(n_{\text{features}}, n_{\text{features}})$, which captures the covariance structure of the data. The matrix $M$ effectively scales and rotates the Euclidean space, allowing NCA to learn a more meaningful distance metric.

In summary, NCA aims to find an optimal linear transformation matrix $W$ that maximizes the sum of probabilities of correct classification, while also implicitly learning a Mahalanobis distance metric through the covariance matrix $M$.

##**"Distance" & "Uniform"**
In the context of the k-nearest neighbors (KNN) algorithm in scikit-learn (sklearn), "distance" and "uniform" refer to the weighting schemes used when aggregating the predictions of the nearest neighbors.

- **Distance Weighted (distance)**: In this scheme, the contribution of each neighbor to the prediction is weighted by the inverse of its distance from the query point. Mathematically, the weight assigned to each neighbor is $1 / d$, where $d$ is the distance between the query point and the neighbor. This means that closer neighbors have a greater influence on the prediction than farther ones. The predicted value is then computed as the weighted average of the targets of the nearest neighbors.

- **Uniform Weighted (uniform)**: In contrast, in the uniform scheme, all neighbors within the specified number (k) are given equal weight in the prediction. Regardless of the distance from the query point, each neighbor contributes equally to the prediction. This means that in the uniform scheme, the nearest neighbors have the same influence on the prediction, irrespective of their distances.

To summarize:

- **Distance weighting**: Closer neighbors have more influence on the prediction.
- **Uniform weighting**: All neighbors within the specified number have equal influence on the prediction.

In sklearn's KNN implementation, you can specify the weighting scheme using the `weights` parameter. For example:

```python
from sklearn.neighbors import KNeighborsClassifier

# Using distance weighting
knn_distance = KNeighborsClassifier(n_neighbors=5, weights='distance')

# Using uniform weighting
knn_uniform = KNeighborsClassifier(n_neighbors=5, weights='uniform')
```

This choice depends on the specific characteristics of your data and the problem at hand. In some cases, one weighting scheme may outperform the other, so it's often a good idea to experiment with both and see which works better for your particular dataset.

## **Decision Boundary for KNN**

In a classification problem with two or more classes, a decision boundary or decision surface is a hypersurface that partitions the underlying vector space into two or more sets, one for each class. The classifier will classify all the points on one side of the decision boundary as belonging to one class and all those on the other side as belonging to the other class.

### What is Decision Boundary?

A decision boundary separates different classes in a classification problem. It's the frontier or border in the feature space that helps determine which class a particular data point belongs to. In simpler terms, it's the line, curve, or surface that divides the input space into regions corresponding to different classes.

### Important Points

- **Applicability Across Algorithms**: Decision boundaries can be drawn for all classification algorithms, including neural networks. They are fundamental to understanding how classifiers make decisions.

- **Linearity vs. Non-Linearity**: Decision boundaries can be either linear (e.g., in the case of Support Vector Machines) or non-linear (e.g., in the case of Decision Tree classifiers or KNN). The nature of the data and the problem at hand determine the complexity of the decision boundary required.

- **Gradual Transitions**: Decision boundaries are not always sharply defined. In many cases, especially with fuzzy logic-based classification algorithms, the transition between classes in the feature space is gradual rather than abrupt. This ambiguity is crucial to consider when interpreting classifier outputs.

- **Hyperplane Representation**: In higher-dimensional problems, the decision boundary acts as a hyperplane for linear classifiers. This hyperplane is a generalization of a line (in 2D) or a plane (in 3D) to higher dimensions and serves the same purpose of separating different classes in the feature space.

Understanding decision boundaries is essential for evaluating classifier performance, interpreting model behavior, and making informed decisions in machine learning applications.



## **Voronoi Diagram**

In mathematics, a **Voronoi diagram** is a partitioning of a plane into regions based on distance to points in a specific subset of the plane.

### How it Works

Given a set of points in a plane, called **sites**, a Voronoi diagram divides the plane into regions, where each region consists of points that are closer to a particular site than to any other site. These regions are called **Voronoi cells** or **Voronoi polygons**.

### Construction

1. **Initial Setup**: Start with a set of points (sites) in the plane.
2. **Determine Regions**: For each site, determine the area of the plane that is closer to that site than to any other site. This creates a polygon around each site.
3. **Voronoi Cells**: The resulting polygons form the Voronoi cells or regions of the diagram.

### Applications

- **Computer Graphics**: Used in computer graphics for various purposes including texture generation, collision detection, and mesh generation.
- **Geography**: Voronoi diagrams are used to model territorial divisions, such as political boundaries or resource allocation.
- **Biology**: In biology, Voronoi diagrams can represent the territories of animals or the distribution of resources in ecosystems.
- **Robotics**: Used in robotics for path planning and navigation.

### Properties

- **Uniqueness**: Each point in the plane belongs to exactly one Voronoi cell.
- **Dual Graph**: The Voronoi diagram is closely related to the Delaunay triangulation, forming its dual graph.

### Conclusion

Voronoi diagrams are a powerful tool in mathematics and various fields for analyzing spatial relationships and partitioning space efficiently based on proximity to points.

## **When to use which type of average parameter?**
The `**''average''** parameter is required for multiclass/multilabel targets and determines the type of averaging performed on the data when calculating metrics like precision, recall, and F1-score. It can take the values **'micro', 'macro', 'samples', 'weighted', or None** (default=**'binary'**). Here's when to use each average parameter:

- **binary**: Use this when dealing with binary classification tasks. It only reports results for the class specified by `pos_label`, which is applicable only if targets (`y_true`, `y_pred`) are binary.

- **micro**: This calculates metrics globally by counting the total true positives, false negatives, and false positives. It's suitable for scenarios where you want to treat all classes equally and don't want to consider class imbalance.

- **macro**: Use this when you want to calculate metrics for each label and find their unweighted mean. It doesn't take label imbalance into account and treats each class equally.

- **weighted**: This calculates metrics for each label and finds their average weighted by support (the number of true instances for each label). It adjusts the 'macro' averaging to account for label imbalance. It's useful when there is a significant class imbalance, as it provides a more balanced view of the model's performance across different classes.

- **samples**: Use this when you want to calculate metrics for each instance and find their average. This is only meaningful for multilabel classification tasks where the average differs from the **'accuracy_score'**.

In summary:
- Use **'binary'** for binary classification.
- Use **'micro'** for equal treatment of all classes.
- Use **'macro'** for unweighted averaging across classes.
- Use **'weighted'** for weighted averaging considering class imbalance.
- Use **'samples'** for average per instance in multilabel classification.



### **1.6.4. Nearest Neighbor Algorithms**

#### 1.6.4.1. Brute Force

Fast computation of nearest neighbors is an active area of research in machine learning. The most naive neighbor search implementation involves the brute-force computation of distances between all pairs of points in the dataset: for *n* samples in *d* dimensions, this approach scales as *O(n^2d)*. Efficient brute-force neighbors searches can be very competitive for small data samples. However, as the number of samples *n* grows, the brute-force approach quickly becomes infeasible. In the classes within `sklearn.neighbors`, brute-force neighbors searches are specified using the keyword `algorithm = 'brute'`, and are computed using the routines available in `sklearn.metrics.pairwise`.

#### 1.6.4.2. K-D Tree

To address the computational inefficiencies of the brute-force approach, a variety of tree-based data structures have been invented. In general, these structures attempt to reduce the required number of distance calculations by efficiently encoding aggregate distance information for the sample. The basic idea is that if point *A* is very distant from point *B*, and point *C* is very close to point *D*, then we know that points *A* and *D* are very distant, without having to explicitly calculate their distance. In this way, the computational cost of a nearest neighbors search can be reduced to *O(n log n)* or better. This is a significant improvement over brute-force for large *n*.

An early approach to taking advantage of this aggregate information was the KD tree data structure (short for K-dimensional tree), which generalizes two-dimensional Quad-trees and 3-dimensional Oct-trees to an arbitrary number of dimensions. The KD tree is a binary tree structure which recursively partitions the parameter space along the data axes, dividing it into nested orthotropic regions into which data points are filed. The construction of a KD tree is very fast: because partitioning is performed only along the data axes, no *d*-dimensional distances need to be computed. Once constructed, the nearest neighbor of a query point can be determined with only *O(log n)* distance computations. Though the KD tree approach is very fast for low-dimensional (*d* < 20) neighbors searches, it becomes inefficient as *d* grows very large: this is one manifestation of the so-called “curse of dimensionality”. In scikit-learn, KD tree neighbors searches are specified using the keyword `algorithm = 'kd_tree'`, and are computed using the class `KDTree`.

#### 1.6.4.3. Ball Tree

To address the inefficiencies of KD Trees in higher dimensions, the ball tree data structure was developed. Where KD trees partition data along Cartesian axes, ball trees partition data in a series of nesting hyper-spheres. This makes tree construction more costly than that of the KD tree, but results in a data structure which can be very efficient on highly structured data, even in very high dimensions.

A ball tree recursively divides the data into nodes defined by a centroid *c* and radius *r*, such that each point in the node lies within the hyper-sphere defined by *c* and *r*. The number of candidate points for a neighbor search is reduced through use of the triangle inequality.

With this setup, a single distance calculation between a test point and the centroid is sufficient to determine a lower and upper bound on the distance to all points within the node. Because of the spherical geometry of the ball tree nodes, it can out-perform a KD-tree in high dimensions, though the actual performance is highly dependent on the structure of the training data. In scikit-learn, ball-tree-based neighbors searches are specified using the keyword `algorithm = 'ball_tree'`, and are computed using the class `BallTree`. Alternatively, the user can work with the `BallTree` class directly.

#### 1.6.4.4. Choice of Nearest Neighbors Algorithm

The optimal algorithm for a given dataset is a complicated choice, and depends on a number of factors:

- **Number of samples *n* (i.e. n_samples) and dimensionality *d* (i.e. n_features)**.
  - Brute force query time grows as *O(n^2d)*.
  - Ball tree query time grows as approximately *O(n log n)*.
  - KD tree query time changes with *d* in a way that is difficult to precisely characterize. For small *d* (less than 20 or so), the cost is approximately *O(n log n)*, and the KD tree query can be very efficient. For larger *d*, the cost increases to nearly *O(n^2)*, and the overhead due to the tree structure can lead to queries which are slower than brute force.
- **Data structure: intrinsic dimensionality of the data and/or sparsity of the data**.
  - Brute force query time is unchanged by data structure.
  - Ball tree and KD tree query times can be greatly influenced by data structure. In general, sparser data with a smaller intrinsic dimensionality leads to faster query times.
- **Number of neighbors *k* requested for a query point**.
  - Brute force query time is largely unaffected by the value of *k*.
  - Ball tree and KD tree query time will become slower as *k* increases.
- **Number of query points**.
  - Both the ball tree and the KD Tree require a construction phase. If only a small number of queries will be performed, however, the construction can make up a significant fraction of the total cost. If very few query points will be required, brute force is better than a tree-based method.

Otherwise, it selects the first out of `'kd_tree'` and `'ball_tree'` that has `effective_metric_` in its `VALID_METRICS` list. This heuristic is based on the following assumptions:
